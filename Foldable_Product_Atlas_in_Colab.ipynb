```python
"""
Build and maintain a curated starter catalog for a Foldable Product Atlas.
"""

import subprocess
import sys
from pathlib import Path
from typing import Any, Final

# =============================================================================
# CONTROL KNOBS
# =============================================================================

INSTALL_PACKAGES: Final[bool] = True

REPO_NAME: Final[str] = "foldable-product-atlas"
REPO_ROOT: Final[Path] = Path("/content") / REPO_NAME
DATA_ROOT: Final[Path] = REPO_ROOT / "data"

DISCOVERY_CANDIDATES_PATH: Final[Path] = DATA_ROOT / "discovery_candidates.csv"
DISCOVERY_ENRICHED_PATH: Final[Path] = (
    DATA_ROOT / "discovery_candidates_enriched.csv"
)
DISCOVERY_BALANCED_PATH: Final[Path] = (
    DATA_ROOT / "discovery_candidates_balanced.csv"
)
DISCOVERY_QUERY_AUDIT_PATH: Final[Path] = DATA_ROOT / "discovery_query_audit.csv"
RAW_VS_BALANCED_PATH: Final[Path] = DATA_ROOT / "raw_vs_balanced_summary.csv"

PRODUCTS_SEED_PATH: Final[Path] = DATA_ROOT / "products_seed.csv"
REVIEW_QUEUE_PATH: Final[Path] = DATA_ROOT / "review_queue.csv"
COVERAGE_GAPS_PATH: Final[Path] = DATA_ROOT / "coverage_gaps.csv"

MPL_DPI: Final[int] = 250

PURE_BLUE: Final[str] = "#0000FF"
AXIS_BLACK: Final[str] = "#000000"
GRID_ALPHA: Final[float] = 0.18
BAR_EDGE_WIDTH: Final[float] = 0.6

TOP_N_CATEGORIES: Final[int] = 24
TOP_N_DOMAINS: Final[int] = 24
TOP_N_PRODUCTS: Final[int] = 24
HISTOGRAM_BINS: Final[int] = 18

DISPLAY_TABLES: Final[bool] = True
DISPLAY_MAX_ROWS: Final[int] = 80
DISPLAY_MAX_COL_WIDTH: Final[int] = 180

# =============================================================================
# DEPENDENCIES
# =============================================================================


def install_dependencies() -> None:
    """Install plotting and dataframe dependencies with uv pip."""
    subprocess.run(
        [sys.executable, "-m", "pip", "install", "-q", "uv"],
        check=True,
    )
    subprocess.run(
        [
            "uv",
            "pip",
            "install",
            "-q",
            "--system",
            "matplotlib",
            "numpy",
            "pandas",
            "tabulate",
        ],
        check=True,
    )


if INSTALL_PACKAGES:
    install_dependencies()

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import HTML, Markdown, display

mpl.rcParams["figure.dpi"] = MPL_DPI
pd.set_option("display.max_rows", DISPLAY_MAX_ROWS)
pd.set_option("display.max_columns", 160)
pd.set_option("display.max_colwidth", DISPLAY_MAX_COL_WIDTH)
pd.set_option("display.width", 220)

# =============================================================================
# DISPLAY HELPERS
# =============================================================================

DISPLAY_CACHE: list[pd.DataFrame] = []


def clean_text(value: Any, max_chars: int | None = None) -> str:
    """Normalize text to one line and optionally truncate it."""
    if value is None:
        return ""

    try:
        if pd.isna(value):
            return ""
    except (TypeError, ValueError):
        pass

    text = " ".join(str(value).split())

    if max_chars is not None and len(text) > max_chars:
        return text[: max_chars - 1].rstrip() + "…"

    return text


def configure_colab_dataframe_display() -> None:
    """Disable Colab interactive dataframe rendering when available."""
    try:
        from google.colab import data_table

        data_table.disable_dataframe_formatter()
    except Exception:
        pass


def prepare_display_dataframe(
    df: pd.DataFrame,
    max_rows: int | None = DISPLAY_MAX_ROWS,
) -> pd.DataFrame:
    """Build a stable dataframe copy for HTML display."""
    if max_rows is None:
        table = df.copy()
    else:
        table = df.head(max_rows).copy()

    for column in table.columns:
        if table[column].dtype == "object":
            table[column] = table[column].map(
                lambda value: clean_text(value, DISPLAY_MAX_COL_WIDTH)
            )

    DISPLAY_CACHE.append(table)
    return table


def display_table(
    title: str,
    df: pd.DataFrame,
    max_rows: int | None = DISPLAY_MAX_ROWS,
) -> None:
    """Display a dataframe as static HTML."""
    if DISPLAY_TABLES is False:
        return

    table = prepare_display_dataframe(df, max_rows=max_rows)
    display(Markdown(f"### {title}"))

    html = table.to_html(index=False, escape=True)
    wrapped_html = (
        '<div style="max-height: 520px; overflow: auto; '
        'border: 1px solid #444; padding: 8px;">'
        f"{html}"
        "</div>"
    )
    display(HTML(wrapped_html))


configure_colab_dataframe_display()

# =============================================================================
# LOADERS
# =============================================================================


def read_csv_if_exists(path: Path) -> pd.DataFrame | None:
    """Read a CSV file if it exists, otherwise return None."""
    if path.exists() is False:
        print(f"Missing optional file: {path}")
        return None

    return pd.read_csv(path)


def require_nonempty_dataframe(
    df: pd.DataFrame | None,
    path: Path,
) -> pd.DataFrame:
    """Require that a dataframe exists and has rows."""
    if df is None:
        raise FileNotFoundError(f"Required CSV was not found: {path}")

    if len(df) == 0:
        raise ValueError(f"Required CSV is empty: {path}")

    return df


def load_available_outputs() -> dict[str, pd.DataFrame]:
    """Load all available Foldable Product Atlas CSV outputs."""
    outputs: dict[str, pd.DataFrame] = {}

    paths = {
        "discovery_candidates": DISCOVERY_CANDIDATES_PATH,
        "discovery_enriched": DISCOVERY_ENRICHED_PATH,
        "discovery_balanced": DISCOVERY_BALANCED_PATH,
        "discovery_query_audit": DISCOVERY_QUERY_AUDIT_PATH,
        "raw_vs_balanced": RAW_VS_BALANCED_PATH,
        "products_seed": PRODUCTS_SEED_PATH,
        "review_queue": REVIEW_QUEUE_PATH,
        "coverage_gaps": COVERAGE_GAPS_PATH,
    }

    for name, path in paths.items():
        df = read_csv_if_exists(path)
        if df is not None and len(df) > 0:
            outputs[name] = df
            print(f"Loaded {name}: {len(df)} rows from {path}")

    if len(outputs) == 0:
        raise FileNotFoundError(
            "No recognized CSV outputs were found under "
            f"{DATA_ROOT}. Run the catalog or discovery notebook first."
        )

    return outputs


# =============================================================================
# PLOT STYLE
# =============================================================================


def apply_pure_blue_axes(ax: plt.Axes) -> None:
    """Apply pure-blue-compatible high-contrast axes styling."""
    ax.tick_params(axis="both", colors=AXIS_BLACK)
    ax.xaxis.label.set_color(AXIS_BLACK)
    ax.yaxis.label.set_color(AXIS_BLACK)
    ax.title.set_color(AXIS_BLACK)

    for spine in ax.spines.values():
        spine.set_color(AXIS_BLACK)
        spine.set_linewidth(1.0)


def draw_barh(
    series: pd.Series,
    title: str,
    xlabel: str,
    ylabel: str,
    figsize: tuple[float, float] = (10.8, 7.0),
) -> None:
    """Render a horizontal bar chart in pure blue."""
    plot_series = series.dropna()

    if len(plot_series) == 0:
        print(f"Skipping empty plot: {title}")
        return

    fig, ax = plt.subplots(figsize=figsize)
    plot_series.sort_values().plot(
        kind="barh",
        ax=ax,
        color=PURE_BLUE,
        edgecolor=AXIS_BLACK,
        linewidth=BAR_EDGE_WIDTH,
    )

    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.grid(axis="x", alpha=GRID_ALPHA, color=AXIS_BLACK)
    apply_pure_blue_axes(ax)

    fig.tight_layout()
    plt.show()


def draw_histogram(
    values: pd.Series,
    title: str,
    xlabel: str,
    ylabel: str,
    bins: int = HISTOGRAM_BINS,
    figsize: tuple[float, float] = (8.6, 4.8),
) -> None:
    """Render a histogram in pure blue."""
    numeric_values = pd.to_numeric(values, errors="coerce").dropna()

    if len(numeric_values) == 0:
        print(f"Skipping empty histogram: {title}")
        return

    fig, ax = plt.subplots(figsize=figsize)
    ax.hist(
        numeric_values,
        bins=bins,
        color=PURE_BLUE,
        edgecolor=AXIS_BLACK,
        linewidth=BAR_EDGE_WIDTH,
    )

    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.grid(axis="y", alpha=GRID_ALPHA, color=AXIS_BLACK)
    apply_pure_blue_axes(ax)

    fig.tight_layout()
    plt.show()


# =============================================================================
# DISCOVERY PLOTS
# =============================================================================


def plot_discovery_family_counts(
    df: pd.DataFrame,
    title: str,
) -> None:
    """Plot discovery candidate counts by transformation family."""
    if "transformation_family" not in df.columns:
        print(f"Skipping {title}: transformation_family column missing")
        return

    counts = df["transformation_family"].fillna("unknown").astype(str).value_counts()
    draw_barh(
        counts,
        title=title,
        xlabel="Candidate count",
        ylabel="Transformation family",
        figsize=(9.5, 5.4),
    )


def plot_discovery_category_counts(
    df: pd.DataFrame,
    title: str,
    top_n: int = TOP_N_CATEGORIES,
) -> None:
    """Plot discovery candidate counts by category."""
    if "category_hint" not in df.columns:
        print(f"Skipping {title}: category_hint column missing")
        return

    counts = df["category_hint"].fillna("unknown").value_counts().head(top_n)
    draw_barh(
        counts,
        title=title,
        xlabel="Candidate count",
        ylabel="Category",
        figsize=(10.8, 7.0),
    )


def plot_discovery_domain_counts(
    df: pd.DataFrame,
    title: str,
    top_n: int = TOP_N_DOMAINS,
) -> None:
    """Plot discovery candidate counts by source domain."""
    if "domain" not in df.columns:
        print(f"Skipping {title}: domain column missing")
        return

    counts = df["domain"].fillna("unknown").value_counts().head(top_n)
    draw_barh(
        counts,
        title=title,
        xlabel="Candidate count",
        ylabel="Domain",
        figsize=(10.4, 6.4),
    )


def plot_discovery_priority_distribution(
    df: pd.DataFrame,
    title: str,
) -> None:
    """Plot discovery source-priority score distribution."""
    if "source_priority_score" not in df.columns:
        print(f"Skipping {title}: source_priority_score column missing")
        return

    draw_histogram(
        df["source_priority_score"],
        title=title,
        xlabel="Source priority score",
        ylabel="Candidate count",
    )


def plot_candidates_per_query(
    audit_df: pd.DataFrame,
    top_n: int = TOP_N_CATEGORIES,
) -> None:
    """Plot candidate yield normalized by query count."""
    required_columns = {"category_hint", "candidates_per_query"}
    if required_columns.issubset(audit_df.columns) is False:
        print("Skipping query-yield plot: required audit columns missing")
        return

    plot_df = audit_df.sort_values(
        "candidates_per_query",
        ascending=False,
        kind="stable",
    ).head(top_n)

    series = plot_df.set_index("category_hint")["candidates_per_query"]
    draw_barh(
        series,
        title="Candidate Yield Normalized by Query Count",
        xlabel="Raw candidates per query",
        ylabel="Category",
        figsize=(10.8, 7.0),
    )


def plot_raw_vs_balanced_summary(summary_df: pd.DataFrame) -> None:
    """Plot raw versus balanced candidate counts from summary table."""
    required_columns = {
        "category_hint",
        "raw_candidate_count",
        "balanced_candidate_count",
    }
    if required_columns.issubset(summary_df.columns) is False:
        print("Skipping raw-vs-balanced plot: required summary columns missing")
        return

    plot_df = summary_df.head(TOP_N_CATEGORIES).copy()
    plot_df = plot_df.set_index("category_hint")

    fig, ax = plt.subplots(figsize=(11.5, 7.2))

    y_positions = np.arange(len(plot_df))
    bar_height = 0.38

    ax.barh(
        y_positions - bar_height / 2,
        plot_df["raw_candidate_count"],
        height=bar_height,
        color=PURE_BLUE,
        edgecolor=AXIS_BLACK,
        linewidth=BAR_EDGE_WIDTH,
        alpha=0.45,
        label="Raw",
    )
    ax.barh(
        y_positions + bar_height / 2,
        plot_df["balanced_candidate_count"],
        height=bar_height,
        color=PURE_BLUE,
        edgecolor=AXIS_BLACK,
        linewidth=BAR_EDGE_WIDTH,
        alpha=1.0,
        label="Balanced",
    )

    ax.set_yticks(y_positions)
    ax.set_yticklabels(plot_df.index)
    ax.set_title("Raw Versus Balanced Candidate Counts")
    ax.set_xlabel("Candidate count")
    ax.set_ylabel("Category")
    ax.grid(axis="x", alpha=GRID_ALPHA, color=AXIS_BLACK)
    ax.legend(frameon=False)

    apply_pure_blue_axes(ax)
    fig.tight_layout()
    plt.show()


# =============================================================================
# CURATED CATALOG PLOTS
# =============================================================================


def plot_catalog_family_counts(
    df: pd.DataFrame,
    title: str = "Curated Catalog Count by Transformation Family",
) -> None:
    """Plot curated catalog counts by transformation family."""
    if "transformation_family" not in df.columns:
        print(f"Skipping {title}: transformation_family column missing")
        return

    counts = df["transformation_family"].fillna("unknown").astype(str).value_counts()
    draw_barh(
        counts,
        title=title,
        xlabel="Product count",
        ylabel="Transformation family",
        figsize=(9.5, 5.4),
    )


def plot_catalog_market_categories(
    df: pd.DataFrame,
    title: str = "Curated Catalog Count by Market Category",
    top_n: int = TOP_N_CATEGORIES,
) -> None:
    """Plot curated catalog counts by market category."""
    if "market_category" not in df.columns:
        print(f"Skipping {title}: market_category column missing")
        return

    counts = df["market_category"].fillna("unknown").value_counts().head(top_n)
    draw_barh(
        counts,
        title=title,
        xlabel="Product count",
        ylabel="Market category",
        figsize=(10.8, 7.0),
    )


def plot_catalog_review_status(
    df: pd.DataFrame,
    title: str = "Curated Catalog Count by Review Status",
) -> None:
    """Plot curated catalog counts by review status."""
    if "review_status" not in df.columns:
        print(f"Skipping {title}: review_status column missing")
        return

    counts = df["review_status"].fillna("unknown").value_counts()
    draw_barh(
        counts,
        title=title,
        xlabel="Product count",
        ylabel="Review status",
        figsize=(9.2, 4.8),
    )


def plot_catalog_source_domains(
    df: pd.DataFrame,
    title: str = "Curated Catalog Source Domains",
    top_n: int = TOP_N_DOMAINS,
) -> None:
    """Plot curated catalog source-domain counts."""
    domain_column = "source_domain"
    if domain_column not in df.columns:
        if "source_url" not in df.columns:
            print(f"Skipping {title}: source_domain and source_url columns missing")
            return

        domain_series = df["source_url"].map(extract_domain_from_url)
    else:
        domain_series = df[domain_column]

    counts = domain_series.fillna("unknown").value_counts().head(top_n)
    draw_barh(
        counts,
        title=title,
        xlabel="Product count",
        ylabel="Source domain",
        figsize=(10.4, 6.4),
    )


def plot_foldability_score_distribution(
    df: pd.DataFrame,
    title: str = "Foldability Score Distribution",
) -> None:
    """Plot foldability-score distribution."""
    if "foldability_score" not in df.columns:
        print(f"Skipping {title}: foldability_score column missing")
        return

    draw_histogram(
        df["foldability_score"],
        title=title,
        xlabel="Foldability score",
        ylabel="Product count",
        bins=12,
    )


def plot_top_foldability_scores(
    df: pd.DataFrame,
    title: str = "Top Foldability Scores",
    top_n: int = TOP_N_PRODUCTS,
) -> None:
    """Plot top product foldability scores."""
    required_columns = {"brand", "product_name", "foldability_score"}
    if required_columns.issubset(df.columns) is False:
        print(f"Skipping {title}: required product-score columns missing")
        return

    plot_df = df.copy()
    plot_df["product_label"] = (
        plot_df["brand"].fillna("").astype(str)
        + " "
        + plot_df["product_name"].fillna("").astype(str)
    )
    plot_df["foldability_score"] = pd.to_numeric(
        plot_df["foldability_score"],
        errors="coerce",
    )

    plot_df = (
        plot_df.dropna(subset=["foldability_score"])
        .sort_values("foldability_score", ascending=False, kind="stable")
        .head(top_n)
    )

    series = plot_df.set_index("product_label")["foldability_score"]
    draw_barh(
        series,
        title=title,
        xlabel="Foldability score",
        ylabel="Product",
        figsize=(11.0, 8.0),
    )


def plot_coverage_gaps(
    coverage_df: pd.DataFrame,
    title: str = "Catalog Coverage by Target Category",
) -> None:
    """Plot covered versus uncovered target categories."""
    required_columns = {"target_market_category", "covered"}
    if required_columns.issubset(coverage_df.columns) is False:
        print(f"Skipping {title}: required coverage columns missing")
        return

    plot_df = coverage_df.copy()
    plot_df["coverage_numeric"] = plot_df["covered"].astype(bool).astype(int)
    series = plot_df.set_index("target_market_category")["coverage_numeric"]

    draw_barh(
        series,
        title=title,
        xlabel="Covered flag, 1 means covered",
        ylabel="Target market category",
        figsize=(10.8, 7.0),
    )


# =============================================================================
# SMALL URL HELPER
# =============================================================================


def extract_domain_from_url(url: Any) -> str:
    """Extract a normalized source domain from a URL-like value."""
    text = clean_text(url)
    if text == "":
        return "unknown"

    try:
        from urllib.parse import urlparse

        parsed = urlparse(text)
        return parsed.netloc.lower().removeprefix("www.") or "unknown"
    except Exception:
        return "unknown"


# =============================================================================
# SUMMARY TABLES
# =============================================================================


def build_loaded_file_summary(outputs: dict[str, pd.DataFrame]) -> pd.DataFrame:
    """Build a summary of loaded CSV files."""
    rows = []

    for name, df in outputs.items():
        rows.append(
            {
                "table_name": name,
                "row_count": len(df),
                "column_count": len(df.columns),
                "columns": ", ".join(df.columns.astype(str).tolist()),
            }
        )

    return pd.DataFrame(rows)


def build_discovery_plot_summary(df: pd.DataFrame) -> pd.DataFrame:
    """Build a compact summary of discovery data."""
    rows = []

    if "transformation_family" in df.columns:
        rows.append(
            {
                "metric": "transformation_family_count",
                "value": df["transformation_family"].nunique(dropna=True),
            }
        )

    if "category_hint" in df.columns:
        rows.append(
            {
                "metric": "category_hint_count",
                "value": df["category_hint"].nunique(dropna=True),
            }
        )

    if "domain" in df.columns:
        rows.append(
            {
                "metric": "source_domain_count",
                "value": df["domain"].nunique(dropna=True),
            }
        )

    rows.append({"metric": "row_count", "value": len(df)})

    return pd.DataFrame(rows)


def build_catalog_plot_summary(df: pd.DataFrame) -> pd.DataFrame:
    """Build a compact summary of curated catalog data."""
    rows = []

    for column in [
        "market_category",
        "transformation_family",
        "review_status",
        "source_domain",
    ]:
        if column in df.columns:
            rows.append(
                {
                    "metric": f"{column}_count",
                    "value": df[column].nunique(dropna=True),
                }
            )

    rows.append({"metric": "row_count", "value": len(df)})

    return pd.DataFrame(rows)


# =============================================================================
# MAIN
# =============================================================================


def render_discovery_plots(outputs: dict[str, pd.DataFrame]) -> None:
    """Render pure-blue plots for discovery outputs."""
    if "discovery_candidates" in outputs:
        raw_df = outputs["discovery_candidates"]

        display(Markdown("## Raw Discovery Plots"))
        display_table(
            "Raw discovery summary",
            build_discovery_plot_summary(raw_df),
            max_rows=None,
        )

        plot_discovery_category_counts(
            raw_df,
            "Raw Discovery Categories",
        )
        plot_discovery_family_counts(
            raw_df,
            "Raw Transformation-Family Counts",
        )
        plot_discovery_priority_distribution(
            raw_df,
            "Raw Candidate Source Priority Distribution",
        )
        plot_discovery_domain_counts(
            raw_df,
            "Raw Top Discovery Domains",
        )

    if "discovery_balanced" in outputs:
        balanced_df = outputs["discovery_balanced"]

        display(Markdown("## Balanced Discovery Plots"))
        display_table(
            "Balanced discovery summary",
            build_discovery_plot_summary(balanced_df),
            max_rows=None,
        )

        plot_discovery_category_counts(
            balanced_df,
            "Balanced Discovery Categories",
        )
        plot_discovery_family_counts(
            balanced_df,
            "Balanced Transformation-Family Counts",
        )
        plot_discovery_priority_distribution(
            balanced_df,
            "Balanced Candidate Source Priority Distribution",
        )
        plot_discovery_domain_counts(
            balanced_df,
            "Balanced Top Discovery Domains",
        )

    if "discovery_query_audit" in outputs:
        plot_candidates_per_query(outputs["discovery_query_audit"])

    if "raw_vs_balanced" in outputs:
        plot_raw_vs_balanced_summary(outputs["raw_vs_balanced"])


def render_catalog_plots(outputs: dict[str, pd.DataFrame]) -> None:
    """Render pure-blue plots for curated catalog outputs."""
    if "products_seed" not in outputs:
        return

    catalog_df = outputs["products_seed"]

    display(Markdown("## Curated Catalog Plots"))
    display_table(
        "Curated catalog summary",
        build_catalog_plot_summary(catalog_df),
        max_rows=None,
    )

    plot_catalog_market_categories(catalog_df)
    plot_catalog_family_counts(catalog_df)
    plot_catalog_review_status(catalog_df)
    plot_catalog_source_domains(catalog_df)
    plot_foldability_score_distribution(catalog_df)
    plot_top_foldability_scores(catalog_df)

    if "coverage_gaps" in outputs:
        plot_coverage_gaps(outputs["coverage_gaps"])


def main() -> dict[str, pd.DataFrame]:
    """Load available outputs and render all plots in pure blue."""
    outputs = load_available_outputs()

    display_table(
        "Loaded Foldable Product Atlas CSV outputs",
        build_loaded_file_summary(outputs),
        max_rows=None,
    )

    render_discovery_plots(outputs)
    render_catalog_plots(outputs)

    print("Pure-blue plotting workflow complete.")
    return outputs


plot_outputs = main()
```

```python
"""
Estimate foldable and collapsible product market volume from public web data.
"""

import hashlib
import json
import math
import re
import subprocess
import sys
import time
import urllib.parse
import urllib.robotparser
from collections import defaultdict
from dataclasses import dataclass
from difflib import SequenceMatcher
from pathlib import Path
from typing import Any, Final

# =============================================================================
# CONTROL KNOBS
# =============================================================================

INSTALL_PACKAGES: Final[bool] = True

REPO_NAME: Final[str] = "foldable-product-atlas"
REPO_ROOT: Final[Path] = Path("/content") / REPO_NAME

OUTPUT_DIR: Final[Path] = REPO_ROOT / "data"
QUERY_LOG_CSV_PATH: Final[Path] = OUTPUT_DIR / "market_volume_queries.csv"
SEARCH_RESULTS_CSV_PATH: Final[Path] = OUTPUT_DIR / "market_volume_search_results.csv"
SCRAPED_PAGES_CSV_PATH: Final[Path] = OUTPUT_DIR / "market_volume_scraped_pages.csv"
PRODUCT_RECORDS_CSV_PATH: Final[Path] = OUTPUT_DIR / "market_volume_product_records.csv"
PRODUCT_MODELS_CSV_PATH: Final[Path] = OUTPUT_DIR / "market_volume_product_models.csv"
MARKET_VOLUME_CSV_PATH: Final[Path] = OUTPUT_DIR / "market_volume_summary.csv"

EXPORT_FILES: Final[bool] = True
MPL_DPI: Final[int] = 250
PURE_BLUE: Final[str] = "#0000FF"

SEARCH_REGION: Final[str] = "us-en"
SAFESEARCH: Final[str] = "moderate"
SEARCH_TIMEOUT_S: Final[float] = 12.0
SEARCH_DELAY_S: Final[float] = 0.35
RESULTS_PER_QUERY: Final[int] = 12

MAX_SEARCH_TERMS_PER_CATEGORY: Final[int] = 3
MAX_GENERAL_QUERIES_PER_CATEGORY: Final[int] = 4
MAX_SITE_QUERIES_PER_CATEGORY: Final[int] = 6
MAX_QUERIES_TOTAL: Final[int | None] = None

SCRAPE_PAGES: Final[bool] = True
RUN_ROBOTS_CHECK: Final[bool] = True
ALLOW_WHEN_ROBOTS_UNAVAILABLE: Final[bool] = True
REQUEST_CONNECT_TIMEOUT_S: Final[float] = 4.0
REQUEST_READ_TIMEOUT_S: Final[float] = 7.0
ROBOTS_CONNECT_TIMEOUT_S: Final[float] = 3.0
ROBOTS_READ_TIMEOUT_S: Final[float] = 4.0
SCRAPE_DELAY_S: Final[float] = 0.12
RETRY_TOTAL: Final[int] = 1
RETRY_BACKOFF_FACTOR: Final[float] = 0.35
MAX_HTML_BYTES: Final[int] = 1_800_000

MAX_TOTAL_PAGES_TO_SCRAPE: Final[int] = 1_000
MAX_SCRAPED_PAGES_PER_CATEGORY: Final[int] = 90
MAX_SCRAPED_PAGES_PER_DOMAIN_CATEGORY: Final[int] = 14
MAX_DISCOVERED_LINKS_PER_PAGE: Final[int] = 18
MAX_DISCOVERED_LINKS_PER_DOMAIN_CATEGORY: Final[int] = 35

MIN_PRODUCT_CONFIDENCE_FOR_VOLUME: Final[float] = 0.55
USE_LOW_CONFIDENCE_CANDIDATES_IN_VOLUME: Final[bool] = False
MIN_NAME_TOKEN_COUNT_FOR_VOLUME: Final[int] = 2

DEDUP_SEQUENCE_THRESHOLD: Final[float] = 0.92
DEDUP_JACCARD_THRESHOLD: Final[float] = 0.78

DISPLAY_MAX_ROWS: Final[int] = 120
DISPLAY_MAX_COL_WIDTH: Final[int] = 180
TOP_N_PLOT: Final[int] = 28

USER_AGENT: Final[str] = (
    "foldable-product-atlas-market-volume-bot/1.0 "
    "(public product-model counting research; respectful bounded crawling)"
)

# =============================================================================
# DEPENDENCY INSTALLATION
# =============================================================================


def install_dependencies() -> None:
    """Install scraping, search, table, and plotting dependencies."""
    subprocess.run(
        [sys.executable, "-m", "pip", "install", "-q", "uv"],
        check=True,
    )
    subprocess.run(
        [
            sys.executable,
            "-m",
            "uv",
            "pip",
            "install",
            "-q",
            "--system",
            "beautifulsoup4",
            "ddgs",
            "lxml",
            "matplotlib",
            "numpy",
            "pandas",
            "requests",
            "tabulate",
        ],
        check=True,
    )


if INSTALL_PACKAGES:
    install_dependencies()

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from ddgs import DDGS
from IPython.display import HTML, Markdown, display
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

mpl.rcParams["figure.dpi"] = MPL_DPI
mpl.rcParams["savefig.dpi"] = MPL_DPI
mpl.rcParams["axes.edgecolor"] = "black"
mpl.rcParams["axes.labelcolor"] = "black"
mpl.rcParams["xtick.color"] = "black"
mpl.rcParams["ytick.color"] = "black"
mpl.rcParams["text.color"] = "black"

pd.set_option("display.max_rows", DISPLAY_MAX_ROWS)
pd.set_option("display.max_columns", 180)
pd.set_option("display.max_colwidth", DISPLAY_MAX_COL_WIDTH)
pd.set_option("display.width", 240)

# =============================================================================
# DATA MODELS
# =============================================================================


@dataclass(frozen=True)
class MarketCategory:
    """One market category whose product-model volume will be estimated."""

    family_hint: str
    category_hint: str
    search_terms: tuple[str, ...]
    category_markers: tuple[str, ...]
    transformation_markers: tuple[str, ...]
    official_domain_fragments: tuple[str, ...] = ()


@dataclass(frozen=True)
class FetchResult:
    """Bounded HTTP fetch result."""

    url: str
    final_url: str
    status_code: int | None
    html: str
    ok: bool
    message: str


# =============================================================================
# MARKET CATEGORY REGISTRY
# =============================================================================

CATEGORIES: Final[tuple[MarketCategory, ...]] = (
    MarketCategory(
        "foldable",
        "computer mouse",
        ("foldable computer mouse", "folding computer mouse", "arc mouse"),
        ("mouse", "computer mouse"),
        ("foldable", "folding", "snap flat", "arc mouse", "bendable"),
        ("microsoft.com",),
    ),
    MarketCategory(
        "foldable",
        "bluetooth keyboard",
        (
            "foldable bluetooth keyboard",
            "folding bluetooth keyboard",
            "foldable travel keyboard",
        ),
        ("keyboard", "bluetooth keyboard", "travel keyboard"),
        ("foldable", "folding", "tri-fold", "bi-fold", "portable"),
        (),
    ),
    MarketCategory(
        "foldable",
        "foldable headphones",
        (
            "foldable headphones",
            "folding headphones",
            "fold-flat headphones",
        ),
        ("headphones", "headphone", "headset"),
        ("foldable", "folding", "fold-flat", "fold flat", "swivel", "hinge"),
        (
            "sony.com",
            "bose.com",
            "soundcore.com",
            "sennheiser-hearing.com",
            "jbl.com",
            "audio-technica.com",
            "beatsbydre.com",
            "skullcandy.com",
            "marshallheadphones.com",
        ),
    ),
    MarketCategory(
        "foldable",
        "folding bicycle",
        (
            "folding bicycle",
            "foldable bicycle",
            "full size folding bike",
        ),
        ("bicycle", "bike", "cycling"),
        ("folding", "foldable", "folded size", "folding frame", "hinge"),
        (
            "montaguebikes.com",
            "brompton.com",
            "ternbicycles.com",
            "dahon.com",
            "bikefriday.com",
            "birdybicycle.com",
            "vello.bike",
            "helix.ca",
            "zizzo.bike",
            "citizenbike.com",
            "strida.com",
            "hummingbirdbike.com",
            "decathlon.com",
        ),
    ),
    MarketCategory(
        "foldable",
        "folding e-bike",
        (
            "folding electric bike",
            "foldable e-bike",
            "folding ebike",
        ),
        ("e-bike", "ebike", "electric bike", "electric bicycle"),
        ("folding", "foldable", "folded size", "folding frame", "hinge"),
        (
            "gocycle.com",
            "lectricebikes.com",
            "radpowerbikes.com",
            "ternbicycles.com",
            "dahon.com",
        ),
    ),
    MarketCategory(
        "foldable",
        "bicycle helmet",
        (
            "foldable bicycle helmet",
            "folding bicycle helmet",
            "collapsible bike helmet",
        ),
        ("helmet", "bike helmet", "bicycle helmet"),
        ("foldable", "folding", "collapsible", "segmented", "nested"),
        (),
    ),
    MarketCategory(
        "foldable",
        "kayak",
        ("folding kayak", "foldable kayak", "origami kayak"),
        ("kayak",),
        ("folding", "foldable", "origami", "folded", "packable"),
        ("orukayak.com",),
    ),
    MarketCategory(
        "foldable",
        "canoe",
        ("folding canoe", "foldable canoe", "origami canoe"),
        ("canoe",),
        ("folding", "foldable", "origami", "folded", "packable"),
        ("oricanoe.com",),
    ),
    MarketCategory(
        "foldable",
        "paddle board",
        ("foldable paddle board", "folding paddle board", "compact paddle board"),
        ("paddle board", "sup", "stand up paddle"),
        ("foldable", "folding", "compact", "portable"),
        (),
    ),
    MarketCategory(
        "foldable",
        "solar panel",
        ("foldable solar panel", "folding solar panel", "portable solar panel"),
        ("solar panel", "solar charger"),
        ("foldable", "folding", "portable", "briefcase"),
        (),
    ),
    MarketCategory(
        "foldable",
        "display",
        ("foldable display", "folding display", "flexible oled display"),
        ("display", "screen", "oled"),
        ("foldable", "folding", "flexible", "rollable"),
        ("samsung.com", "lg.com", "lenovo.com", "asus.com"),
    ),
    MarketCategory(
        "foldable",
        "OLED laptop",
        ("foldable OLED laptop", "folding OLED laptop", "dual screen laptop"),
        ("laptop", "notebook", "oled"),
        ("foldable", "folding", "dual screen", "flexible oled"),
        ("lenovo.com", "asus.com", "hp.com"),
    ),
    MarketCategory(
        "foldable",
        "smartphone",
        ("foldable smartphone", "folding smartphone", "foldable phone"),
        ("smartphone", "phone"),
        ("foldable", "folding", "flip", "book-style", "hinge"),
        ("samsung.com", "motorola.com", "oneplus.com", "google.com"),
    ),
    MarketCategory(
        "foldable",
        "monitor",
        ("foldable monitor", "folding monitor", "portable folding monitor"),
        ("monitor", "display", "screen"),
        ("foldable", "folding", "portable", "dual screen"),
        (),
    ),
    MarketCategory(
        "foldable",
        "drone",
        ("folding drone", "foldable drone", "portable folding drone"),
        ("drone", "quadcopter"),
        ("folding", "foldable", "foldable arms", "folded"),
        ("dji.com", "autelrobotics.com"),
    ),
    MarketCategory(
        "foldable",
        "stroller",
        ("foldable stroller", "folding stroller", "compact stroller"),
        ("stroller",),
        ("foldable", "folding", "compact fold", "one-hand fold"),
        (),
    ),
    MarketCategory(
        "foldable",
        "wagon",
        ("folding wagon", "foldable wagon", "collapsible wagon"),
        ("wagon", "cart"),
        ("folding", "foldable", "collapsible"),
        (),
    ),
    MarketCategory(
        "foldable",
        "camping chair",
        ("folding camping chair", "foldable camping chair", "compact camp chair"),
        ("chair", "camp chair", "camping chair"),
        ("folding", "foldable", "collapsible", "packable"),
        (),
    ),
    MarketCategory(
        "foldable",
        "camping table",
        ("folding camping table", "foldable camping table", "roll top table"),
        ("table", "camping table"),
        ("folding", "foldable", "collapsible", "roll top"),
        (),
    ),
    MarketCategory(
        "foldable",
        "treadmill",
        ("folding treadmill", "foldable treadmill", "compact treadmill"),
        ("treadmill",),
        ("folding", "foldable", "folded"),
        (),
    ),
    MarketCategory(
        "foldable",
        "wheelchair",
        ("folding wheelchair", "foldable wheelchair", "portable wheelchair"),
        ("wheelchair",),
        ("folding", "foldable", "collapsible"),
        (),
    ),
    MarketCategory(
        "foldable",
        "paper microscope",
        ("foldable paper microscope", "foldscope", "paper microscope"),
        ("microscope", "foldscope"),
        ("foldable", "folding", "paper", "origami"),
        ("foldscope.com",),
    ),
    MarketCategory(
        "collapsible",
        "water bottle",
        ("collapsible water bottle", "foldable water bottle", "soft water bottle"),
        ("water bottle", "bottle"),
        ("collapsible", "foldable", "soft", "squeeze", "roll up"),
        (),
    ),
    MarketCategory(
        "collapsible",
        "soft flask",
        ("collapsible soft flask", "soft flask", "hydration soft flask"),
        ("soft flask", "hydration flask", "flask"),
        ("collapsible", "soft", "squeeze", "compressible"),
        ("hydrapak.com", "salomon.com"),
    ),
    MarketCategory(
        "collapsible",
        "cup",
        ("collapsible cup", "foldable cup", "telescoping cup"),
        ("cup",),
        ("collapsible", "foldable", "telescoping", "nesting"),
        (),
    ),
    MarketCategory(
        "collapsible",
        "coffee cup",
        ("collapsible coffee cup", "foldable coffee cup", "collapsible travel cup"),
        ("coffee cup", "travel cup", "cup"),
        ("collapsible", "foldable", "telescoping", "compact"),
        ("stojo.co",),
    ),
    MarketCategory(
        "collapsible",
        "bowl",
        ("collapsible bowl", "foldable bowl", "collapsible camping bowl"),
        ("bowl",),
        ("collapsible", "foldable", "telescoping", "nesting"),
        (),
    ),
    MarketCategory(
        "collapsible",
        "food container",
        (
            "collapsible food container",
            "foldable food container",
            "collapsible storage container",
        ),
        ("food container", "container"),
        ("collapsible", "foldable", "telescoping", "nesting"),
        (),
    ),
    MarketCategory(
        "collapsible",
        "cooler",
        ("collapsible cooler", "foldable cooler", "soft collapsible cooler"),
        ("cooler",),
        ("collapsible", "foldable", "soft sided", "folds flat"),
        (),
    ),
    MarketCategory(
        "collapsible",
        "bucket",
        ("collapsible bucket", "foldable bucket", "collapsible camping bucket"),
        ("bucket",),
        ("collapsible", "foldable", "folds flat"),
        (),
    ),
    MarketCategory(
        "collapsible",
        "laundry basket",
        ("collapsible laundry basket", "foldable laundry basket", "pop up hamper"),
        ("laundry basket", "hamper", "basket"),
        ("collapsible", "foldable", "pop up", "folds flat"),
        (),
    ),
    MarketCategory(
        "collapsible",
        "storage bin",
        ("collapsible storage bin", "foldable storage bin", "collapsible crate"),
        ("storage bin", "bin", "crate"),
        ("collapsible", "foldable", "folding", "folds flat"),
        (),
    ),
    MarketCategory(
        "collapsible",
        "traffic cone",
        ("collapsible traffic cone", "foldable traffic cone", "collapsible safety cone"),
        ("traffic cone", "safety cone", "cone"),
        ("collapsible", "foldable", "telescoping"),
        (),
    ),
    MarketCategory(
        "collapsible",
        "lantern",
        ("collapsible lantern", "foldable lantern", "collapsible camping lantern"),
        ("lantern",),
        ("collapsible", "foldable", "telescoping", "pop up"),
        (),
    ),
    MarketCategory(
        "collapsible",
        "trekking pole",
        ("collapsible trekking pole", "folding trekking pole", "telescoping trekking pole"),
        ("trekking pole", "hiking pole"),
        ("collapsible", "folding", "telescoping", "foldable"),
        (),
    ),
    MarketCategory(
        "collapsible",
        "pet bowl",
        ("collapsible pet bowl", "foldable dog bowl", "collapsible dog bowl"),
        ("pet bowl", "dog bowl", "cat bowl"),
        ("collapsible", "foldable", "folds flat"),
        (),
    ),
    MarketCategory(
        "collapsible",
        "measuring cups",
        ("collapsible measuring cups", "foldable measuring cups", "collapsible measuring cup set"),
        ("measuring cup", "measuring cups"),
        ("collapsible", "foldable", "telescoping", "nested"),
        (),
    ),
    MarketCategory(
        "rollable",
        "solar charger",
        ("roll up solar charger", "rollable solar charger", "roll up solar panel"),
        ("solar charger", "solar panel"),
        ("roll up", "rollable", "flexible", "roll-up"),
        (),
    ),
    MarketCategory(
        "rollable",
        "keyboard",
        ("rollable keyboard", "roll up keyboard", "flexible roll up keyboard"),
        ("keyboard",),
        ("rollable", "roll up", "roll-up", "flexible"),
        (),
    ),
    MarketCategory(
        "inflatable",
        "air mattress",
        ("inflatable air mattress", "camping air mattress", "deflatable air mattress"),
        ("air mattress", "sleeping pad"),
        ("inflatable", "deflatable", "air"),
        (),
    ),
    MarketCategory(
        "inflatable",
        "inflatable kayak",
        ("inflatable kayak", "deflatable kayak", "packable inflatable kayak"),
        ("kayak",),
        ("inflatable", "deflatable", "packable"),
        (),
    ),
    MarketCategory(
        "inflatable",
        "inflatable paddle board",
        ("inflatable paddle board", "inflatable sup", "deflatable paddle board"),
        ("paddle board", "sup", "stand up paddle"),
        ("inflatable", "deflatable", "packable"),
        (),
    ),
    MarketCategory(
        "modular",
        "furniture",
        ("flat pack furniture", "modular collapsible furniture", "foldable furniture"),
        ("furniture", "chair", "table", "shelf"),
        ("flat pack", "modular", "collapsible", "foldable", "disassemblable"),
        ("ikea.com",),
    ),
)

CATEGORY_LOOKUP: Final[dict[str, MarketCategory]] = {
    category.category_hint: category for category in CATEGORIES
}

DENYLIST_DOMAINS: Final[tuple[str, ...]] = (
    "facebook.com",
    "instagram.com",
    "pinterest.com",
    "tiktok.com",
    "youtube.com",
    "reddit.com",
    "x.com",
    "twitter.com",
    "linkedin.com",
    "wikipedia.org",
    "quora.com",
)

MARKETPLACE_DOMAIN_FRAGMENTS: Final[tuple[str, ...]] = (
    "amazon.",
    "walmart.",
    "target.",
    "bestbuy.",
    "ebay.",
    "etsy.",
    "aliexpress.",
    "alibaba.",
    "temu.",
    "homedepot.",
    "lowes.",
    "rei.",
    "newegg.",
    "dickssportinggoods.",
)

LISTICLE_DOMAIN_FRAGMENTS: Final[tuple[str, ...]] = (
    "wirecutter.",
    "nytimes.",
    "forbes.",
    "cnet.",
    "tomsguide.",
    "techradar.",
    "rtings.",
    "gearpatrol.",
    "bestproducts.",
    "outdoorgearlab.",
    "bicycling.",
)

GENERIC_PRODUCT_WORDS: Final[set[str]] = {
    "official",
    "product",
    "products",
    "shop",
    "store",
    "buy",
    "sale",
    "new",
    "portable",
    "compact",
    "foldable",
    "folding",
    "folded",
    "collapsible",
    "rollable",
    "inflatable",
    "modular",
    "bike",
    "bicycle",
    "headphone",
    "headphones",
    "keyboard",
    "mouse",
    "water",
    "bottle",
    "cup",
    "bowl",
    "chair",
    "table",
    "kayak",
    "canoe",
    "helmet",
    "display",
    "monitor",
    "laptop",
    "phone",
    "smartphone",
    "drone",
    "stroller",
    "wagon",
    "treadmill",
    "wheelchair",
    "solar",
    "panel",
    "charger",
    "air",
    "mattress",
    "furniture",
    "set",
    "pack",
    "with",
    "and",
    "the",
    "for",
}

INACTIVE_MARKERS: Final[tuple[str, ...]] = (
    "discontinued",
    "no longer available",
    "archived",
    "retired",
    "out of production",
)

OUT_OF_STOCK_MARKERS: Final[tuple[str, ...]] = (
    "out of stock",
    "sold out",
    "currently unavailable",
    "unavailable",
)

# =============================================================================
# DISPLAY HELPERS
# =============================================================================


def configure_colab_dataframe_display() -> None:
    """Disable Colab's interactive dataframe formatter when available."""
    try:
        from google.colab import data_table

        data_table.disable_dataframe_formatter()
    except Exception:
        pass


def clean_text(value: Any, max_chars: int | None = None) -> str:
    """Normalize a value to a single-line string."""
    if value is None:
        return ""

    try:
        if isinstance(pd.isna(value), bool) and pd.isna(value):
            return ""
    except (TypeError, ValueError):
        pass

    text = re.sub(r"\s+", " ", str(value)).strip()

    if max_chars is not None and len(text) > max_chars:
        return text[: max_chars - 1].rstrip() + "…"

    return text


def display_table(
    title: str,
    df: pd.DataFrame,
    max_rows: int | None = DISPLAY_MAX_ROWS,
) -> None:
    """Display a dataframe as bounded static HTML."""
    table = df.copy() if max_rows is None else df.head(max_rows).copy()

    for column in table.columns:
        if table[column].dtype == "object":
            table[column] = table[column].map(
                lambda value: clean_text(value, DISPLAY_MAX_COL_WIDTH)
            )

    display(Markdown(f"### {title}"))
    html = table.to_html(index=False, escape=True)
    wrapped = (
        '<div style="max-height: 560px; overflow: auto; '
        'border: 1px solid #444; padding: 8px;">'
        f"{html}"
        "</div>"
    )
    display(HTML(wrapped))


configure_colab_dataframe_display()

# =============================================================================
# URL AND DOMAIN UTILITIES
# =============================================================================


def normalize_url(url: str) -> str:
    """Normalize a URL for deduplication."""
    parsed = urllib.parse.urlparse(clean_text(url))
    scheme = parsed.scheme.lower() or "https"
    netloc = parsed.netloc.lower().removeprefix("www.")
    path = re.sub(r"/+$", "", parsed.path)
    return urllib.parse.urlunparse((scheme, netloc, path, "", "", ""))


def extract_domain(url: str) -> str:
    """Extract a normalized domain from a URL."""
    parsed = urllib.parse.urlparse(clean_text(url))
    return parsed.netloc.lower().removeprefix("www.")


def absolute_url(base_url: str, href: str) -> str:
    """Build an absolute URL from a base URL and a link target."""
    return urllib.parse.urljoin(base_url, clean_text(href))


def domain_contains(domain: str, fragments: tuple[str, ...]) -> bool:
    """Return whether a domain contains any configured fragment."""
    return any(fragment in domain for fragment in fragments)


def domain_is_denied(domain: str) -> bool:
    """Return whether a domain should be skipped."""
    return any(blocked in domain for blocked in DENYLIST_DOMAINS)


def domain_role(domain: str, category: MarketCategory) -> str:
    """Classify source domain role for market-volume evidence."""
    if domain_contains(domain, category.official_domain_fragments):
        return "known_official"

    if domain_contains(domain, MARKETPLACE_DOMAIN_FRAGMENTS):
        return "marketplace"

    if domain_contains(domain, LISTICLE_DOMAIN_FRAGMENTS):
        return "review_or_listicle"

    return "general_web"


def build_robots_url(url: str) -> str:
    """Build robots.txt URL for a page URL."""
    parsed = urllib.parse.urlparse(url)
    root = f"{parsed.scheme}://{parsed.netloc}"
    return urllib.parse.urljoin(root, "/robots.txt")


# =============================================================================
# QUERY CONSTRUCTION AND SEARCH
# =============================================================================


def install_query_id(query_index: int) -> str:
    """Build a stable query identifier."""
    return f"mq-{query_index:05d}"


def build_general_queries(category: MarketCategory) -> list[dict[str, str]]:
    """Build bounded general web queries for one category."""
    templates = (
        "{term} official product",
        "{term} product model",
        "{term} manufacturer product",
        "{term} buy product",
        "{term} current model",
    )

    rows: list[dict[str, str]] = []
    terms = category.search_terms[:MAX_SEARCH_TERMS_PER_CATEGORY]

    for term in terms:
        for template in templates:
            rows.append(
                {
                    "family_hint": category.family_hint,
                    "category_hint": category.category_hint,
                    "query_kind": "general",
                    "query": template.format(term=term),
                }
            )

    return rows[:MAX_GENERAL_QUERIES_PER_CATEGORY]


def build_site_queries(category: MarketCategory) -> list[dict[str, str]]:
    """Build bounded site-directed queries for official domains."""
    rows: list[dict[str, str]] = []
    terms = category.search_terms[:MAX_SEARCH_TERMS_PER_CATEGORY]

    for domain in category.official_domain_fragments:
        for term in terms[:1]:
            rows.append(
                {
                    "family_hint": category.family_hint,
                    "category_hint": category.category_hint,
                    "query_kind": "known_official_site",
                    "query": f"site:{domain} {term} product",
                }
            )

    return rows[:MAX_SITE_QUERIES_PER_CATEGORY]


def build_query_table() -> pd.DataFrame:
    """Build the full market-volume query table."""
    rows: list[dict[str, str]] = []

    for category in CATEGORIES:
        rows.extend(build_general_queries(category))
        rows.extend(build_site_queries(category))

    unique_rows: list[dict[str, str]] = []
    seen: set[tuple[str, str, str]] = set()

    for row in rows:
        key = (
            row["category_hint"].lower(),
            row["query_kind"].lower(),
            row["query"].lower(),
        )

        if key in seen:
            continue

        seen.add(key)
        unique_rows.append(row)

    if MAX_QUERIES_TOTAL is not None:
        unique_rows = unique_rows[:MAX_QUERIES_TOTAL]

    query_df = pd.DataFrame(unique_rows)
    query_df["query_id"] = [
        install_query_id(index) for index in range(1, len(query_df) + 1)
    ]

    return query_df[
        ["query_id", "family_hint", "category_hint", "query_kind", "query"]
    ]


def run_single_search(query: str, max_results: int) -> list[dict[str, str]]:
    """Run one DDGS text search."""
    try:
        with DDGS(timeout=SEARCH_TIMEOUT_S) as ddgs:
            results = ddgs.text(
                query,
                region=SEARCH_REGION,
                safesearch=SAFESEARCH,
                max_results=max_results,
            )
        return list(results)
    except Exception as exc:
        print(f"Search failed: {query}")
        print(f"Reason: {exc}")
        return []


def run_market_searches(query_df: pd.DataFrame) -> pd.DataFrame:
    """Run all market-volume search queries."""
    rows: list[dict[str, Any]] = []
    total_queries = len(query_df)

    for index, query_row in query_df.iterrows():
        query = clean_text(query_row["query"])
        started_at = time.perf_counter()
        print(f"[{index + 1}/{total_queries}] Searching: {query}")

        results = run_single_search(query, RESULTS_PER_QUERY)
        elapsed_s = time.perf_counter() - started_at
        print(f"    results={len(results)} | elapsed={elapsed_s:.2f} s")

        category = CATEGORY_LOOKUP[query_row["category_hint"]]

        for rank, result in enumerate(results, start=1):
            url = clean_text(result.get("href") or result.get("url"))
            title = clean_text(result.get("title"))
            snippet = clean_text(result.get("body"))

            if url == "":
                continue

            domain = extract_domain(url)

            if domain_is_denied(domain):
                continue

            rows.append(
                {
                    "query_id": query_row["query_id"],
                    "query": query,
                    "query_kind": query_row["query_kind"],
                    "family_hint": query_row["family_hint"],
                    "category_hint": query_row["category_hint"],
                    "search_rank": rank,
                    "source_title": title,
                    "source_snippet": snippet,
                    "url": url,
                    "normalized_url": normalize_url(url),
                    "domain": domain,
                    "source_role": domain_role(domain, category),
                }
            )

        if index < total_queries - 1:
            time.sleep(SEARCH_DELAY_S)

    columns = [
        "query_id",
        "query",
        "query_kind",
        "family_hint",
        "category_hint",
        "search_rank",
        "source_title",
        "source_snippet",
        "url",
        "normalized_url",
        "domain",
        "source_role",
    ]

    if len(rows) == 0:
        return pd.DataFrame(columns=columns)

    output_df = pd.DataFrame(rows).drop_duplicates(subset=["normalized_url"])
    return output_df[columns].reset_index(drop=True)


# =============================================================================
# HTTP FETCHING
# =============================================================================


def build_http_session() -> requests.Session:
    """Build a retry-enabled HTTP session."""
    retry = Retry(
        total=RETRY_TOTAL,
        connect=RETRY_TOTAL,
        read=RETRY_TOTAL,
        backoff_factor=RETRY_BACKOFF_FACTOR,
        status_forcelist=(429, 500, 502, 503, 504),
        allowed_methods=("GET", "HEAD"),
    )

    adapter = HTTPAdapter(max_retries=retry)
    session = requests.Session()
    session.headers.update({"User-Agent": USER_AGENT})
    session.mount("https://", adapter)
    session.mount("http://", adapter)
    return session


def can_fetch_url(
    session: requests.Session,
    url: str,
    robots_cache: dict[str, tuple[bool, str]],
) -> tuple[bool, str]:
    """Check robots.txt for a URL with bounded timeouts."""
    if RUN_ROBOTS_CHECK is False:
        return True, "robots_check_disabled"

    domain = extract_domain(url)

    if domain in robots_cache:
        return robots_cache[domain]

    robots_url = build_robots_url(url)

    try:
        response = session.get(
            robots_url,
            timeout=(ROBOTS_CONNECT_TIMEOUT_S, ROBOTS_READ_TIMEOUT_S),
            allow_redirects=True,
        )

        if response.ok is False:
            if ALLOW_WHEN_ROBOTS_UNAVAILABLE:
                result = (True, f"robots_unavailable_allowed:{response.status_code}")
            else:
                result = (False, f"robots_unavailable_blocked:{response.status_code}")

            robots_cache[domain] = result
            return result

        parser = urllib.robotparser.RobotFileParser()
        parser.parse(response.text.splitlines())

        allowed = parser.can_fetch(USER_AGENT, url)
        result = (allowed, "allowed" if allowed else "blocked_by_robots_txt")
        robots_cache[domain] = result
        return result

    except requests.RequestException as exc:
        if ALLOW_WHEN_ROBOTS_UNAVAILABLE:
            result = (True, f"robots_unavailable_allowed:{exc}")
        else:
            result = (False, f"robots_unavailable_blocked:{exc}")

        robots_cache[domain] = result
        return result


def response_looks_like_html(response: requests.Response) -> bool:
    """Return whether a response appears to contain HTML or text."""
    content_type = response.headers.get("content-type", "").lower()
    return "html" in content_type or "text" in content_type


def read_limited_response_text(response: requests.Response) -> str:
    """Read response content with an upper byte cap."""
    chunks: list[bytes] = []
    total_bytes = 0

    for chunk in response.iter_content(chunk_size=32_768):
        if chunk:
            chunks.append(chunk)
            total_bytes += len(chunk)

        if total_bytes >= MAX_HTML_BYTES:
            break

    raw = b"".join(chunks)
    encoding = response.encoding or response.apparent_encoding or "utf-8"
    return raw.decode(encoding, errors="replace")


def fetch_html(
    session: requests.Session,
    url: str,
    robots_cache: dict[str, tuple[bool, str]],
) -> FetchResult:
    """Fetch one HTML page with bounded network behavior."""
    allowed, robots_message = can_fetch_url(session, url, robots_cache)

    if allowed is False:
        return FetchResult(url, url, None, "", False, robots_message)

    try:
        response = session.get(
            url,
            timeout=(REQUEST_CONNECT_TIMEOUT_S, REQUEST_READ_TIMEOUT_S),
            allow_redirects=True,
            stream=True,
        )

        if response.ok is False:
            return FetchResult(
                url=url,
                final_url=response.url,
                status_code=response.status_code,
                html="",
                ok=False,
                message=f"{robots_message}; http_status:{response.status_code}",
            )

        if response_looks_like_html(response) is False:
            return FetchResult(
                url=url,
                final_url=response.url,
                status_code=response.status_code,
                html="",
                ok=False,
                message=f"{robots_message}; non_html",
            )

        html = read_limited_response_text(response)
        html_is_valid = html.lstrip().startswith("<") or "<html" in html[:1000].lower()

        return FetchResult(
            url=url,
            final_url=response.url,
            status_code=response.status_code,
            html=html if html_is_valid else "",
            ok=bool(html_is_valid),
            message=robots_message if html_is_valid else "invalid_html",
        )

    except requests.RequestException as exc:
        return FetchResult(url, url, None, "", False, f"{robots_message}; {exc}")


# =============================================================================
# HTML AND JSON-LD EXTRACTION
# =============================================================================


def parse_json_safely(raw_json: str) -> Any:
    """Parse JSON while tolerating common control characters."""
    text = clean_text(raw_json)

    if text == "":
        return None

    try:
        return json.loads(text)
    except json.JSONDecodeError:
        text = re.sub(r"[\x00-\x1f\x7f]", " ", text)
        try:
            return json.loads(text)
        except json.JSONDecodeError:
            return None


def as_list(value: Any) -> list[Any]:
    """Convert scalar, missing, or list values into a list."""
    if value is None:
        return []

    if isinstance(value, list):
        return value

    return [value]


def flatten_json_ld(value: Any) -> list[dict[str, Any]]:
    """Flatten JSON-LD objects and graphs."""
    flattened: list[dict[str, Any]] = []

    if isinstance(value, list):
        for item in value:
            flattened.extend(flatten_json_ld(item))
        return flattened

    if isinstance(value, dict):
        flattened.append(value)

        for key in ("@graph", "mainEntity", "itemListElement", "offers"):
            child = value.get(key)

            if isinstance(child, (dict, list)):
                flattened.extend(flatten_json_ld(child))

    return flattened


def extract_json_ld_records(soup: BeautifulSoup) -> list[dict[str, Any]]:
    """Extract parseable JSON-LD records from a page."""
    records: list[dict[str, Any]] = []

    for script in soup.find_all("script", attrs={"type": "application/ld+json"}):
        parsed = parse_json_safely(script.get_text())
        if parsed is None:
            continue

        records.extend(flatten_json_ld(parsed))

    return records


def record_has_type(record: dict[str, Any], target_type: str) -> bool:
    """Return whether a JSON-LD record has a target type."""
    raw_type = record.get("@type")
    values = [str(item).lower() for item in as_list(raw_type)]
    return target_type.lower() in values


def extract_brand_name(value: Any) -> str:
    """Extract a brand name from JSON-LD brand metadata."""
    if isinstance(value, dict):
        return clean_text(value.get("name"))

    if isinstance(value, list):
        for item in value:
            brand_name = extract_brand_name(item)
            if brand_name:
                return brand_name

    return clean_text(value)


def extract_image_url(value: Any) -> str:
    """Extract the first image URL from JSON-LD image metadata."""
    for item in as_list(value):
        if isinstance(item, dict):
            image_url = clean_text(item.get("url") or item.get("contentUrl"))
        else:
            image_url = clean_text(item)

        if image_url:
            return image_url

    return ""


def extract_offer(product: dict[str, Any]) -> dict[str, Any]:
    """Extract the first usable offer dictionary from JSON-LD."""
    offers = as_list(product.get("offers"))

    for offer in offers:
        if isinstance(offer, dict):
            return offer

    return {}


def get_meta_content(soup: BeautifulSoup, *names: str) -> str:
    """Read content from common meta tags."""
    for name in names:
        for attrs in ({"property": name}, {"name": name}, {"itemprop": name}):
            tag = soup.find("meta", attrs=attrs)

            if tag is not None and tag.get("content"):
                return clean_text(tag.get("content"))

    return ""


def get_page_title(soup: BeautifulSoup) -> str:
    """Extract page title from visible or metadata sources."""
    h1 = soup.find("h1")
    title = soup.find("title")

    for tag in (h1, title):
        if tag is None:
            continue

        text = clean_text(tag.get_text(" ", strip=True))
        if text:
            return text

    return ""


def get_canonical_url(soup: BeautifulSoup, fallback_url: str) -> str:
    """Extract canonical URL when available."""
    tag = soup.find("link", rel=lambda value: value and "canonical" in value)

    if tag is not None and tag.get("href"):
        return absolute_url(fallback_url, clean_text(tag.get("href")))

    return fallback_url


def page_visible_text(soup: BeautifulSoup, max_chars: int = 8_000) -> str:
    """Extract bounded visible page text for evidence checks."""
    for tag in soup(["script", "style", "noscript", "svg"]):
        tag.decompose()

    text = clean_text(soup.get_text(" ", strip=True))
    return text[:max_chars]


def page_metadata(soup: BeautifulSoup, fetch_result: FetchResult) -> dict[str, str]:
    """Extract common page metadata."""
    canonical_url = get_canonical_url(soup, fetch_result.final_url)

    return {
        "canonical_url": canonical_url,
        "page_title": clean_text(
            get_meta_content(soup, "og:title", "twitter:title")
            or get_page_title(soup)
        ),
        "page_description": clean_text(
            get_meta_content(
                soup,
                "og:description",
                "twitter:description",
                "description",
            ),
            520,
        ),
        "page_image_url": clean_text(
            get_meta_content(soup, "og:image", "twitter:image")
        ),
    }


def url_looks_product_like(url: str) -> bool:
    """Return whether a URL path looks like a product detail page."""
    parsed = urllib.parse.urlparse(url)
    path = parsed.path.lower()

    product_path_markers = (
        "/product/",
        "/products/",
        "/shop/",
        "/store/",
        "/p/",
        "/dp/",
        "/item/",
        "/collections/",
    )

    return any(marker in path for marker in product_path_markers)


def extract_product_links(
    soup: BeautifulSoup,
    base_url: str,
    category: MarketCategory,
) -> list[str]:
    """Extract likely product links from a category or search landing page."""
    links: list[str] = []
    seen: set[str] = set()

    for anchor in soup.find_all("a", href=True):
        href = clean_text(anchor.get("href"))
        text = clean_text(anchor.get_text(" ", strip=True)).lower()
        candidate_url = absolute_url(base_url, href)
        normalized = normalize_url(candidate_url)
        domain = extract_domain(candidate_url)

        if domain_is_denied(domain):
            continue

        if normalized in seen:
            continue

        marker_hit = any(
            marker.lower() in text or marker.lower() in candidate_url.lower()
            for marker in category.category_markers + category.transformation_markers
        )

        if url_looks_product_like(candidate_url) and marker_hit:
            seen.add(normalized)
            links.append(candidate_url)

        if len(links) >= MAX_DISCOVERED_LINKS_PER_PAGE:
            break

    return links


# =============================================================================
# EVIDENCE, CONFIDENCE, AND PRODUCT RECORD EXTRACTION
# =============================================================================


def text_has_any(text: str, markers: tuple[str, ...]) -> bool:
    """Return whether any marker appears in a text."""
    lowered = text.lower()
    return any(marker.lower() in lowered for marker in markers)


def category_evidence_score(text: str, category: MarketCategory) -> float:
    """Score text support for the intended product category."""
    lowered = text.lower()
    hits = sum(marker.lower() in lowered for marker in category.category_markers)
    return min(1.0, hits / max(len(category.category_markers), 1))


def transformation_evidence_score(text: str, category: MarketCategory) -> float:
    """Score text support for the compacting transformation."""
    lowered = text.lower()
    hits = sum(marker.lower() in lowered for marker in category.transformation_markers)
    return min(1.0, hits / max(min(len(category.transformation_markers), 4), 1))


def first_nonempty(*values: Any) -> str:
    """Return the first nonempty cleaned string from a sequence."""
    for value in values:
        text = clean_text(value)

        if text:
            return text

    return ""


def normalize_availability(value: Any) -> str:
    """Normalize availability strings from schema.org or page text."""
    text = clean_text(value)

    if text.startswith("http"):
        text = text.rstrip("/").split("/")[-1]

    return text


def infer_market_status(evidence_text: str, availability: str) -> str:
    """Infer market status from text and availability."""
    lowered = f"{evidence_text} {availability}".lower()

    if text_has_any(lowered, INACTIVE_MARKERS):
        return "inactive_or_discontinued"

    if text_has_any(lowered, OUT_OF_STOCK_MARKERS):
        return "active_unavailable_or_out_of_stock"

    if "preorder" in lowered or "pre-order" in lowered:
        return "preorder"

    return "active_or_listed"


def source_role_score(source_role: str) -> float:
    """Convert source role into confidence contribution."""
    scores = {
        "known_official": 0.18,
        "marketplace": 0.10,
        "general_web": 0.07,
        "review_or_listicle": -0.08,
    }
    return scores.get(source_role, 0.0)


def product_confidence_score(row: dict[str, Any], category: MarketCategory) -> float:
    """Compute confidence that a row is an active product-model record."""
    name = clean_text(row.get("product_name"))
    brand = clean_text(row.get("brand"))
    description = clean_text(row.get("description"))
    title = clean_text(row.get("source_title"))
    snippet = clean_text(row.get("source_snippet"))
    page_title = clean_text(row.get("page_title"))
    url = clean_text(row.get("product_url"))
    availability = clean_text(row.get("availability"))
    source_role = clean_text(row.get("source_role"))

    evidence_text = " | ".join(
        [name, brand, description, title, snippet, page_title, url, availability]
    )

    score = 0.0
    score += 0.16 if name else 0.0
    score += 0.08 if brand else 0.0
    score += 0.14 if bool(row.get("json_ld_product_found")) else 0.0
    score += 0.08 if clean_text(row.get("price")) else 0.0
    score += 0.06 if availability else 0.0
    score += 0.10 if url_looks_product_like(url) else 0.0
    score += 0.16 * category_evidence_score(evidence_text, category)
    score += 0.16 * transformation_evidence_score(evidence_text, category)
    score += source_role_score(source_role)

    if infer_market_status(evidence_text, availability) == "inactive_or_discontinued":
        score -= 0.25

    if len(product_tokens(name)) < MIN_NAME_TOKEN_COUNT_FOR_VOLUME:
        score -= 0.10

    return round(float(np.clip(score, 0.0, 1.0)), 3)


def build_product_record(
    context: dict[str, Any],
    metadata: dict[str, str],
    product: dict[str, Any] | None,
    page_text: str,
) -> dict[str, Any]:
    """Build one product evidence record from JSON-LD or page fallback."""
    category = CATEGORY_LOOKUP[context["category_hint"]]
    product_data = product or {}
    offer = extract_offer(product_data)

    product_url = first_nonempty(
        product_data.get("url"),
        metadata.get("canonical_url"),
        context.get("url"),
    )
    product_url = absolute_url(context["url"], product_url)

    product_name = first_nonempty(
        product_data.get("name"),
        metadata.get("page_title"),
        context.get("source_title"),
    )

    brand = first_nonempty(
        extract_brand_name(product_data.get("brand")),
        infer_brand_from_text(product_name, context["domain"], category),
    )

    description = clean_text(
        first_nonempty(
            product_data.get("description"),
            metadata.get("page_description"),
            context.get("source_snippet"),
        ),
        520,
    )

    availability = normalize_availability(
        first_nonempty(
            offer.get("availability"),
            "",
        )
    )

    evidence_text = " | ".join(
        [
            product_name,
            brand,
            description,
            metadata.get("page_title", ""),
            metadata.get("page_description", ""),
            context.get("source_title", ""),
            context.get("source_snippet", ""),
            page_text[:1200],
        ]
    )

    row: dict[str, Any] = {
        "family_hint": context["family_hint"],
        "category_hint": context["category_hint"],
        "query_id": context.get("query_id", ""),
        "query": context.get("query", ""),
        "query_kind": context.get("query_kind", ""),
        "source_role": context.get("source_role", ""),
        "source_rank": context.get("search_rank", ""),
        "source_url": context.get("url", ""),
        "source_domain": context.get("domain", ""),
        "source_title": context.get("source_title", ""),
        "source_snippet": context.get("source_snippet", ""),
        "page_title": metadata.get("page_title", ""),
        "page_description": metadata.get("page_description", ""),
        "product_url": product_url,
        "normalized_product_url": normalize_url(product_url),
        "product_domain": extract_domain(product_url),
        "product_name": product_name,
        "brand": brand,
        "description": description,
        "price": first_nonempty(
            offer.get("price"),
            offer.get("lowPrice"),
            offer.get("highPrice"),
        ),
        "currency": first_nonempty(offer.get("priceCurrency")),
        "availability": availability,
        "image_url": first_nonempty(
            extract_image_url(product_data.get("image")),
            metadata.get("page_image_url"),
        ),
        "json_ld_product_found": product is not None,
        "category_evidence": round(category_evidence_score(evidence_text, category), 3),
        "transformation_evidence": round(
            transformation_evidence_score(evidence_text, category),
            3,
        ),
        "market_status": infer_market_status(evidence_text, availability),
    }

    row["product_confidence"] = product_confidence_score(row, category)
    return row


def extract_product_records_from_page(
    context: dict[str, Any],
    fetch_result: FetchResult,
) -> tuple[list[dict[str, Any]], dict[str, Any], list[str]]:
    """Extract product records and product links from one fetched page."""
    if fetch_result.ok is False:
        page_row = {
            **context,
            "final_url": fetch_result.final_url,
            "status_code": fetch_result.status_code,
            "fetch_ok": False,
            "fetch_message": fetch_result.message,
            "page_title": "",
            "page_description": "",
            "product_record_count": 0,
            "discovered_product_link_count": 0,
        }
        return [], page_row, []

    soup = BeautifulSoup(fetch_result.html, "lxml")
    category = CATEGORY_LOOKUP[context["category_hint"]]
    metadata = page_metadata(soup, fetch_result)
    visible_text = page_visible_text(soup)

    json_ld_records = extract_json_ld_records(soup)
    products = [
        record
        for record in json_ld_records
        if isinstance(record, dict) and record_has_type(record, "Product")
    ]

    product_rows: list[dict[str, Any]] = []

    for product in products:
        product_rows.append(
            build_product_record(
                context=context,
                metadata=metadata,
                product=product,
                page_text=visible_text,
            )
        )

    if len(product_rows) == 0:
        fallback_text = " | ".join(
            [
                metadata.get("page_title", ""),
                metadata.get("page_description", ""),
                context.get("source_title", ""),
                context.get("source_snippet", ""),
                visible_text[:1200],
            ]
        )
        has_category = category_evidence_score(fallback_text, category) > 0
        has_transform = transformation_evidence_score(fallback_text, category) > 0

        if has_category and has_transform:
            product_rows.append(
                build_product_record(
                    context=context,
                    metadata=metadata,
                    product=None,
                    page_text=visible_text,
                )
            )

    discovered_links = extract_product_links(
        soup=soup,
        base_url=fetch_result.final_url,
        category=category,
    )

    page_row = {
        **context,
        "final_url": fetch_result.final_url,
        "status_code": fetch_result.status_code,
        "fetch_ok": True,
        "fetch_message": fetch_result.message,
        "page_title": metadata.get("page_title", ""),
        "page_description": metadata.get("page_description", ""),
        "product_record_count": len(product_rows),
        "discovered_product_link_count": len(discovered_links),
    }

    return product_rows, page_row, discovered_links


# =============================================================================
# BRAND AND PRODUCT IDENTITY DEDUPLICATION
# =============================================================================


def root_domain_token(domain: str) -> str:
    """Extract a coarse brand-like token from a domain."""
    domain = clean_text(domain).lower().removeprefix("www.")
    parts = domain.split(".")

    if len(parts) == 0:
        return ""

    skip = {"com", "co", "net", "org", "us", "ca", "uk", "shop", "store"}
    tokens = [part for part in parts if part not in skip]

    return tokens[0] if tokens else parts[0]


def infer_brand_from_text(
    product_name: str,
    domain: str,
    category: MarketCategory,
) -> str:
    """Infer a brand name from title text or known official domains."""
    lowered_name = clean_text(product_name).lower()

    for fragment in category.official_domain_fragments:
        token = root_domain_token(fragment)

        if token and token in clean_text(domain).lower():
            return token

        if token and token in lowered_name:
            return token

    return ""


def slugify_text(text: str) -> str:
    """Convert text into a stable lowercase token string."""
    lowered = clean_text(text).lower()
    lowered = re.sub(r"[^a-z0-9]+", " ", lowered)
    lowered = re.sub(r"\s+", " ", lowered).strip()
    return lowered


def product_tokens(name: str) -> list[str]:
    """Tokenize a product name while suppressing generic category words."""
    tokens = slugify_text(name).split()
    filtered = [
        token
        for token in tokens
        if len(token) >= 2 and token not in GENERIC_PRODUCT_WORDS
    ]
    return filtered


def normalize_brand(brand: str, domain: str) -> str:
    """Normalize a brand string or infer a weak brand from domain."""
    cleaned = slugify_text(brand)

    if cleaned:
        return cleaned

    domain_token = root_domain_token(domain)

    if domain_token in {
        "amazon",
        "walmart",
        "target",
        "bestbuy",
        "ebay",
        "etsy",
        "aliexpress",
        "homedepot",
        "lowes",
    }:
        return "unknown"

    return domain_token or "unknown"


def normalized_product_name(row: pd.Series) -> str:
    """Build a normalized product-name signature."""
    name = first_nonempty(row.get("product_name"), row.get("page_title"))
    brand = clean_text(row.get("normalized_brand"))

    name_slug = slugify_text(name)
    brand_tokens = set(brand.split())

    tokens = [
        token
        for token in name_slug.split()
        if token not in brand_tokens and token not in GENERIC_PRODUCT_WORDS
    ]

    if len(tokens) == 0:
        tokens = product_tokens(name)

    return " ".join(tokens[:12])


def token_jaccard(left: str, right: str) -> float:
    """Compute token-set Jaccard similarity."""
    left_set = set(left.split())
    right_set = set(right.split())

    if len(left_set) == 0 or len(right_set) == 0:
        return 0.0

    return len(left_set & right_set) / len(left_set | right_set)


def names_are_same_model(left: str, right: str) -> bool:
    """Return whether two normalized names likely refer to one model."""
    if left == "" or right == "":
        return False

    if left == right:
        return True

    sequence = SequenceMatcher(None, left, right).ratio()
    jaccard = token_jaccard(left, right)

    return sequence >= DEDUP_SEQUENCE_THRESHOLD or jaccard >= DEDUP_JACCARD_THRESHOLD


def stable_hash_id(prefix: str, text: str, length: int = 12) -> str:
    """Build a stable short hash identifier."""
    digest = hashlib.sha1(text.encode("utf-8")).hexdigest()[:length]
    return f"{prefix}-{digest}"


def prepare_records_for_dedup(product_df: pd.DataFrame) -> pd.DataFrame:
    """Prepare product records with normalized brand and name columns."""
    output_df = product_df.copy()

    output_df["normalized_brand"] = output_df.apply(
        lambda row: normalize_brand(row.get("brand", ""), row.get("product_domain", "")),
        axis=1,
    )
    output_df["normalized_name"] = output_df.apply(normalized_product_name, axis=1)
    output_df["name_token_count"] = output_df["normalized_name"].map(
        lambda value: len(str(value).split())
    )

    output_df["record_id"] = [
        stable_hash_id(
            "rec",
            "|".join(
                [
                    clean_text(row.get("category_hint")),
                    clean_text(row.get("normalized_brand")),
                    clean_text(row.get("normalized_name")),
                    clean_text(row.get("normalized_product_url")),
                    str(index),
                ]
            ),
        )
        for index, row in output_df.iterrows()
    ]

    return output_df


def assign_product_model_ids(product_df: pd.DataFrame) -> pd.DataFrame:
    """Cluster product records into unique product-model identifiers."""
    if len(product_df) == 0:
        return product_df.assign(product_model_id=[])

    work_df = prepare_records_for_dedup(product_df)
    work_df = work_df.sort_values(
        [
            "category_hint",
            "normalized_brand",
            "product_confidence",
            "normalized_name",
        ],
        ascending=[True, True, False, True],
        kind="stable",
    ).reset_index(drop=True)

    model_ids = [""] * len(work_df)
    representatives: dict[tuple[str, str], list[tuple[str, str]]] = defaultdict(list)

    for index, row in work_df.iterrows():
        category = clean_text(row["category_hint"])
        brand = clean_text(row["normalized_brand"])
        name = clean_text(row["normalized_name"])
        group_key = (category, brand)
        assigned_id = ""

        for model_id, representative_name in representatives[group_key]:
            if names_are_same_model(name, representative_name):
                assigned_id = model_id
                break

        if assigned_id == "":
            id_basis = "|".join([category, brand, name, clean_text(row["product_domain"])])
            assigned_id = stable_hash_id("model", id_basis)
            representatives[group_key].append((assigned_id, name))

        model_ids[index] = assigned_id

    work_df["product_model_id"] = model_ids
    return work_df


def select_best_text(series: pd.Series) -> str:
    """Select the longest nonempty text value from a series."""
    values = [clean_text(value) for value in series.tolist()]
    values = [value for value in values if value]

    if len(values) == 0:
        return ""

    return sorted(values, key=len, reverse=True)[0]


def build_product_model_table(product_df: pd.DataFrame) -> pd.DataFrame:
    """Aggregate product evidence records into product-model rows."""
    if len(product_df) == 0:
        return pd.DataFrame()

    dedup_df = assign_product_model_ids(product_df)

    grouped = dedup_df.groupby("product_model_id", dropna=False)

    model_df = grouped.agg(
        family_hint=("family_hint", "first"),
        category_hint=("category_hint", "first"),
        brand=("brand", select_best_text),
        normalized_brand=("normalized_brand", "first"),
        product_name=("product_name", select_best_text),
        normalized_name=("normalized_name", "first"),
        confidence=("product_confidence", "max"),
        mean_confidence=("product_confidence", "mean"),
        evidence_count=("record_id", "count"),
        source_domain_count=("source_domain", "nunique"),
        product_domain_count=("product_domain", "nunique"),
        source_roles=("source_role", lambda x: ", ".join(sorted(set(map(str, x))))),
        official_source_count=(
            "source_role",
            lambda x: int(sum(value == "known_official" for value in x)),
        ),
        marketplace_source_count=(
            "source_role",
            lambda x: int(sum(value == "marketplace" for value in x)),
        ),
        json_ld_record_count=("json_ld_product_found", "sum"),
        best_url=("product_url", select_best_text),
        best_image_url=("image_url", select_best_text),
        price=("price", select_best_text),
        currency=("currency", select_best_text),
        availability=("availability", select_best_text),
        market_status=("market_status", lambda x: sorted(set(map(str, x)))[0]),
        category_evidence=("category_evidence", "max"),
        transformation_evidence=("transformation_evidence", "max"),
        name_token_count=("name_token_count", "max"),
    ).reset_index()

    model_df["mean_confidence"] = model_df["mean_confidence"].round(3)
    model_df["verified_for_volume"] = (
        (
            model_df["confidence"] >= MIN_PRODUCT_CONFIDENCE_FOR_VOLUME
        )
        & (
            model_df["name_token_count"] >= MIN_NAME_TOKEN_COUNT_FOR_VOLUME
        )
        & (
            model_df["market_status"] != "inactive_or_discontinued"
        )
    )

    if USE_LOW_CONFIDENCE_CANDIDATES_IN_VOLUME:
        model_df["included_in_volume"] = (
            model_df["market_status"] != "inactive_or_discontinued"
        )
    else:
        model_df["included_in_volume"] = model_df["verified_for_volume"]

    model_df = model_df.sort_values(
        ["included_in_volume", "confidence", "evidence_count"],
        ascending=[False, False, False],
        kind="stable",
    ).reset_index(drop=True)

    return model_df


# =============================================================================
# SCRAPING WORKFLOW
# =============================================================================


def build_initial_page_queue(search_df: pd.DataFrame) -> list[dict[str, Any]]:
    """Convert search-result rows into an initial page queue."""
    queue: list[dict[str, Any]] = []

    for _, row in search_df.iterrows():
        queue.append(
            {
                "query_id": row["query_id"],
                "query": row["query"],
                "query_kind": row["query_kind"],
                "family_hint": row["family_hint"],
                "category_hint": row["category_hint"],
                "search_rank": int(row["search_rank"]),
                "source_title": row["source_title"],
                "source_snippet": row["source_snippet"],
                "url": row["url"],
                "normalized_url": row["normalized_url"],
                "domain": row["domain"],
                "source_role": row["source_role"],
                "discovery_stage": "search_result",
            }
        )

    return queue


def enqueue_discovered_links(
    queue: list[dict[str, Any]],
    context: dict[str, Any],
    discovered_links: list[str],
    discovered_link_counts: dict[tuple[str, str], int],
) -> None:
    """Add discovered product links to the scrape queue with bounded counts."""
    category_hint = context["category_hint"]
    category = CATEGORY_LOOKUP[category_hint]

    for link in discovered_links:
        domain = extract_domain(link)
        key = (category_hint, domain)

        if discovered_link_counts[key] >= MAX_DISCOVERED_LINKS_PER_DOMAIN_CATEGORY:
            continue

        queue.append(
            {
                "query_id": context.get("query_id", ""),
                "query": context.get("query", ""),
                "query_kind": "discovered_product_link",
                "family_hint": category.family_hint,
                "category_hint": category_hint,
                "search_rank": "",
                "source_title": "",
                "source_snippet": "",
                "url": link,
                "normalized_url": normalize_url(link),
                "domain": domain,
                "source_role": domain_role(domain, category),
                "discovery_stage": "product_link",
            }
        )
        discovered_link_counts[key] += 1


def scrape_market_pages(search_df: pd.DataFrame) -> tuple[pd.DataFrame, pd.DataFrame]:
    """Scrape pages and return product records plus page audit rows."""
    session = build_http_session()
    robots_cache: dict[str, tuple[bool, str]] = {}
    queue = build_initial_page_queue(search_df)

    seen_urls: set[str] = set()
    product_rows: list[dict[str, Any]] = []
    page_rows: list[dict[str, Any]] = []
    category_scrape_counts: dict[str, int] = defaultdict(int)
    domain_category_scrape_counts: dict[tuple[str, str], int] = defaultdict(int)
    discovered_link_counts: dict[tuple[str, str], int] = defaultdict(int)

    index = 0
    total_scraped = 0

    while index < len(queue) and total_scraped < MAX_TOTAL_PAGES_TO_SCRAPE:
        context = queue[index]
        index += 1

        normalized = normalize_url(context["url"])
        category_hint = context["category_hint"]
        domain = context["domain"]
        domain_category_key = (category_hint, domain)

        if normalized in seen_urls:
            continue

        if category_scrape_counts[category_hint] >= MAX_SCRAPED_PAGES_PER_CATEGORY:
            continue

        if (
            domain_category_scrape_counts[domain_category_key]
            >= MAX_SCRAPED_PAGES_PER_DOMAIN_CATEGORY
        ):
            continue

        seen_urls.add(normalized)
        total_scraped += 1
        category_scrape_counts[category_hint] += 1
        domain_category_scrape_counts[domain_category_key] += 1

        print(
            f"[{total_scraped}/{MAX_TOTAL_PAGES_TO_SCRAPE}] "
            f"{category_hint} | {context['source_role']} | {context['url']}"
        )

        fetch_result = fetch_html(session, context["url"], robots_cache)
        records, page_row, discovered_links = extract_product_records_from_page(
            context=context,
            fetch_result=fetch_result,
        )

        product_rows.extend(records)
        page_rows.append(page_row)

        if fetch_result.ok:
            enqueue_discovered_links(
                queue=queue,
                context=context,
                discovered_links=discovered_links,
                discovered_link_counts=discovered_link_counts,
            )

        print(
            "    "
            f"ok={fetch_result.ok} | "
            f"status={fetch_result.status_code} | "
            f"records={len(records)} | "
            f"links={len(discovered_links)} | "
            f"message={clean_text(fetch_result.message, 120)}"
        )

        time.sleep(SCRAPE_DELAY_S)

    product_df = pd.DataFrame(product_rows)
    page_df = pd.DataFrame(page_rows)

    return product_df, page_df


# =============================================================================
# MARKET-VOLUME ESTIMATION
# =============================================================================


def chao1_estimate(observation_counts: pd.Series) -> tuple[float, int, int]:
    """Estimate total product richness using the Chao1 lower-bound estimator."""
    counts = observation_counts.astype(int)
    observed = int(len(counts))
    f1 = int((counts == 1).sum())
    f2 = int((counts == 2).sum())

    if observed == 0:
        return 0.0, f1, f2

    if f2 > 0:
        estimate = observed + (f1 * f1) / (2.0 * f2)
    else:
        estimate = observed + (f1 * (f1 - 1)) / 2.0

    return round(float(estimate), 2), f1, f2


def build_market_volume_summary(model_df: pd.DataFrame) -> pd.DataFrame:
    """Build category-level market-volume summary."""
    rows: list[dict[str, Any]] = []

    for category in CATEGORIES:
        category_models = model_df.loc[
            model_df["category_hint"] == category.category_hint
        ].copy()

        included = category_models.loc[category_models["included_in_volume"]].copy()
        observation_counts = included["evidence_count"] if len(included) else pd.Series(dtype=int)
        chao1, singleton_count, doubleton_count = chao1_estimate(observation_counts)

        rows.append(
            {
                "family_hint": category.family_hint,
                "category_hint": category.category_hint,
                "observed_unique_active_models": int(len(included)),
                "chao1_estimated_active_models": chao1,
                "unseen_model_lower_bound": round(max(0.0, chao1 - len(included)), 2),
                "candidate_model_count_all_confidences": int(len(category_models)),
                "verified_model_fraction": round(
                    len(included) / max(len(category_models), 1),
                    3,
                ),
                "singleton_model_count": singleton_count,
                "doubleton_model_count": doubleton_count,
                "known_official_model_count": int(
                    (included["official_source_count"] > 0).sum()
                    if len(included)
                    else 0
                ),
                "marketplace_model_count": int(
                    (included["marketplace_source_count"] > 0).sum()
                    if len(included)
                    else 0
                ),
                "mean_confidence": round(
                    float(included["confidence"].mean()) if len(included) else 0.0,
                    3,
                ),
                "mean_evidence_count": round(
                    float(included["evidence_count"].mean()) if len(included) else 0.0,
                    3,
                ),
                "median_evidence_count": round(
                    float(included["evidence_count"].median()) if len(included) else 0.0,
                    3,
                ),
            }
        )

    summary_df = pd.DataFrame(rows)
    summary_df = summary_df.sort_values(
        ["observed_unique_active_models", "chao1_estimated_active_models"],
        ascending=[False, False],
        kind="stable",
    ).reset_index(drop=True)

    return summary_df


def build_family_summary(market_summary_df: pd.DataFrame) -> pd.DataFrame:
    """Aggregate market-volume estimates by transformation family."""
    return (
        market_summary_df.groupby("family_hint", dropna=False)
        .agg(
            observed_unique_active_models=("observed_unique_active_models", "sum"),
            chao1_estimated_active_models=("chao1_estimated_active_models", "sum"),
            category_count=("category_hint", "count"),
            mean_confidence=("mean_confidence", "mean"),
        )
        .reset_index()
        .assign(mean_confidence=lambda table: table["mean_confidence"].round(3))
        .sort_values(
            "observed_unique_active_models",
            ascending=False,
            kind="stable",
        )
        .reset_index(drop=True)
    )


# =============================================================================
# PLOTS
# =============================================================================


def plot_horizontal_bar(
    series: pd.Series,
    title: str,
    xlabel: str,
    ylabel: str,
    figsize: tuple[float, float] = (11.0, 7.0),
) -> None:
    """Render a pure-blue horizontal bar chart."""
    plot_series = series.sort_values()

    fig, ax = plt.subplots(figsize=figsize)
    ax.barh(
        plot_series.index.astype(str),
        plot_series.values,
        color=PURE_BLUE,
        edgecolor="black",
        linewidth=0.8,
    )
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.grid(axis="x", alpha=0.25, color="black")
    fig.tight_layout()
    plt.show()


def plot_market_volume_by_category(
    market_summary_df: pd.DataFrame,
    value_column: str,
    title: str,
) -> None:
    """Plot category-level market-volume values."""
    plot_df = market_summary_df.head(TOP_N_PLOT).copy()
    series = plot_df.set_index("category_hint")[value_column]
    plot_horizontal_bar(
        series=series,
        title=title,
        xlabel="Unique active product-model count",
        ylabel="Category",
        figsize=(11.2, 7.4),
    )


def plot_family_market_volume(family_summary_df: pd.DataFrame) -> None:
    """Plot observed unique active model counts by transformation family."""
    series = family_summary_df.set_index("family_hint")[
        "observed_unique_active_models"
    ]
    plot_horizontal_bar(
        series=series,
        title="Observed Active Product-Model Volume by Transformation Family",
        xlabel="Unique active product-model count",
        ylabel="Transformation family",
        figsize=(9.2, 5.6),
    )


def plot_confidence_distribution(model_df: pd.DataFrame) -> None:
    """Plot product-model confidence distribution."""
    fig, ax = plt.subplots(figsize=(8.8, 5.0))
    ax.hist(
        model_df["confidence"].astype(float),
        bins=18,
        color=PURE_BLUE,
        edgecolor="black",
        linewidth=0.8,
    )
    ax.axvline(
        MIN_PRODUCT_CONFIDENCE_FOR_VOLUME,
        color="black",
        linewidth=1.1,
        linestyle=":",
        label="volume threshold",
    )
    ax.set_title("Product-Model Confidence Distribution")
    ax.set_xlabel("Confidence score")
    ax.set_ylabel("Product-model count")
    ax.grid(axis="y", alpha=0.25, color="black")
    ax.legend(frameon=False)
    fig.tight_layout()
    plt.show()


def plot_coverage_vs_volume(market_summary_df: pd.DataFrame) -> None:
    """Plot evidence coverage versus observed model volume."""
    fig, ax = plt.subplots(figsize=(8.8, 5.4))
    ax.scatter(
        market_summary_df["observed_unique_active_models"],
        market_summary_df["mean_evidence_count"],
        color=PURE_BLUE,
        edgecolor="black",
        linewidth=0.8,
        s=55,
    )
    ax.set_title("Coverage Diagnostic: Volume versus Mean Evidence Count")
    ax.set_xlabel("Observed unique active models")
    ax.set_ylabel("Mean evidence records per model")
    ax.grid(alpha=0.25, color="black")
    fig.tight_layout()
    plt.show()


# =============================================================================
# EXPORTS
# =============================================================================


def export_outputs(
    query_df: pd.DataFrame,
    search_df: pd.DataFrame,
    page_df: pd.DataFrame,
    product_df: pd.DataFrame,
    model_df: pd.DataFrame,
    market_summary_df: pd.DataFrame,
) -> None:
    """Export workflow tables to CSV files."""
    if EXPORT_FILES is False:
        return

    OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

    query_df.to_csv(QUERY_LOG_CSV_PATH, index=False)
    search_df.to_csv(SEARCH_RESULTS_CSV_PATH, index=False)
    page_df.to_csv(SCRAPED_PAGES_CSV_PATH, index=False)
    product_df.to_csv(PRODUCT_RECORDS_CSV_PATH, index=False)
    model_df.to_csv(PRODUCT_MODELS_CSV_PATH, index=False)
    market_summary_df.to_csv(MARKET_VOLUME_CSV_PATH, index=False)

    display(Markdown(f"Query log written to `{QUERY_LOG_CSV_PATH}`."))
    display(Markdown(f"Search results written to `{SEARCH_RESULTS_CSV_PATH}`."))
    display(Markdown(f"Scraped pages written to `{SCRAPED_PAGES_CSV_PATH}`."))
    display(Markdown(f"Product records written to `{PRODUCT_RECORDS_CSV_PATH}`."))
    display(Markdown(f"Product models written to `{PRODUCT_MODELS_CSV_PATH}`."))
    display(Markdown(f"Market volume summary written to `{MARKET_VOLUME_CSV_PATH}`."))


# =============================================================================
# MAIN WORKFLOW
# =============================================================================


def main() -> tuple[pd.DataFrame, pd.DataFrame, pd.DataFrame, pd.DataFrame]:
    """Run market-volume estimation workflow."""
    started_at = time.perf_counter()

    query_df = build_query_table()
    print(f"Market-volume queries loaded: {len(query_df)}")
    display_table("Market-volume query table", query_df, max_rows=140)

    search_df = run_market_searches(query_df)
    print(f"Unique search-result pages: {len(search_df)}")

    if len(search_df) == 0:
        raise RuntimeError("No search results were collected.")

    display_table("Search-result source audit", search_df, max_rows=120)

    if SCRAPE_PAGES:
        product_df, page_df = scrape_market_pages(search_df)
    else:
        product_df = pd.DataFrame()
        page_df = pd.DataFrame()

    print(f"Scraped page audit rows: {len(page_df)}")
    print(f"Extracted product evidence records: {len(product_df)}")

    if len(product_df) == 0:
        raise RuntimeError(
            "No product records were extracted. Increase scrape limits or "
            "review robots/network failures."
        )

    model_df = build_product_model_table(product_df)
    market_summary_df = build_market_volume_summary(model_df)
    family_summary_df = build_family_summary(market_summary_df)

    display_table("Scraped page audit", page_df, max_rows=120)
    display_table("Product evidence records", product_df, max_rows=120)
    display_table("Deduplicated product-model table", model_df, max_rows=160)
    display_table("Market-volume summary by category", market_summary_df, max_rows=None)
    display_table("Market-volume summary by family", family_summary_df, max_rows=None)

    plot_market_volume_by_category(
        market_summary_df,
        "observed_unique_active_models",
        "Observed Active Product-Model Volume by Category",
    )
    plot_market_volume_by_category(
        market_summary_df,
        "chao1_estimated_active_models",
        "Chao1 Estimated Active Product-Model Volume by Category",
    )
    plot_family_market_volume(family_summary_df)
    plot_confidence_distribution(model_df)
    plot_coverage_vs_volume(market_summary_df)

    export_outputs(
        query_df=query_df,
        search_df=search_df,
        page_df=page_df,
        product_df=product_df,
        model_df=model_df,
        market_summary_df=market_summary_df,
    )

    elapsed_min = (time.perf_counter() - started_at) / 60.0
    print(f"Workflow complete in {elapsed_min:.2f} min")

    return product_df, model_df, market_summary_df, family_summary_df


product_records, product_models, market_volume, family_market_volume = main()
```
