```python
"""
Build and maintain a curated starter catalog for a Foldable Product Atlas.
"""

import json
import math
import re
import subprocess
import sys
import urllib.parse
from dataclasses import dataclass
from pathlib import Path
from typing import Any, Final

# =============================================================================
# CONTROL KNOBS
# =============================================================================

INSTALL_PACKAGES: Final[bool] = True

REPO_NAME: Final[str] = "foldable-product-atlas"
OUTPUT_ROOT: Final[Path] = Path("/content") / REPO_NAME

MPL_DPI: Final[int] = 250
DISPLAY_MAX_ROWS: Final[int] = 120
DISPLAY_MAX_COLUMNS: Final[int] = 120

EXPORT_REPO_FILES: Final[bool] = True
RUN_LINK_AUDIT: Final[bool] = False
LINK_TIMEOUT_S: Final[float] = 8.0

SCORE_DECIMALS: Final[int] = 3
MIN_SCORE: Final[float] = 0.0
MAX_SCORE: Final[float] = 5.0

FOLDABILITY_WEIGHTS: Final[dict[str, float]] = {
    "portability_gain_score": 0.28,
    "mechanism_novelty_score": 0.19,
    "market_availability_score": 0.17,
    "evidence_quality_score": 0.18,
    "measurement_completeness_score": 0.10,
    "certification_relevance_score": 0.08,
}

REQUIRED_COLUMNS: Final[list[str]] = [
    "product_id",
    "brand",
    "product_name",
    "market_category",
    "transformation_family",
    "transformation_subtype",
    "fold_mechanism",
    "fold_architecture",
    "source_url",
    "source_type",
    "review_status",
]

CANONICAL_COLUMNS: Final[list[str]] = [
    "product_id",
    "brand",
    "product_name",
    "market_category",
    "transformation_family",
    "transformation_subtype",
    "fold_mechanism",
    "fold_architecture",
    "compacting_motion",
    "deployment_state",
    "review_status",
    "release_year",
    "price_usd",
    "display_size_in",
    "mass_g",
    "deployed_length_mm",
    "deployed_width_mm",
    "deployed_height_mm",
    "folded_length_mm",
    "folded_width_mm",
    "folded_height_mm",
    "compression_ratio",
    "claimed_fold_cycles",
    "certification_or_standard",
    "primary_materials",
    "primary_use_case",
    "region_or_market",
    "portability_gain_score",
    "mechanism_novelty_score",
    "market_availability_score",
    "evidence_quality_score",
    "measurement_completeness_score",
    "certification_relevance_score",
    "foldability_score",
    "price_usd_per_score",
    "source_type",
    "source_url",
    "normalized_source_url",
    "source_domain",
    "remarks",
    "missing_measurements",
]

MEASUREMENT_COLUMNS: Final[list[str]] = [
    "price_usd",
    "mass_g",
    "deployed_length_mm",
    "deployed_width_mm",
    "deployed_height_mm",
    "folded_length_mm",
    "folded_width_mm",
    "folded_height_mm",
    "compression_ratio",
    "claimed_fold_cycles",
    "certification_or_standard",
]

TARGET_MARKET_CATEGORIES: Final[list[str]] = [
    "computer peripheral",
    "portable keyboard",
    "foldable computer",
    "foldable smartphone",
    "foldable monitor",
    "folding bicycle",
    "folding e-bike",
    "urban mobility safety",
    "folding watercraft",
    "foldable solar panel",
    "folding drone",
    "portable workstation",
    "folding furniture",
    "collapsible drinkware",
    "collapsible cookware",
    "collapsible storage",
    "collapsible outdoor gear",
    "rollable electronics",
    "inflatable recreation",
    "scientific education",
    "foldable headphones",
]

REVIEW_STATUS_ORDER: Final[list[str]] = [
    "canonical_seed",
    "needs_measurements",
    "needs_source_review",
    "candidate_only",
]

# =============================================================================
# DEPENDENCY INSTALLATION
# =============================================================================


def install_dependencies() -> None:
    """Install lightweight notebook dependencies with uv pip."""
    subprocess.run(
        [sys.executable, "-m", "pip", "install", "-q", "uv"],
        check=True,
    )
    subprocess.run(
        [
            "uv",
            "pip",
            "install",
            "-q",
            "--system",
            "jsonschema",
            "matplotlib",
            "numpy",
            "pandas",
            "requests",
            "tabulate",
        ],
        check=True,
    )


if INSTALL_PACKAGES:
    install_dependencies()

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
from IPython.display import Markdown, display
from jsonschema import Draft202012Validator
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

mpl.rcParams["figure.dpi"] = MPL_DPI
pd.set_option("display.max_rows", DISPLAY_MAX_ROWS)
pd.set_option("display.max_columns", DISPLAY_MAX_COLUMNS)
pd.set_option("display.max_colwidth", 160)
pd.set_option("display.width", 200)

# =============================================================================
# TAXONOMY MODEL
# =============================================================================


@dataclass(frozen=True)
class TaxonomyRule:
    """Keyword rule for assigning a transformation family and subtype."""

    family: str
    subtype: str
    keywords: tuple[str, ...]
    priority: int


TAXONOMY_RULES: Final[list[TaxonomyRule]] = [
    TaxonomyRule(
        family="foldable",
        subtype="flexible-display-fold",
        keywords=(
            "foldable oled",
            "folding oled",
            "flexible oled",
            "foldable display",
            "folding display",
            "ultra-thin glass",
            "book-fold display",
            "foldable smartphone",
            "foldable laptop",
            "foldable computer",
            "foldable monitor",
        ),
        priority=120,
    ),
    TaxonomyRule(
        family="foldable",
        subtype="headphone-hinge-fold",
        keywords=(
            "foldable headphone",
            "foldable headphones",
            "folding headphone",
            "folding headphones",
            "fold-flat",
            "fold flat",
            "foldable design",
            "swivel foldable structure",
            "rotating earcups",
            "rotating ear cups",
            "swiveling earcups",
            "swiveling ear cups",
            "earcup rotation",
            "earcup swivel",
            "folding headband",
            "collapsible headband",
            "fold into carrying case",
            "compact case storage",
        ),
        priority=118,
    ),
    TaxonomyRule(
        family="foldable",
        subtype="rigid-hinge-fold",
        keywords=(
            "hinge",
            "hinged",
            "multi-joint",
            "frame fold",
            "full-size frame folding",
            "folding bicycle",
            "folding e-bike",
            "folding drone",
            "folding stroller",
            "folding chair",
            "folding table",
            "folding treadmill",
        ),
        priority=110,
    ),
    TaxonomyRule(
        family="foldable",
        subtype="origami-panel-fold",
        keywords=(
            "origami",
            "folded panel",
            "corrugated polymer",
            "folding kayak",
            "folding canoe",
            "paper microscope",
            "folded paper",
            "crease pattern",
        ),
        priority=105,
    ),
    TaxonomyRule(
        family="foldable",
        subtype="snap-flexure-fold",
        keywords=(
            "snap-flat",
            "snap flat",
            "flexure",
            "flexible shell",
            "curved flexible shell",
            "bendable shell",
        ),
        priority=100,
    ),
    TaxonomyRule(
        family="collapsible",
        subtype="segmented-shell-collapse",
        keywords=(
            "collapsible helmet",
            "nested shell",
            "segmented collapsible helmet",
            "telescoping rings",
            "segmented hardshell",
            "smart-fold shell",
        ),
        priority=112,
    ),
    TaxonomyRule(
        family="collapsible",
        subtype="soft-vessel-collapse",
        keywords=(
            "collapsible bottle",
            "soft flask",
            "hydration flask",
            "soft bottle",
            "water pouch",
            "volume evacuation",
            "squeeze-collapse",
        ),
        priority=104,
    ),
    TaxonomyRule(
        family="collapsible",
        subtype="telescoping-or-nesting-collapse",
        keywords=(
            "collapsible cup",
            "collapsible bowl",
            "collapsible container",
            "collapsible cookware",
            "telescoping",
            "telescopic",
            "nested",
            "nesting",
        ),
        priority=102,
    ),
    TaxonomyRule(
        family="collapsible",
        subtype="frame-assisted-collapse",
        keywords=(
            "collapsible wagon",
            "collapsible cart",
            "collapsible storage",
            "collapsible crate",
            "collapsible bin",
            "pop-up",
            "pop up",
            "scissor frame",
        ),
        priority=96,
    ),
    TaxonomyRule(
        family="rollable",
        subtype="rollable-sheet",
        keywords=(
            "rollable",
            "roll-up",
            "roll up",
            "rolled sheet",
            "rolltop",
            "roll-top",
        ),
        priority=86,
    ),
    TaxonomyRule(
        family="inflatable",
        subtype="deflation-compaction",
        keywords=(
            "inflatable",
            "deflatable",
            "air mattress",
            "inflatable kayak",
            "inflatable paddle board",
        ),
        priority=80,
    ),
    TaxonomyRule(
        family="modular",
        subtype="flat-pack-or-disassemblable",
        keywords=(
            "flat-pack",
            "flat pack",
            "modular",
            "disassemblable",
            "breakdown",
            "snap together",
        ),
        priority=65,
    ),
]

# =============================================================================
# TEXT AND URL HELPERS
# =============================================================================


def clean_text(value: Any, max_chars: int | None = None) -> str:
    """Return a normalized one-line string."""
    if value is None:
        return ""

    try:
        if pd.isna(value):
            return ""
    except (TypeError, ValueError):
        pass

    text = re.sub(r"\s+", " ", str(value)).strip()

    if max_chars is not None and len(text) > max_chars:
        return f"{text[: max_chars - 1].rstrip()}…"

    return text


def slugify(value: str) -> str:
    """Convert text into a stable lowercase identifier."""
    slug = re.sub(r"[^a-z0-9]+", "-", value.lower().strip())
    return slug.strip("-")


def normalize_url(url: str) -> str:
    """Normalize a source URL for deduplication and source comparison."""
    parsed = urllib.parse.urlparse(clean_text(url))
    scheme = parsed.scheme.lower() or "https"
    netloc = parsed.netloc.lower().removeprefix("www.")
    path = re.sub(r"/+$", "", parsed.path)
    return urllib.parse.urlunparse((scheme, netloc, path, "", "", ""))


def extract_domain(url: str) -> str:
    """Extract the normalized domain from a URL."""
    parsed = urllib.parse.urlparse(clean_text(url))
    return parsed.netloc.lower().removeprefix("www.")


def value_is_present(value: Any) -> bool:
    """Return whether a value contains useful catalog information."""
    if value is None:
        return False

    if isinstance(value, float) and math.isnan(value):
        return False

    text = clean_text(value)
    return text != ""


def parse_numeric(value: Any) -> float:
    """Parse a number from a field, returning NaN when parsing fails."""
    if value_is_present(value) is False:
        return float("nan")

    try:
        return float(value)
    except (TypeError, ValueError):
        return float("nan")


def assign_product_id(record: dict[str, Any]) -> dict[str, Any]:
    """Assign a stable product ID to one record."""
    output = record.copy()
    product_key = f"{output['brand']} {output['product_name']}"
    output.setdefault("product_id", slugify(product_key))
    return output


# =============================================================================
# SEED DATA
# =============================================================================


def build_seed_records() -> list[dict[str, Any]]:
    """Build starter catalog records with a broader category footprint."""
    records: list[dict[str, Any]] = [
        {
            "brand": "Microsoft",
            "product_name": "Surface Arc Mouse",
            "market_category": "computer peripheral",
            "fold_mechanism": "snap-flat flexure body",
            "fold_architecture": "curved flexible shell",
            "compacting_motion": "elastic curvature reversal",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "release_year": 2017,
            "price_usd": 89.99,
            "primary_materials": "polymer shell, electronics",
            "primary_use_case": "portable pointing device",
            "portability_gain_score": 3.5,
            "mechanism_novelty_score": 3.5,
            "market_availability_score": 5.0,
            "evidence_quality_score": 4.5,
            "certification_relevance_score": 1.0,
            "source_type": "official product page",
            "source_url": (
                "https://www.microsoft.com/en-us/d/"
                "surface-arc-mouse-black/8nr554s5qxn7"
            ),
            "remarks": "Snaps flat for travel and arcs for use.",
        },
        {
            "brand": "Closca",
            "product_name": "Loop Helmet",
            "market_category": "urban mobility safety",
            "fold_mechanism": "collapsible telescoping rings",
            "fold_architecture": "nested shell segments",
            "compacting_motion": "telescoping shell collapse",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "release_year": 2019,
            "compression_ratio": 1.82,
            "certification_or_standard": "helmet safety certification claimed",
            "primary_materials": "EPS, polycarbonate",
            "primary_use_case": "bike and scooter helmet",
            "portability_gain_score": 3.8,
            "mechanism_novelty_score": 4.0,
            "market_availability_score": 4.0,
            "evidence_quality_score": 4.0,
            "certification_relevance_score": 5.0,
            "source_type": "official product page",
            "source_url": "https://closca.com/en-us/products/loop-aurora-abyss",
            "remarks": "Collapsible safety product with a nested shell concept.",
        },
        {
            "brand": "Overade",
            "product_name": "Plixi Foldable Helmet",
            "market_category": "urban mobility safety",
            "fold_mechanism": "multi-hinge folding shell",
            "fold_architecture": "segmented hardshell",
            "compacting_motion": "multi-axis shell folding",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "release_year": 2018,
            "compression_ratio": 3.0,
            "certification_or_standard": "CPSC, CE EN1078",
            "primary_materials": "EPS, shell, hinge hardware",
            "primary_use_case": "bike and scooter helmet",
            "portability_gain_score": 4.5,
            "mechanism_novelty_score": 4.5,
            "market_availability_score": 4.0,
            "evidence_quality_score": 4.0,
            "certification_relevance_score": 5.0,
            "source_type": "official product page",
            "source_url": (
                "https://overade.com/en/products/"
                "plixi-foldable-urban-bike-helmet"
            ),
            "remarks": "Folded storage volume claim is roughly three times smaller.",
        },
        {
            "brand": "FEND",
            "product_name": "FEND One Helmet",
            "market_category": "urban mobility safety",
            "fold_mechanism": "patented smart-fold shell",
            "fold_architecture": "segmented collapsible helmet",
            "compacting_motion": "segmented shell folding",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "release_year": 2020,
            "price_usd": 119.00,
            "compression_ratio": 2.0,
            "certification_or_standard": "US CPSC, EU EN1078",
            "primary_materials": "ABS, PC, EPS",
            "primary_use_case": "bike, e-bike, scooter, skateboard helmet",
            "portability_gain_score": 4.2,
            "mechanism_novelty_score": 4.2,
            "market_availability_score": 4.5,
            "evidence_quality_score": 4.0,
            "certification_relevance_score": 5.0,
            "source_type": "official product page",
            "source_url": "https://fend.io/products/fend-helmet",
            "remarks": "Folds by about 50 percent for bag storage.",
        },
        {
            "brand": "Lenovo",
            "product_name": "ThinkPad X1 Fold",
            "market_category": "foldable computer",
            "fold_mechanism": "folding flexible OLED display",
            "fold_architecture": "book-fold display with hinge spine",
            "compacting_motion": "controlled-radius flexible display fold",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "release_year": 2020,
            "display_size_in": 16.3,
            "primary_materials": "flexible OLED, hinge, computing hardware",
            "primary_use_case": "portable computer and tablet workstation",
            "portability_gain_score": 4.0,
            "mechanism_novelty_score": 5.0,
            "market_availability_score": 3.5,
            "evidence_quality_score": 4.5,
            "certification_relevance_score": 1.0,
            "source_type": "official product page",
            "source_url": "https://www.lenovo.com/us/en/thinkpad-x1-fold/",
            "remarks": "Foldable OLED PC with multiple use modes.",
        },
        {
            "brand": "ASUS",
            "product_name": "Zenbook 17 Fold OLED",
            "market_category": "foldable computer",
            "fold_mechanism": "folding flexible OLED display",
            "fold_architecture": "17.3 inch book-fold OLED computer",
            "compacting_motion": "controlled-radius flexible display fold",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "release_year": 2022,
            "display_size_in": 17.3,
            "claimed_fold_cycles": 30000,
            "certification_or_standard": "Intel Evo",
            "primary_materials": "flexible OLED, hinge, computing hardware",
            "primary_use_case": "portable large-screen computing",
            "portability_gain_score": 4.0,
            "mechanism_novelty_score": 5.0,
            "market_availability_score": 3.5,
            "evidence_quality_score": 4.5,
            "certification_relevance_score": 1.5,
            "source_type": "official press release",
            "source_url": "https://www.asus.com/us/news/s7pu8a5n2p8gfrpi/",
            "remarks": "Official release cites 17.3 inch OLED and hinge testing.",
        },
        {
            "brand": "Samsung",
            "product_name": "Galaxy Z Fold",
            "market_category": "foldable smartphone",
            "fold_mechanism": "folding flexible OLED display",
            "fold_architecture": "book-fold smartphone",
            "compacting_motion": "controlled-radius flexible display fold",
            "deployment_state": "commercial",
            "review_status": "needs_source_review",
            "display_size_in": 7.6,
            "primary_materials": "flexible OLED, ultra-thin glass, hinge",
            "primary_use_case": "phone-tablet hybrid computing",
            "portability_gain_score": 4.5,
            "mechanism_novelty_score": 4.5,
            "market_availability_score": 5.0,
            "evidence_quality_score": 4.0,
            "certification_relevance_score": 2.5,
            "source_type": "official product family page",
            "source_url": "https://www.samsung.com/us/smartphones/galaxy-z-fold/",
            "remarks": "Canonical book-fold smartphone family entry.",
        },
        {
            "brand": "Samsung",
            "product_name": "Galaxy Z Flip",
            "market_category": "foldable smartphone",
            "fold_mechanism": "folding flexible OLED display",
            "fold_architecture": "clamshell-fold smartphone",
            "compacting_motion": "controlled-radius flexible display fold",
            "deployment_state": "commercial",
            "review_status": "needs_source_review",
            "primary_materials": "flexible OLED, ultra-thin glass, hinge",
            "primary_use_case": "compact smartphone",
            "portability_gain_score": 4.2,
            "mechanism_novelty_score": 4.4,
            "market_availability_score": 5.0,
            "evidence_quality_score": 4.0,
            "certification_relevance_score": 2.5,
            "source_type": "official product family page",
            "source_url": "https://www.samsung.com/us/smartphones/galaxy-z-flip/",
            "remarks": "Canonical clamshell smartphone family entry.",
        },
        {
            "brand": "Brompton",
            "product_name": "C Line",
            "market_category": "folding bicycle",
            "fold_mechanism": "multi-joint frame fold",
            "fold_architecture": "hinged steel bicycle frame",
            "compacting_motion": "rigid-body frame folding",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "release_year": 1975,
            "price_usd": 1720.00,
            "mass_g": 11299.0,
            "primary_materials": "steel frame, drivetrain, wheels",
            "primary_use_case": "multimodal urban transportation",
            "portability_gain_score": 5.0,
            "mechanism_novelty_score": 4.0,
            "market_availability_score": 5.0,
            "evidence_quality_score": 4.5,
            "certification_relevance_score": 2.0,
            "source_type": "official product page",
            "source_url": "https://us.brompton.com/c-line",
            "remarks": "Canonical commercial folding bicycle benchmark.",
        },
        {
            "brand": "Tern",
            "product_name": "Link D8",
            "market_category": "folding bicycle",
            "fold_mechanism": "multi-joint frame fold",
            "fold_architecture": "hinged aluminum bicycle frame",
            "compacting_motion": "rigid-body frame folding",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "aluminum frame, drivetrain, wheels",
            "primary_use_case": "urban transportation",
            "portability_gain_score": 4.7,
            "mechanism_novelty_score": 3.7,
            "market_availability_score": 4.5,
            "evidence_quality_score": 3.8,
            "certification_relevance_score": 2.0,
            "source_type": "official product page",
            "source_url": "https://www.ternbicycles.com/us/bikes/472/link-d8",
            "remarks": "Useful comparison point against Brompton-style folding geometry.",
        },
        {
            "brand": "Gocycle",
            "product_name": "G4",
            "market_category": "folding e-bike",
            "fold_mechanism": "hinged e-bike frame fold",
            "fold_architecture": "folding electric bicycle frame",
            "compacting_motion": "rigid-body frame folding",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "composite frame, battery, motor, drivetrain",
            "primary_use_case": "electric urban transportation",
            "portability_gain_score": 4.6,
            "mechanism_novelty_score": 4.2,
            "market_availability_score": 4.0,
            "evidence_quality_score": 3.8,
            "certification_relevance_score": 2.2,
            "source_type": "official product page",
            "source_url": "https://gocycle.com/models/g4/",
            "remarks": "Folding e-bike entry with powered drivetrain constraints.",
        },
        {
            "brand": "Montague",
            "product_name": "Paratrooper",
            "market_category": "folding bicycle",
            "fold_mechanism": "patented full-size frame folding system",
            "fold_architecture": "full-size folding mountain bicycle frame",
            "compacting_motion": "rigid-body frame folding",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "aluminum frame, drivetrain, wheels",
            "primary_use_case": "folding mountain bicycle for rugged transport",
            "region_or_market": "US and international",
            "portability_gain_score": 4.4,
            "mechanism_novelty_score": 4.0,
            "market_availability_score": 4.0,
            "evidence_quality_score": 4.2,
            "certification_relevance_score": 2.0,
            "source_type": "official product page",
            "source_url": "https://www.montaguebikes.com/product/paratrooper/",
            "remarks": (
                "Military-derived full-size folding mountain-bike family entry; "
                "add folded dimensions, mass, and frame-size data during review."
            ),
        },
        {
            "brand": "Montague",
            "product_name": "Paratrooper Pro",
            "market_category": "folding bicycle",
            "fold_mechanism": "patented full-size frame folding system",
            "fold_architecture": "upgraded full-size folding mountain bicycle frame",
            "compacting_motion": "rigid-body frame folding",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "aluminum frame, drivetrain, wheels",
            "primary_use_case": "performance folding mountain bicycle",
            "region_or_market": "US and international",
            "portability_gain_score": 4.4,
            "mechanism_novelty_score": 4.0,
            "market_availability_score": 4.0,
            "evidence_quality_score": 4.2,
            "certification_relevance_score": 2.0,
            "source_type": "official product page",
            "source_url": "https://www.montaguebikes.com/product/paratrooper-pro/",
            "remarks": (
                "Performance-oriented Paratrooper variant; useful for comparing "
                "folding-bike geometry across component tiers."
            ),
        },
        {
            "brand": "Montague",
            "product_name": "Paratrooper Elite",
            "market_category": "folding bicycle",
            "fold_mechanism": "patented full-size frame folding system",
            "fold_architecture": "high-performance folding hardtail mountain bicycle frame",
            "compacting_motion": "rigid-body frame folding",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "aluminum frame, drivetrain, wheels",
            "primary_use_case": "technical-trail folding mountain bicycle",
            "region_or_market": "US and international",
            "portability_gain_score": 4.3,
            "mechanism_novelty_score": 4.2,
            "market_availability_score": 3.8,
            "evidence_quality_score": 4.2,
            "certification_relevance_score": 2.0,
            "source_type": "official product page",
            "source_url": "https://www.montaguebikes.com/product/paratrooper-elite/",
            "remarks": (
                "High-end Montague folding mountain-bike anchor; add component, "
                "mass, folded-volume, and deployed geometry measurements."
            ),
        },
        {
            "brand": "Montague",
            "product_name": "Paratrooper Express",
            "market_category": "folding bicycle",
            "fold_mechanism": "patented full-size frame folding system",
            "fold_architecture": "full-size folding mountain bicycle frame",
            "compacting_motion": "rigid-body frame folding",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "aluminum frame, Shimano drivetrain, wheels",
            "primary_use_case": "folding mountain bicycle with broad gearing",
            "region_or_market": "US and international",
            "portability_gain_score": 4.3,
            "mechanism_novelty_score": 3.9,
            "market_availability_score": 3.8,
            "evidence_quality_score": 4.1,
            "certification_relevance_score": 2.0,
            "source_type": "official product page",
            "source_url": "https://www.montaguebikes.com/product/paratrooper-express/",
            "remarks": (
                "Montague folding mountain-bike variant with an 18-speed drivetrain; "
                "add geometry and folded-dimension measurements during review."
            ),
        },
        {
            "brand": "Montague",
            "product_name": "Paratrooper Highline",
            "market_category": "folding bicycle",
            "fold_mechanism": "patented full-size frame folding system",
            "fold_architecture": "27.5 inch wheel folding mountain bicycle frame",
            "compacting_motion": "rigid-body frame folding",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "aluminum frame, suspension fork, drivetrain, wheels",
            "primary_use_case": "cross-country folding mountain bicycle",
            "region_or_market": "US and international",
            "portability_gain_score": 4.3,
            "mechanism_novelty_score": 4.1,
            "market_availability_score": 3.8,
            "evidence_quality_score": 4.1,
            "certification_relevance_score": 2.0,
            "source_type": "official product page",
            "source_url": "https://www.montaguebikes.com/product/paratrooper-highline/",
            "remarks": (
                "Rugged 27.5 inch Montague folding mountain-bike entry; add mass, "
                "folded footprint, and suspension component data."
            ),
        },
        {
            "brand": "Montague",
            "product_name": "Navigator",
            "market_category": "folding bicycle",
            "fold_mechanism": "patented full-size frame folding system",
            "fold_architecture": "700c-wheel multi-use folding bicycle frame",
            "compacting_motion": "rigid-body frame folding",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "aluminum frame, drivetrain, wheels",
            "primary_use_case": "commuting, fitness, and mixed-surface riding",
            "region_or_market": "US and international",
            "portability_gain_score": 4.4,
            "mechanism_novelty_score": 4.0,
            "market_availability_score": 4.0,
            "evidence_quality_score": 4.2,
            "certification_relevance_score": 2.0,
            "source_type": "official product page",
            "source_url": "https://www.montaguebikes.com/product/navigator/",
            "remarks": (
                "Full-size 700c folding-bike entry; useful contrast against "
                "small-wheel folding bicycle architectures."
            ),
        },
        {
            "brand": "Montague",
            "product_name": "Boston",
            "market_category": "folding bicycle",
            "fold_mechanism": "patented full-size frame folding system",
            "fold_architecture": "minimalist single-speed full-size folding bicycle frame",
            "compacting_motion": "rigid-body frame folding",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "aluminum frame, drivetrain, wheels",
            "primary_use_case": "simple urban folding bicycle transport",
            "region_or_market": "US and international",
            "portability_gain_score": 4.2,
            "mechanism_novelty_score": 3.8,
            "market_availability_score": 4.0,
            "evidence_quality_score": 4.1,
            "certification_relevance_score": 2.0,
            "source_type": "official product page",
            "source_url": "https://www.montaguebikes.com/product/boston/",
            "remarks": (
                "Simplified full-size Montague folding-bike entry; useful baseline "
                "for separating mechanism complexity from drivetrain complexity."
            ),
        },
        {
            "brand": "Montague",
            "product_name": "M-E1",
            "market_category": "folding e-bike",
            "fold_mechanism": "patented full-size e-bike frame folding system",
            "fold_architecture": "full-size folding electric bicycle frame",
            "compacting_motion": (
                "rigid-body frame folding with integrated electric-assist hardware"
            ),
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": (
                "aluminum frame, battery, mid-drive motor, drivetrain, wheels"
            ),
            "primary_use_case": (
                "folding electric bicycle for commuting, RV, aviation, and marine storage"
            ),
            "region_or_market": "US and international",
            "portability_gain_score": 4.5,
            "mechanism_novelty_score": 4.4,
            "market_availability_score": 4.0,
            "evidence_quality_score": 4.3,
            "certification_relevance_score": 2.5,
            "source_type": "official product page",
            "source_url": "https://www.montaguebikes.com/product/m-e1/",
            "remarks": (
                "Full-size folding e-bike entry; review should capture battery "
                "mass, motor type, folded dimensions, deployed dimensions, and "
                "transport use-case constraints."
            ),
        },
        {
            "brand": "Oru Kayak",
            "product_name": "Inlet",
            "market_category": "folding watercraft",
            "fold_mechanism": "origami panel fold",
            "fold_architecture": "folded corrugated polymer hull",
            "compacting_motion": "crease-pattern hull deployment",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "release_year": 2012,
            "mass_g": 8165.0,
            "primary_materials": "corrugated polypropylene",
            "primary_use_case": "portable recreational kayak",
            "portability_gain_score": 5.0,
            "mechanism_novelty_score": 5.0,
            "market_availability_score": 4.5,
            "evidence_quality_score": 4.0,
            "certification_relevance_score": 2.0,
            "source_type": "official product page",
            "source_url": (
                "https://www.orukayak.com/products/"
                "oru-kayak-portable-folding-lightweight-recreational-kayak-for-beginners"
            ),
            "remarks": "Origami-inspired box-to-boat product class.",
        },
        {
            "brand": "Tucktec",
            "product_name": "Folding Kayak",
            "market_category": "folding watercraft",
            "fold_mechanism": "folded polymer hull",
            "fold_architecture": "flat-sheet folding kayak hull",
            "compacting_motion": "crease-pattern hull deployment",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "polymer sheet, fastening hardware",
            "primary_use_case": "portable recreational kayak",
            "portability_gain_score": 4.7,
            "mechanism_novelty_score": 4.4,
            "market_availability_score": 4.0,
            "evidence_quality_score": 3.8,
            "certification_relevance_score": 2.0,
            "source_type": "official product page",
            "source_url": "https://tucktec.com/products/tucktec-folding-kayak",
            "remarks": "Complementary folding watercraft architecture.",
        },
        {
            "brand": "Foldscope Instruments",
            "product_name": "Foldscope Paper Microscope",
            "market_category": "scientific education",
            "fold_mechanism": "origami paper assembly",
            "fold_architecture": "folded paper optical instrument",
            "compacting_motion": "crease-pattern optical structure assembly",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "release_year": 2014,
            "price_usd": 10.99,
            "primary_materials": "paper, lens, adhesive, optional LED",
            "primary_use_case": "portable microscopy education",
            "portability_gain_score": 5.0,
            "mechanism_novelty_score": 5.0,
            "market_availability_score": 4.5,
            "evidence_quality_score": 5.0,
            "certification_relevance_score": 1.0,
            "source_type": "official product page",
            "source_url": "https://foldscope.com/",
            "remarks": "Low-cost foldable paper microscope.",
        },
        {
            "brand": "Logitech",
            "product_name": "Casa Pop-Up Desk",
            "market_category": "portable workstation",
            "fold_mechanism": "case-to-stand fold",
            "fold_architecture": "folding book-like laptop stand kit",
            "compacting_motion": "case conversion into ergonomic stand",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "release_year": 2023,
            "price_usd": 179.99,
            "primary_materials": "fabric case, keyboard, touchpad, stand",
            "primary_use_case": "compact home-office workstation",
            "portability_gain_score": 4.0,
            "mechanism_novelty_score": 3.5,
            "market_availability_score": 4.0,
            "evidence_quality_score": 4.0,
            "certification_relevance_score": 1.0,
            "source_type": "official product page",
            "source_url": "https://www.logitech.com/en-us/shop/p/casa-pop-up-desk",
            "remarks": "Desk kit packs into a book-like case.",
        },
        {
            "brand": "Jackery",
            "product_name": "SolarSaga 100W",
            "market_category": "foldable solar panel",
            "fold_mechanism": "panel bi-fold",
            "fold_architecture": "folding photovoltaic panel",
            "compacting_motion": "hinged planar panel folding",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "photovoltaic cells, laminate, fabric, wiring",
            "primary_use_case": "portable power generation",
            "portability_gain_score": 4.2,
            "mechanism_novelty_score": 3.2,
            "market_availability_score": 4.5,
            "evidence_quality_score": 3.8,
            "certification_relevance_score": 2.5,
            "source_type": "official product page",
            "source_url": (
                "https://www.jackery.com/products/solarsaga-100w-solar-panel"
            ),
            "remarks": "Portable panel category anchor.",
        },
        {
            "brand": "Goal Zero",
            "product_name": "Nomad Solar Panel",
            "market_category": "foldable solar panel",
            "fold_mechanism": "folding photovoltaic panel",
            "fold_architecture": "multi-panel folding solar charger",
            "compacting_motion": "hinged planar panel folding",
            "deployment_state": "commercial",
            "review_status": "needs_source_review",
            "primary_materials": "photovoltaic cells, fabric, wiring",
            "primary_use_case": "portable power generation",
            "portability_gain_score": 4.1,
            "mechanism_novelty_score": 3.2,
            "market_availability_score": 4.0,
            "evidence_quality_score": 3.5,
            "certification_relevance_score": 2.5,
            "source_type": "official product family page",
            "source_url": "https://www.goalzero.com/collections/portable-solar-panels",
            "remarks": "Solar panel family entry for later product-level expansion.",
        },
        {
            "brand": "DJI",
            "product_name": "Mavic Series",
            "market_category": "folding drone",
            "fold_mechanism": "folding arm mechanism",
            "fold_architecture": "compact quadcopter with folding arms",
            "compacting_motion": "rotating arm fold",
            "deployment_state": "commercial",
            "review_status": "needs_source_review",
            "primary_materials": "polymer shell, motors, electronics, battery",
            "primary_use_case": "portable aerial imaging",
            "portability_gain_score": 4.0,
            "mechanism_novelty_score": 3.8,
            "market_availability_score": 4.5,
            "evidence_quality_score": 3.5,
            "certification_relevance_score": 3.0,
            "source_type": "official product family page",
            "source_url": "https://www.dji.com/products/camera-drones",
            "remarks": "Use a family entry until a model-specific row is reviewed.",
        },
        {
            "brand": "MOFT",
            "product_name": "Snap Phone Stand and Wallet",
            "market_category": "portable workstation",
            "fold_mechanism": "thin origami stand fold",
            "fold_architecture": "folding phone support accessory",
            "compacting_motion": "thin-panel crease deployment",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "polymer, vegan leather, magnets",
            "primary_use_case": "portable phone stand and grip",
            "portability_gain_score": 3.7,
            "mechanism_novelty_score": 3.9,
            "market_availability_score": 4.2,
            "evidence_quality_score": 3.8,
            "certification_relevance_score": 1.0,
            "source_type": "official product page",
            "source_url": (
                "https://www.moft.us/products/"
                "moft-snap-on-phone-stand-wallet-magsafe-compatible"
            ),
            "remarks": "Compact crease-based accessory architecture.",
        },
        {
            "brand": "Sea to Summit",
            "product_name": "X-Cup",
            "market_category": "collapsible drinkware",
            "fold_mechanism": "collapsible nested silicone rings",
            "fold_architecture": "telescoping drink cup",
            "compacting_motion": "axial nesting collapse",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "silicone",
            "primary_use_case": "camping drinkware",
            "portability_gain_score": 4.2,
            "mechanism_novelty_score": 3.5,
            "market_availability_score": 4.5,
            "evidence_quality_score": 3.8,
            "certification_relevance_score": 1.5,
            "source_type": "official product page",
            "source_url": "https://seatosummit.com/products/x-cup",
            "remarks": "Useful telescoping cookware category anchor.",
        },
        {
            "brand": "Stojo",
            "product_name": "Collapsible Cup",
            "market_category": "collapsible drinkware",
            "fold_mechanism": "collapsible silicone cup",
            "fold_architecture": "nested elastomer cup",
            "compacting_motion": "axial nesting collapse",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "silicone, polymer lid",
            "primary_use_case": "portable drinkware",
            "portability_gain_score": 4.0,
            "mechanism_novelty_score": 3.3,
            "market_availability_score": 4.2,
            "evidence_quality_score": 3.5,
            "certification_relevance_score": 1.5,
            "source_type": "official product page",
            "source_url": "https://stojo.co/products/cup",
            "remarks": "Everyday collapsible drinkware entry.",
        },
        {
            "brand": "HydraPak",
            "product_name": "Stow",
            "market_category": "collapsible outdoor gear",
            "fold_mechanism": "soft-vessel collapse",
            "fold_architecture": "flexible hydration bottle",
            "compacting_motion": "volume evacuation and flattening",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "TPU, cap hardware",
            "primary_use_case": "portable hydration",
            "portability_gain_score": 4.4,
            "mechanism_novelty_score": 3.4,
            "market_availability_score": 4.2,
            "evidence_quality_score": 3.6,
            "certification_relevance_score": 1.5,
            "source_type": "official product page",
            "source_url": "https://hydrapak.com/products/stow-500-ml-2021",
            "remarks": "Soft-vessel collapsible bottle entry.",
        },
        {
            "brand": "Vapur",
            "product_name": "Anti-Bottle",
            "market_category": "collapsible outdoor gear",
            "fold_mechanism": "soft-vessel collapse",
            "fold_architecture": "flexible water bottle pouch",
            "compacting_motion": "volume evacuation and rolling or flattening",
            "deployment_state": "commercial",
            "review_status": "needs_source_review",
            "primary_materials": "flexible multilayer film, cap",
            "primary_use_case": "portable hydration",
            "portability_gain_score": 4.4,
            "mechanism_novelty_score": 3.5,
            "market_availability_score": 4.0,
            "evidence_quality_score": 3.2,
            "certification_relevance_score": 1.5,
            "source_type": "official site",
            "source_url": "https://vapur.us/",
            "remarks": "Soft-bottle family entry for manual product-level expansion.",
        },
        {
            "brand": "Helinox",
            "product_name": "Chair One",
            "market_category": "folding furniture",
            "fold_mechanism": "collapsible pole frame",
            "fold_architecture": "shock-corded frame plus fabric seat",
            "compacting_motion": "frame disassembly and fabric folding",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "aluminum poles, textile seat",
            "primary_use_case": "portable outdoor seating",
            "portability_gain_score": 4.6,
            "mechanism_novelty_score": 3.5,
            "market_availability_score": 4.5,
            "evidence_quality_score": 3.8,
            "certification_relevance_score": 1.5,
            "source_type": "official product page",
            "source_url": "https://helinox.com/products/chair-one",
            "remarks": "Collapsible frame product with load-bearing textile surface.",
        },
        {
            "brand": "Lifetime",
            "product_name": "Folding Table",
            "market_category": "folding furniture",
            "fold_mechanism": "center-fold table hinge",
            "fold_architecture": "folding rigid tabletop with collapsible legs",
            "compacting_motion": "hinged tabletop and leg collapse",
            "deployment_state": "commercial",
            "review_status": "needs_source_review",
            "primary_materials": "polyethylene, steel",
            "primary_use_case": "portable furniture",
            "portability_gain_score": 3.8,
            "mechanism_novelty_score": 2.6,
            "market_availability_score": 4.5,
            "evidence_quality_score": 3.2,
            "certification_relevance_score": 1.0,
            "source_type": "official product family page",
            "source_url": "https://www.lifetime.com/folding-tables",
            "remarks": "Common rigid-hinge furniture category anchor.",
        },
        {
            "brand": "LG",
            "product_name": "SIGNATURE OLED R",
            "market_category": "rollable electronics",
            "fold_mechanism": "rollable flexible OLED panel",
            "fold_architecture": "rollable television display",
            "compacting_motion": "flexible sheet winding",
            "deployment_state": "commercial or limited commercial",
            "review_status": "needs_source_review",
            "primary_materials": "flexible OLED stack, roll mechanism, base",
            "primary_use_case": "deployable large-area display",
            "portability_gain_score": 3.5,
            "mechanism_novelty_score": 5.0,
            "market_availability_score": 2.5,
            "evidence_quality_score": 3.5,
            "certification_relevance_score": 1.5,
            "source_type": "official product family page",
            "source_url": (
                "https://www.lg.com/us/tvs/"
                "lg-oled65r1pua-signature-oled-4k-tv"
            ),
            "remarks": "Rollable display category anchor, requiring availability review.",
        },
        {
            "brand": "Intex",
            "product_name": "Dura-Beam Air Mattress",
            "market_category": "inflatable recreation",
            "fold_mechanism": "deflation compaction",
            "fold_architecture": "inflatable textile-polymer mattress",
            "compacting_motion": "air evacuation and folded storage",
            "deployment_state": "commercial",
            "review_status": "needs_source_review",
            "primary_materials": "polymer, internal fibers, valve",
            "primary_use_case": "portable sleeping surface",
            "portability_gain_score": 4.0,
            "mechanism_novelty_score": 2.7,
            "market_availability_score": 4.5,
            "evidence_quality_score": 3.3,
            "certification_relevance_score": 1.5,
            "source_type": "official product family page",
            "source_url": "https://intexcorp.com/products/airbeds/",
            "remarks": "Inflatable compaction category anchor.",
        },
        {
            "brand": "Bose",
            "product_name": "QuietComfort Ultra Headphones (2nd Gen)",
            "market_category": "foldable headphones",
            "fold_mechanism": "fold-flat earcup and headband hinge system",
            "fold_architecture": "over-ear noise-cancelling headphones with compact case storage",
            "compacting_motion": "earcup rotation and headband folding into carrying case",
            "deployment_state": "commercial",
            "review_status": "needs_measurements",
            "primary_materials": "polymer, metal hinge hardware, acoustic drivers, electronics",
            "primary_use_case": "portable active-noise-cancelling listening",
            "region_or_market": "US and international",
            "portability_gain_score": 3.8,
            "mechanism_novelty_score": 2.8,
            "market_availability_score": 4.7,
            "evidence_quality_score": 4.5,
            "certification_relevance_score": 1.5,
            "source_type": "official product page",
            "source_url": (
                "https://www.bose.com/p/headphones/"
                "bose-quietcomfort-ultra-headphones-2nd-gen/"
                "QCUH2-HEADPHONEARN.html"
            ),
            "remarks": (
                "Premium Bose over-ear entry; add case dimensions, folded dimensions, "
                "mass, and hinge-cycle evidence if available."
            ),
        },
        {
            "brand": "Beats",
            "product_name": "Solo 4",
            "market_category": "foldable headphones",
            "fold_mechanism": "folding on-ear headphone hinge",
            "fold_architecture": "compact on-ear wireless headphone with foldable travel form",
            "compacting_motion": "earcup and headband hinge folding",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "primary_materials": "polymer, metal hinge hardware, acoustic drivers, electronics",
            "primary_use_case": "portable everyday wireless listening",
            "region_or_market": "US and international",
            "portability_gain_score": 4.0,
            "mechanism_novelty_score": 2.6,
            "market_availability_score": 5.0,
            "evidence_quality_score": 4.5,
            "certification_relevance_score": 1.5,
            "source_type": "official product page",
            "source_url": "https://www.beatsbydre.com/headphones/solo4-wireless",
            "remarks": (
                "Official page describes a compact foldable design with soft "
                "protective case; useful on-ear headphone baseline."
            ),
        },
        {
            "brand": "Beats",
            "product_name": "Studio Pro",
            "market_category": "foldable headphones",
            "fold_mechanism": "folding over-ear headphone hinge",
            "fold_architecture": "over-ear active-noise-cancelling headphone with travel hinge geometry",
            "compacting_motion": "earcup and headband hinge folding",
            "deployment_state": "commercial",
            "review_status": "needs_source_review",
            "primary_materials": "polymer, metal hinge hardware, acoustic drivers, electronics",
            "primary_use_case": "portable premium wireless listening",
            "region_or_market": "US and international",
            "portability_gain_score": 3.7,
            "mechanism_novelty_score": 2.7,
            "market_availability_score": 5.0,
            "evidence_quality_score": 4.0,
            "certification_relevance_score": 1.5,
            "source_type": "official product page",
            "source_url": "https://www.beatsbydre.com/headphones/studio-pro-wireless",
            "remarks": (
                "Popular Beats over-ear entry; keep under source review until "
                "folded dimensions and official fold-language are verified."
            ),
        },
        {
            "brand": "Marshall",
            "product_name": "Major V",
            "market_category": "foldable headphones",
            "fold_mechanism": "rugged foldable on-ear headphone hinge",
            "fold_architecture": "on-ear wireless headphone with compact storage geometry",
            "compacting_motion": "earcup and headband hinge folding",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "primary_materials": "polymer, metal hinge hardware, acoustic drivers, electronics",
            "primary_use_case": "portable on-ear wireless listening",
            "region_or_market": "US and international",
            "portability_gain_score": 4.2,
            "mechanism_novelty_score": 2.8,
            "market_availability_score": 4.7,
            "evidence_quality_score": 4.7,
            "certification_relevance_score": 1.5,
            "source_type": "official product page",
            "source_url": "https://www.marshall.com/us/en/product/major-v",
            "remarks": (
                "Officially positioned as a foldable on-ear headphone design; "
                "strong canonical seed for compact headphone geometry."
            ),
        },
        {
            "brand": "Sony",
            "product_name": "WH-1000XM4",
            "market_category": "foldable headphones",
            "fold_mechanism": "swivel foldable over-ear headphone hinge",
            "fold_architecture": "over-ear wireless noise-cancelling headphone with travel hinge geometry",
            "compacting_motion": "earcup swivel inward into compact carrying case",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "primary_materials": "polymer, metal hinge hardware, acoustic drivers, electronics",
            "primary_use_case": "portable active-noise-cancelling listening",
            "region_or_market": "US and international",
            "portability_gain_score": 3.9,
            "mechanism_novelty_score": 2.8,
            "market_availability_score": 4.7,
            "evidence_quality_score": 4.5,
            "certification_relevance_score": 1.5,
            "source_type": "official product page",
            "source_url": "https://www.sony.com/lr/electronics/headband-headphones/wh-1000xm4",
            "remarks": (
                "Official page describes a swivel foldable structure where earcups "
                "swivel inward for compact case packing."
            ),
        },
        {
            "brand": "Audio-Technica",
            "product_name": "ATH-M50xBT2",
            "market_category": "foldable headphones",
            "fold_mechanism": "fold-flat studio-monitor headphone hinge",
            "fold_architecture": "wireless over-ear monitor headphone with 90 degree swiveling earcups",
            "compacting_motion": "earcup swivel and hinge folding for storage",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "primary_materials": "polymer, metal hinge hardware, acoustic drivers, electronics",
            "primary_use_case": "portable studio-monitor-style wireless listening",
            "region_or_market": "US and international",
            "portability_gain_score": 3.8,
            "mechanism_novelty_score": 2.6,
            "market_availability_score": 4.5,
            "evidence_quality_score": 4.5,
            "certification_relevance_score": 1.5,
            "source_type": "official press release",
            "source_url": (
                "https://www.audio-technica.com/en-us/press/"
                "audio-technica-updates-flagship-wireless-ath-m50x-model-"
                "with-ath-m50xbt2-wireless-over-ear-headphones/"
            ),
            "remarks": (
                "Official Audio-Technica press release describes fold-flat design "
                "with 90 degree swiveling earcups."
            ),
        },
        {
            "brand": "Audio-Technica",
            "product_name": "ATH-M50x",
            "market_category": "foldable headphones",
            "fold_mechanism": "collapsible fold-flat studio-monitor headphone hinge",
            "fold_architecture": "wired over-ear monitor headphone with 90 degree swiveling earcups",
            "compacting_motion": "earcup swivel and hinge folding for transport",
            "deployment_state": "commercial",
            "review_status": "canonical_seed",
            "primary_materials": "polymer, metal hinge hardware, acoustic drivers",
            "primary_use_case": "portable professional monitoring and listening",
            "region_or_market": "US and international",
            "portability_gain_score": 3.8,
            "mechanism_novelty_score": 2.6,
            "market_availability_score": 4.8,
            "evidence_quality_score": 4.5,
            "certification_relevance_score": 1.5,
            "source_type": "official product page",
            "source_url": "https://www.audio-technica.com/en-eu/ath-m50x",
            "remarks": (
                "Official page describes 90 degree swiveling earcups and a "
                "collapsible fold-flat form for travel."
            ),
        },
        {
            "brand": "Jabra",
            "product_name": "Elite 85h",
            "market_category": "foldable headphones",
            "fold_mechanism": "rotating earcup fold-to-off mechanism",
            "fold_architecture": "over-ear wireless headphone with fold-state power behavior",
            "compacting_motion": "earcup rotation into folded storage state",
            "deployment_state": "commercial or legacy commercial",
            "review_status": "canonical_seed",
            "primary_materials": "polymer, metal hinge hardware, acoustic drivers, electronics",
            "primary_use_case": "portable active-noise-cancelling listening",
            "region_or_market": "US and international",
            "portability_gain_score": 3.7,
            "mechanism_novelty_score": 3.0,
            "market_availability_score": 3.4,
            "evidence_quality_score": 4.2,
            "certification_relevance_score": 1.5,
            "source_type": "official support page",
            "source_url": (
                "https://www.jabra.com/supportpages/jabra-elite-85h/"
                "100-99030004-02/faq/how-do-i-turn-on_off-my-jabra-elite-85h"
            ),
            "remarks": (
                "Official support page links earcup rotation and folded state to "
                "power behavior, making it mechanism-relevant."
            ),
        },
    ]

    return [assign_product_id(record) for record in records]


# =============================================================================
# TAXONOMY CLASSIFICATION
# =============================================================================


def classify_transformation(record: dict[str, Any]) -> dict[str, str]:
    """Infer transformation family and subtype from mechanism text."""
    combined_text = " | ".join(
        clean_text(record.get(key))
        for key in (
            "market_category",
            "product_name",
            "fold_mechanism",
            "fold_architecture",
            "compacting_motion",
            "remarks",
        )
    ).lower()

    matches: list[TaxonomyRule] = []
    for rule in TAXONOMY_RULES:
        keyword_hit = any(
            keyword.lower() in combined_text for keyword in rule.keywords
        )
        if keyword_hit:
            matches.append(rule)

    if len(matches) == 0:
        return {
            "transformation_family": "unclassified",
            "transformation_subtype": "unclassified",
        }

    matches.sort(key=lambda rule: rule.priority, reverse=True)
    best_rule = matches[0]
    family_set = {rule.family for rule in matches}

    family = "hybrid" if len(family_set) > 1 else best_rule.family

    return {
        "transformation_family": family,
        "transformation_subtype": best_rule.subtype,
    }


def enrich_records(records: list[dict[str, Any]]) -> list[dict[str, Any]]:
    """Add derived taxonomy and source fields to raw records."""
    enriched: list[dict[str, Any]] = []

    for record in records:
        output = record.copy()
        classification = classify_transformation(output)
        output.setdefault(
            "transformation_family",
            classification["transformation_family"],
        )
        output.setdefault(
            "transformation_subtype",
            classification["transformation_subtype"],
        )

        normalized_url = normalize_url(clean_text(output.get("source_url")))
        output["normalized_source_url"] = normalized_url
        output["source_domain"] = extract_domain(normalized_url)

        for measurement in MEASUREMENT_COLUMNS:
            output.setdefault(measurement, np.nan)

        output.setdefault("region_or_market", "various")
        enriched.append(output)

    return enriched


# =============================================================================
# VALIDATION AND SCORING
# =============================================================================


def build_schema() -> dict[str, Any]:
    """Build a JSON Schema for the product catalog."""
    numeric_score_schema = {
        "type": "number",
        "minimum": MIN_SCORE,
        "maximum": MAX_SCORE,
    }

    return {
        "$schema": "https://json-schema.org/draft/2020-12/schema",
        "title": "Foldable Product Catalog Record",
        "type": "object",
        "required": REQUIRED_COLUMNS,
        "properties": {
            "product_id": {"type": "string", "minLength": 3},
            "brand": {"type": "string", "minLength": 1},
            "product_name": {"type": "string", "minLength": 1},
            "market_category": {"type": "string", "minLength": 1},
            "transformation_family": {"type": "string", "minLength": 1},
            "transformation_subtype": {"type": "string", "minLength": 1},
            "fold_mechanism": {"type": "string", "minLength": 1},
            "fold_architecture": {"type": "string", "minLength": 1},
            "source_url": {"type": "string", "format": "uri"},
            "source_type": {"type": "string", "minLength": 1},
            "review_status": {"type": "string", "enum": REVIEW_STATUS_ORDER},
            "portability_gain_score": numeric_score_schema,
            "mechanism_novelty_score": numeric_score_schema,
            "market_availability_score": numeric_score_schema,
            "evidence_quality_score": numeric_score_schema,
            "certification_relevance_score": numeric_score_schema,
        },
        "additionalProperties": True,
    }


def validate_records(records: list[dict[str, Any]]) -> list[str]:
    """Validate records against schema, uniqueness, and source constraints."""
    validator = Draft202012Validator(build_schema())
    errors: list[str] = []

    product_ids = [clean_text(record.get("product_id")) for record in records]
    duplicated_ids = sorted(
        product_id
        for product_id in set(product_ids)
        if product_ids.count(product_id) > 1
    )

    for product_id in duplicated_ids:
        errors.append(f"Duplicate product_id found: {product_id}")

    normalized_urls = [
        clean_text(record.get("normalized_source_url")) for record in records
    ]
    duplicated_urls = sorted(
        url for url in set(normalized_urls) if normalized_urls.count(url) > 1
    )

    for url in duplicated_urls:
        errors.append(f"Duplicate normalized source URL found: {url}")

    for index, record in enumerate(records):
        for error in sorted(validator.iter_errors(record), key=str):
            errors.append(f"Row {index}: {error.message}")

        parsed = urllib.parse.urlparse(clean_text(record.get("source_url")))
        if parsed.scheme not in {"http", "https"}:
            errors.append(
                f"Row {index}: source_url must start with http or https"
            )

    return errors


def compute_measurement_completeness(row: pd.Series) -> float:
    """Compute measurement completeness on a 0 to 5 scale."""
    present_count = sum(
        value_is_present(row.get(column)) for column in MEASUREMENT_COLUMNS
    )
    return round(MAX_SCORE * present_count / len(MEASUREMENT_COLUMNS), SCORE_DECIMALS)


def compute_missing_measurements(row: pd.Series) -> str:
    """List missing measurement fields for review."""
    missing = [
        column
        for column in MEASUREMENT_COLUMNS
        if value_is_present(row.get(column)) is False
    ]
    return ", ".join(missing)


def compute_foldability_scores(df: pd.DataFrame) -> pd.DataFrame:
    """Compute scalar foldability and value scores."""
    scored_df = df.copy()

    scored_df["measurement_completeness_score"] = scored_df.apply(
        compute_measurement_completeness,
        axis=1,
    )
    scored_df["missing_measurements"] = scored_df.apply(
        compute_missing_measurements,
        axis=1,
    )

    score = np.zeros(len(scored_df), dtype=float)
    for column, weight in FOLDABILITY_WEIGHTS.items():
        values = scored_df[column].astype(float).to_numpy()
        score += weight * values

    scored_df["foldability_score"] = np.round(score, SCORE_DECIMALS)

    price = pd.to_numeric(scored_df["price_usd"], errors="coerce")
    denominator = scored_df["foldability_score"].replace(0.0, np.nan)
    scored_df["price_usd_per_score"] = np.round(price / denominator, 2)

    return scored_df


def build_catalog_dataframe() -> pd.DataFrame:
    """Build, validate, score, and order the seed catalog."""
    records = enrich_records(build_seed_records())
    validation_errors = validate_records(records)

    if len(validation_errors) > 0:
        print("Catalog validation errors were found.")
        for error in validation_errors:
            print(f"- {error}")
        raise ValueError("Catalog validation failed.")

    catalog_df = pd.DataFrame(records)
    catalog_df = compute_foldability_scores(catalog_df)

    for column in CANONICAL_COLUMNS:
        if column not in catalog_df.columns:
            catalog_df[column] = np.nan

    return catalog_df[CANONICAL_COLUMNS].sort_values(
        [
            "review_status",
            "transformation_family",
            "market_category",
            "brand",
            "product_name",
        ],
        kind="stable",
    ).reset_index(drop=True)


# =============================================================================
# AUDIT, COVERAGE, AND REVIEW TABLES
# =============================================================================


def build_http_session() -> requests.Session:
    """Create a retry-enabled HTTP session for optional link audits."""
    retry_strategy = Retry(
        total=2,
        connect=2,
        read=2,
        backoff_factor=0.6,
        status_forcelist=(429, 500, 502, 503, 504),
        allowed_methods=("GET", "HEAD"),
    )

    adapter = HTTPAdapter(max_retries=retry_strategy)
    session = requests.Session()
    session.headers.update({"User-Agent": "foldable-product-atlas-audit/0.4"})
    session.mount("https://", adapter)
    session.mount("http://", adapter)
    return session


def audit_links(df: pd.DataFrame) -> pd.DataFrame:
    """Check whether source URLs respond, when enabled."""
    session = build_http_session()
    audit_rows: list[dict[str, Any]] = []

    for _, row in df.iterrows():
        source_url = clean_text(row["source_url"])
        status_code: int | None = None
        ok = False
        error_message = ""

        try:
            response = session.head(
                source_url,
                timeout=LINK_TIMEOUT_S,
                allow_redirects=True,
            )
            if response.status_code >= 400:
                response = session.get(source_url, timeout=LINK_TIMEOUT_S)
            status_code = response.status_code
            ok = 200 <= response.status_code < 400
        except requests.RequestException as exc:
            error_message = str(exc)

        audit_rows.append(
            {
                "product_id": row["product_id"],
                "source_domain": row["source_domain"],
                "status_code": status_code,
                "ok": ok,
                "error_message": error_message,
                "source_url": source_url,
            }
        )

    return pd.DataFrame(audit_rows)


def build_coverage_table(df: pd.DataFrame) -> pd.DataFrame:
    """Compare current catalog coverage against target market categories."""
    rows: list[dict[str, Any]] = []

    for category in TARGET_MARKET_CATEGORIES:
        category_df = df[df["market_category"] == category]
        rows.append(
            {
                "target_market_category": category,
                "covered": len(category_df) > 0,
                "record_count": len(category_df),
                "families_present": ", ".join(
                    sorted(category_df["transformation_family"].dropna().unique())
                ),
            }
        )

    return pd.DataFrame(rows)


def build_review_queue(df: pd.DataFrame) -> pd.DataFrame:
    """Build a triage queue for records needing source or measurement review."""
    review_df = df[df["review_status"] != "canonical_seed"].copy()

    review_df["review_priority_score"] = (
        (MAX_SCORE - review_df["measurement_completeness_score"].astype(float))
        + (MAX_SCORE - review_df["evidence_quality_score"].astype(float))
        + 0.5 * (MAX_SCORE - review_df["market_availability_score"].astype(float))
    ).round(SCORE_DECIMALS)

    return review_df.sort_values(
        ["review_priority_score", "review_status"],
        ascending=[False, True],
        kind="stable",
    )[
        [
            "product_id",
            "brand",
            "product_name",
            "review_status",
            "review_priority_score",
            "market_category",
            "transformation_family",
            "transformation_subtype",
            "measurement_completeness_score",
            "evidence_quality_score",
            "missing_measurements",
            "source_type",
            "source_url",
        ]
    ].reset_index(drop=True)


def build_query_seed_table(coverage_df: pd.DataFrame) -> pd.DataFrame:
    """Build query seeds for a companion discovery notebook."""
    rows: list[dict[str, str]] = []

    for _, row in coverage_df.iterrows():
        category = clean_text(row["target_market_category"])
        intent = "expand" if bool(row["covered"]) else "fill_gap"

        rows.append(
            {
                "intent": intent,
                "target_market_category": category,
                "query": f"{category} foldable collapsible product official",
            }
        )

    rows.extend(
        [
            {
                "intent": "expand",
                "target_market_category": "folding bicycle",
                "query": "Montague folding bicycle official product",
            },
            {
                "intent": "expand",
                "target_market_category": "folding e-bike",
                "query": "Montague folding e-bike official product",
            },
            {
                "intent": "expand",
                "target_market_category": "foldable headphones",
                "query": "foldable headphones official product fold flat",
            },
            {
                "intent": "expand",
                "target_market_category": "foldable headphones",
                "query": "swivel foldable headphones official product",
            },
        ]
    )

    return pd.DataFrame(rows).drop_duplicates().reset_index(drop=True)


# =============================================================================
# REPOSITORY EXPORTS
# =============================================================================


def build_readme(df: pd.DataFrame, coverage_df: pd.DataFrame) -> str:
    """Create the repository README content."""
    product_count = len(df)
    category_count = df["market_category"].nunique()
    family_count = df["transformation_family"].nunique()
    gap_count = int((coverage_df["covered"] == False).sum())

    return f"""# Foldable Product Atlas

This repository tracks commercial and historically significant products whose
value depends on a reversible compacting transformation. The catalog includes
foldable, collapsible, rollable, inflatable, and modular products.

The starter catalog currently contains {product_count} product records across
{category_count} market categories and {family_count} transformation families.
The coverage tracker currently identifies {gap_count} target categories that
still need additional review.

## Working definition

A foldable or collapsible product changes physical configuration through a
designed hinge, crease, flexure, telescoping motion, soft-vessel collapse,
rolling motion, inflation state, or flat-pack transformation. The transformation
should improve portability, storage density, deployed area, or use-mode
multiplexing.

## Data files

- `data/products_seed.csv` stores the curated starter records.
- `data/product_sources.csv` stores source URLs and source domains.
- `data/review_queue.csv` stores rows that need source or measurement review.
- `data/coverage_gaps.csv` compares current records against target categories.
- `data/query_seeds.csv` provides search prompts for a discovery notebook.
- `schema/products.schema.json` stores the validation schema.
- `docs/taxonomy.md` documents the transformation families.
- `docs/codebook.md` documents major fields and scoring semantics.

## Scoring model

The foldability score is a weighted triage score using portability gain,
mechanism novelty, market availability, evidence quality, measurement
completeness, and certification relevance. It is a review aid, while the
canonical catalog should remain evidence-driven.
"""


def build_taxonomy_markdown() -> str:
    """Create a mechanism-first taxonomy document."""
    return """# Foldable Product Taxonomy

## Foldable

Foldable products compact through designed creases, hinges, flexures, flexible
display laminates, headphone hinge systems, or origami-like panel networks. The
deployed structure typically preserves a clear geometric relationship to the
stored structure.

Representative subtypes include flexible-display folds, rigid-hinge folds,
origami-panel folds, snap-flexure folds, and headphone-hinge folds.

## Collapsible

Collapsible products compact through nesting, telescoping, shell collapse,
soft-vessel volume evacuation, or frame-assisted collapse. The compacting path
often reduces volume more directly than it preserves shape.

Representative subtypes include segmented-shell collapse, soft-vessel collapse,
telescoping collapse, nesting collapse, and frame-assisted collapse.

## Rollable

Rollable products compact by winding a flexible sheet, membrane, display, panel,
or mat into a smaller stored radius.

## Inflatable

Inflatable products compact through deflation. Their deployed geometry depends
on gas pressure, membrane mechanics, and valve integrity.

## Modular

Modular products compact by disassembly, flat-pack conversion, or repeated
snap-fit assembly. This family is useful when a product gains portability
through reversible assembly rather than a single hinge or fold.
"""


def build_codebook_markdown() -> str:
    """Create a field codebook for catalog contributors."""
    return """# Catalog Codebook

## Core identity fields

`product_id` is a stable lowercase identifier derived from brand and product
name. `brand`, `product_name`, and `market_category` describe the product-level
identity.

## Mechanism fields

`transformation_family` and `transformation_subtype` encode the high-level
taxonomy. `fold_mechanism`, `fold_architecture`, and `compacting_motion` record
the physical mechanism in progressively more descriptive language.

## Evidence fields

`source_type`, `source_url`, `normalized_source_url`, and `source_domain`
support source tracking. `review_status` separates canonical seed records from
records needing source or measurement review.

## Measurement fields

Folded and deployed dimensions use millimeters. Mass uses grams. Price uses USD
when available. Compression ratio should compare deployed volume to folded
volume when both volumes are measured consistently.

## Score fields

All sub-scores use a 0 to 5 scale. `foldability_score` is a weighted review
score. It should guide triage while direct evidence should decide catalog
promotion.
"""


def build_contributing_markdown() -> str:
    """Create contribution rules for catalog entries."""
    return """# Contributing

Add one product per row and include at least one source URL.

## Required fields

- product_id
- brand
- product_name
- market_category
- transformation_family
- transformation_subtype
- fold_mechanism
- fold_architecture
- source_url
- source_type
- review_status

## Evidence expectations

Use higher evidence scores for official product pages, standards documents,
peer-reviewed papers, teardown reports, and direct measurements. Use lower
scores for marketplace listings, social posts, and unsupported claims.

## Suggested measurements

Add folded dimensions, deployed dimensions, mass, materials, hinge cycle claims,
certifications, release year, price, region, availability, and transformation
mechanism notes when available.
"""


def write_repo_files(
    df: pd.DataFrame,
    coverage_df: pd.DataFrame,
    review_df: pd.DataFrame,
    query_df: pd.DataFrame,
    output_root: Path,
) -> None:
    """Write starter repository files without saving plot images."""
    data_dir = output_root / "data"
    docs_dir = output_root / "docs"
    schema_dir = output_root / "schema"
    notebooks_dir = output_root / "notebooks"

    for directory in (data_dir, docs_dir, schema_dir, notebooks_dir):
        directory.mkdir(parents=True, exist_ok=True)

    df.to_csv(data_dir / "products_seed.csv", index=False)
    coverage_df.to_csv(data_dir / "coverage_gaps.csv", index=False)
    review_df.to_csv(data_dir / "review_queue.csv", index=False)
    query_df.to_csv(data_dir / "query_seeds.csv", index=False)

    source_columns = [
        "product_id",
        "brand",
        "product_name",
        "source_type",
        "source_domain",
        "source_url",
    ]
    df[source_columns].to_csv(data_dir / "product_sources.csv", index=False)

    (schema_dir / "products.schema.json").write_text(
        json.dumps(build_schema(), indent=2),
        encoding="utf-8",
    )
    (output_root / "README.md").write_text(
        build_readme(df, coverage_df),
        encoding="utf-8",
    )
    (output_root / "CONTRIBUTING.md").write_text(
        build_contributing_markdown(),
        encoding="utf-8",
    )
    (docs_dir / "taxonomy.md").write_text(
        build_taxonomy_markdown(),
        encoding="utf-8",
    )
    (docs_dir / "codebook.md").write_text(
        build_codebook_markdown(),
        encoding="utf-8",
    )
    (output_root / ".gitignore").write_text(
        ".ipynb_checkpoints/\n__pycache__/\n*.pyc\n.DS_Store\n",
        encoding="utf-8",
    )


def print_file_tree(root: Path) -> None:
    """Print a compact repository file tree."""
    paths = sorted(path for path in root.rglob("*") if path.is_file())
    print(f"\nFiles created under {root}:\n")
    for path in paths:
        print(f"- {path.relative_to(root)}")


# =============================================================================
# DISPLAY AND PLOTS
# =============================================================================


def display_catalog_tables(
    df: pd.DataFrame,
    coverage_df: pd.DataFrame,
    review_df: pd.DataFrame,
) -> None:
    """Display notebook tables for review."""
    display(Markdown("### Catalog overview"))
    display(
        df[
            [
                "brand",
                "product_name",
                "market_category",
                "transformation_family",
                "transformation_subtype",
                "review_status",
                "foldability_score",
                "measurement_completeness_score",
                "source_type",
            ]
        ].sort_values("foldability_score", ascending=False)
    )

    display(Markdown("### Montague folding-bike entries"))
    display(
        df[df["brand"].str.lower() == "montague"][
            [
                "brand",
                "product_name",
                "market_category",
                "transformation_subtype",
                "foldability_score",
                "review_status",
                "source_url",
            ]
        ]
    )

    display(Markdown("### Foldable headphone entries"))
    display(
        df[df["market_category"] == "foldable headphones"][
            [
                "brand",
                "product_name",
                "transformation_subtype",
                "foldability_score",
                "review_status",
                "source_url",
            ]
        ].sort_values(["foldability_score", "brand"], ascending=[False, True])
    )

    display(Markdown("### Coverage gaps"))
    display(coverage_df.sort_values(["covered", "target_market_category"]))

    display(Markdown("### Review queue"))
    display(review_df.head(80))


def plot_category_counts(df: pd.DataFrame) -> None:
    """Render product counts by market category."""
    counts = df["market_category"].value_counts().sort_values()

    fig, ax = plt.subplots(figsize=(10.5, 7.0))
    counts.plot(kind="barh", ax=ax)
    ax.set_title("Seed Catalog Coverage by Market Category")
    ax.set_xlabel("Product count")
    ax.set_ylabel("Market category")
    ax.grid(axis="x", alpha=0.25)
    fig.tight_layout()
    plt.show()


def plot_family_counts(df: pd.DataFrame) -> None:
    """Render product counts by transformation family."""
    counts = df["transformation_family"].value_counts().sort_values()

    fig, ax = plt.subplots(figsize=(8.5, 4.8))
    counts.plot(kind="barh", ax=ax)
    ax.set_title("Seed Catalog Coverage by Transformation Family")
    ax.set_xlabel("Product count")
    ax.set_ylabel("Transformation family")
    ax.grid(axis="x", alpha=0.25)
    fig.tight_layout()
    plt.show()


def plot_score_ranking(df: pd.DataFrame) -> None:
    """Render the foldability score ranking."""
    ranked_df = df.sort_values("foldability_score", ascending=True).tail(30)
    labels = ranked_df["brand"] + " " + ranked_df["product_name"]

    fig, ax = plt.subplots(figsize=(11.5, 9.0))
    ax.barh(labels, ranked_df["foldability_score"])
    ax.set_title("Top Preliminary Foldability Scores")
    ax.set_xlabel("Weighted score from 0 to 5")
    ax.set_ylabel("")
    ax.set_xlim(0.0, 5.2)
    ax.grid(axis="x", alpha=0.25)
    fig.tight_layout()
    plt.show()


def make_compact_plot_label(row: pd.Series, max_chars: int = 30) -> str:
    """Build a compact plot label from brand and product name."""
    label = f"{clean_text(row['brand'])} {clean_text(row['product_name'])}"
    return clean_text(label, max_chars=max_chars)


def compute_deterministic_jitter(
    values: np.ndarray,
    amplitude: float,
    salt: int,
) -> np.ndarray:
    """Compute small deterministic jitter for overlapping scatter points."""
    indices = np.arange(len(values), dtype=float)
    phase = 1.618_033_988_75 * (indices + 1.0 + salt)
    return amplitude * np.sin(2.0 * np.pi * phase)


def select_labeled_points(
    df: pd.DataFrame,
    max_labels: int = 16,
) -> pd.DataFrame:
    """Select a bounded set of points for non-overlapping annotations."""
    label_df = df.copy()

    label_df["label_priority"] = (
        1.30 * label_df["foldability_score"].astype(float)
        + 0.75 * label_df["evidence_quality_score"].astype(float)
        + 0.40 * label_df["market_availability_score"].astype(float)
        + 0.60 * label_df["measurement_completeness_score"].astype(float)
    )

    priority_brands = {
        "Montague",
        "Brompton",
        "Oru Kayak",
        "Foldscope Instruments",
        "Beats",
        "Marshall",
        "Sony",
        "Audio-Technica",
        "Jabra",
    }

    priority_mask = label_df["brand"].isin(priority_brands)
    selected_df = pd.concat(
        [
            label_df.loc[priority_mask],
            label_df.sort_values("label_priority", ascending=False),
        ],
        axis=0,
    )

    selected_df = selected_df.drop_duplicates(subset=["product_id"])
    return selected_df.head(max_labels).copy()


def plot_measurement_vs_score(df: pd.DataFrame) -> None:
    """Render measurement completeness against foldability score cleanly."""
    plot_df = df.copy().reset_index(drop=True)

    x_values = plot_df["measurement_completeness_score"].astype(float).to_numpy()
    y_values = plot_df["foldability_score"].astype(float).to_numpy()
    evidence_values = plot_df["evidence_quality_score"].astype(float).to_numpy()

    x_jitter = compute_deterministic_jitter(
        values=x_values,
        amplitude=0.035,
        salt=3,
    )
    y_jitter = compute_deterministic_jitter(
        values=y_values,
        amplitude=0.030,
        salt=11,
    )

    plot_df["plot_x"] = np.clip(x_values + x_jitter, 0.0, 5.0)
    plot_df["plot_y"] = np.clip(y_values + y_jitter, 0.0, 5.0)

    marker_sizes = 55.0 + 22.0 * evidence_values

    fig, ax = plt.subplots(figsize=(11.5, 7.2))

    ax.scatter(
        plot_df["plot_x"],
        plot_df["plot_y"],
        s=marker_sizes,
        alpha=0.62,
        edgecolors="black",
        linewidths=0.35,
    )

    label_df = select_labeled_points(plot_df, max_labels=16)
    label_df = label_df.sort_values(
        ["plot_x", "plot_y", "foldability_score"],
        ascending=[True, False, False],
    ).reset_index(drop=True)

    lane_offsets = [
        (0.22, 0.42),
        (0.34, 0.28),
        (0.46, 0.14),
        (0.58, 0.00),
        (0.70, -0.14),
        (0.82, -0.28),
        (0.94, -0.42),
        (1.06, 0.35),
        (1.18, 0.21),
        (1.30, 0.07),
        (1.42, -0.07),
        (1.54, -0.21),
        (1.66, -0.35),
        (1.78, 0.49),
        (1.90, -0.49),
        (2.02, 0.00),
    ]

    for index, row in label_df.iterrows():
        dx, dy = lane_offsets[index % len(lane_offsets)]
        label_x = min(float(row["plot_x"]) + dx, 5.05)
        label_y = float(np.clip(float(row["plot_y"]) + dy, 0.18, 5.05))

        ax.annotate(
            make_compact_plot_label(row),
            xy=(row["plot_x"], row["plot_y"]),
            xytext=(label_x, label_y),
            textcoords="data",
            fontsize=8,
            ha="left",
            va="center",
            arrowprops={
                "arrowstyle": "-",
                "linewidth": 0.55,
                "alpha": 0.55,
                "shrinkA": 2.0,
                "shrinkB": 2.0,
            },
            bbox={
    "boxstyle": "round,pad=0.18",
    "facecolor": (1.0, 1.0, 1.0, 0.58),
    "edgecolor": (0.45, 0.45, 0.45, 0.55),
    "linewidth": 0.45,
    },
            clip_on=True,
        )

    ax.set_title("Measurement Completeness versus Foldability Score")
    ax.set_xlabel("Measurement completeness score")
    ax.set_ylabel("Foldability score")
    ax.set_xlim(-0.08, 5.25)
    ax.set_ylim(0.0, 5.25)
    ax.grid(alpha=0.25)

    footnote = (
        "Small deterministic jitter separates overlapping records. "
        "Labels show selected high-priority entries only."
    )
    ax.text(
        0.01,
        0.015,
        footnote,
        transform=ax.transAxes,
        fontsize=8,
        ha="left",
        va="bottom",
        alpha=0.75,
    )

    fig.tight_layout()
    plt.show()


def plot_coverage_gap_status(coverage_df: pd.DataFrame) -> None:
    """Render target category coverage status."""
    table = coverage_df.copy()
    table["coverage_flag"] = table["covered"].map({True: 1, False: 0})
    table = table.sort_values(["coverage_flag", "target_market_category"])

    fig, ax = plt.subplots(figsize=(10.5, 6.8))
    ax.barh(table["target_market_category"], table["coverage_flag"])
    ax.set_title("Target Category Coverage Status")
    ax.set_xlabel("Covered flag")
    ax.set_ylabel("Target market category")
    ax.set_xlim(0.0, 1.1)
    ax.set_xticks([0, 1])
    ax.set_xticklabels(["gap", "covered"])
    ax.grid(axis="x", alpha=0.25)
    fig.tight_layout()
    plt.show()


# =============================================================================
# MAIN WORKFLOW
# =============================================================================


def main() -> pd.DataFrame:
    """Run the catalog build, validation, diagnostics, export, and plots."""
    catalog_df = build_catalog_dataframe()
    coverage_df = build_coverage_table(catalog_df)
    review_df = build_review_queue(catalog_df)
    query_seed_df = build_query_seed_table(coverage_df)

    print("Catalog validation passed.")
    print(f"Seed record count: {len(catalog_df)}")
    print(f"Market category count: {catalog_df['market_category'].nunique()}")
    print(
        "Transformation family count: "
        f"{catalog_df['transformation_family'].nunique()}"
    )
    print(f"Review queue count: {len(review_df)}")

    display_catalog_tables(catalog_df, coverage_df, review_df)

    if RUN_LINK_AUDIT:
        audit_df = audit_links(catalog_df)
        display(Markdown("### Optional link audit"))
        display(audit_df)

    plot_category_counts(catalog_df)
    plot_family_counts(catalog_df)
    plot_score_ranking(catalog_df)
    plot_measurement_vs_score(catalog_df)
    plot_coverage_gap_status(coverage_df)

    if EXPORT_REPO_FILES:
        write_repo_files(
            df=catalog_df,
            coverage_df=coverage_df,
            review_df=review_df,
            query_df=query_seed_df,
            output_root=OUTPUT_ROOT,
        )
        print_file_tree(OUTPUT_ROOT)
        display(Markdown(f"Repository starter files written to `{OUTPUT_ROOT}`."))

    return catalog_df

catalog = main()
```

```python
"""
Discover foldable and collapsible product candidates from web search.

This Colab-ready script expands a foldable-product atlas by searching broad
product-category queries, brand-directed queries, and model-aware queries.
It deduplicates candidate URLs, classifies candidates by transformation family,
scrapes bounded product metadata when allowed, exports review tables, and
renders summary plots.

This version removes the brand_hint column entirely and avoids Colab's
interactive dataframe display reference error by rendering tables as static
HTML output.
"""

from __future__ import annotations

import json
import re
import subprocess
import sys
import time
import urllib.parse
import urllib.robotparser
from dataclasses import dataclass
from pathlib import Path
from typing import Any, Final

# =============================================================================
# CONTROL KNOBS
# =============================================================================

INSTALL_PACKAGES: Final[bool] = True

REPO_NAME: Final[str] = "foldable-product-atlas"
REPO_ROOT: Final[Path] = Path("/content") / REPO_NAME

SEED_CSV_PATH: Final[Path] = REPO_ROOT / "data" / "products_seed.csv"
CANDIDATE_CSV_PATH: Final[Path] = (
    REPO_ROOT / "data" / "discovery_candidates.csv"
)
ENRICHED_CSV_PATH: Final[Path] = (
    REPO_ROOT / "data" / "discovery_candidates_enriched.csv"
)
QUERY_LOG_CSV_PATH: Final[Path] = REPO_ROOT / "data" / "discovery_queries.csv"
DISCOVERY_DOC_PATH: Final[Path] = REPO_ROOT / "docs" / "discovery_protocol.md"

MPL_DPI: Final[int] = 250
DISPLAY_MAX_ROWS: Final[int] = 120
DISPLAY_HTML_MAX_ROWS: Final[int] = 120
DISPLAY_HTML_MAX_COL_WIDTH: Final[int] = 180

SEARCH_REGION: Final[str] = "us-en"
SAFESEARCH: Final[str] = "moderate"
RESULTS_PER_QUERY: Final[int] = 12
MAX_QUERIES_TO_RUN: Final[int | None] = None

SCRAPE_CANDIDATES: Final[bool] = True
RUN_ROBOTS_CHECK: Final[bool] = True
ALLOW_WHEN_ROBOTS_UNAVAILABLE: Final[bool] = True
EXPORT_FILES: Final[bool] = True

MAX_CANDIDATES_TO_SCRAPE: Final[int] = 70
SCRAPE_PRIORITY_MIN: Final[float] = 0.18
SCRAPE_ONLY_NON_SEED_ROWS: Final[bool] = True

SEARCH_DELAY_S: Final[float] = 0.35
SCRAPE_DELAY_S: Final[float] = 0.15

SEARCH_TIMEOUT_S: Final[float] = 10.0
REQUEST_CONNECT_TIMEOUT_S: Final[float] = 4.0
REQUEST_READ_TIMEOUT_S: Final[float] = 6.0
ROBOTS_CONNECT_TIMEOUT_S: Final[float] = 3.0
ROBOTS_READ_TIMEOUT_S: Final[float] = 4.0

RETRY_TOTAL: Final[int] = 1
RETRY_BACKOFF_FACTOR: Final[float] = 0.35
BLOCK_DOMAIN_AFTER_FAILURES: Final[int] = 3

MAX_HTML_BYTES: Final[int] = 1_500_000
MAX_DESCRIPTION_CHARS: Final[int] = 420

USER_AGENT: Final[str] = (
    "foldable-product-atlas-discovery-bot/0.6 "
    "(candidate discovery for public product taxonomy research)"
)

MARKETPLACE_DOMAIN_PENALTY: Final[float] = 0.25
LISTICLE_DOMAIN_PENALTY: Final[float] = 0.15
OFFICIAL_LOOKING_DOMAIN_BONUS: Final[float] = 0.18
OFFICIAL_DOMAIN_BONUS: Final[float] = 0.12
KNOWN_BRAND_TEXT_BONUS: Final[float] = 0.06
EXACT_CATEGORY_QUERY_BONUS: Final[float] = 0.08

# =============================================================================
# DEPENDENCY INSTALLATION
# =============================================================================


def install_dependencies() -> None:
    """Install discovery, scraping, table, and plotting dependencies."""
    subprocess.run(
        [sys.executable, "-m", "pip", "install", "-q", "uv"],
        check=True,
    )
    subprocess.run(
        [
            "uv",
            "pip",
            "install",
            "-q",
            "--system",
            "beautifulsoup4",
            "ddgs",
            "lxml",
            "matplotlib",
            "numpy",
            "pandas",
            "requests",
            "tabulate",
        ],
        check=True,
    )


if INSTALL_PACKAGES:
    install_dependencies()

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from ddgs import DDGS
from IPython.display import HTML, Markdown, display
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

mpl.rcParams["figure.dpi"] = MPL_DPI
pd.set_option("display.max_rows", DISPLAY_MAX_ROWS)
pd.set_option("display.max_columns", 150)
pd.set_option("display.max_colwidth", 180)
pd.set_option("display.width", 220)

# =============================================================================
# SAFE COLAB DISPLAY HELPERS
# =============================================================================

DISPLAY_CACHE: list[pd.DataFrame] = []


def configure_colab_dataframe_display() -> None:
    """Disable Colab interactive dataframe rendering when available."""
    try:
        from google.colab import data_table

        data_table.disable_dataframe_formatter()
    except Exception:
        pass


def truncate_cell_text(value: Any, max_chars: int) -> str:
    """Truncate one table cell for static HTML display."""
    text = clean_text(value)

    if len(text) <= max_chars:
        return text

    return text[: max_chars - 1].rstrip() + "…"


def prepare_display_dataframe(
    df: pd.DataFrame,
    max_rows: int | None = None,
    max_col_width: int = DISPLAY_HTML_MAX_COL_WIDTH,
) -> pd.DataFrame:
    """Build a stable dataframe copy for output rendering."""
    if max_rows is None:
        table = df.copy()
    else:
        table = df.head(max_rows).copy()

    for column in table.columns:
        if table[column].dtype == "object":
            table[column] = table[column].map(
                lambda value: truncate_cell_text(value, max_col_width)
            )

    DISPLAY_CACHE.append(table)
    return table


def display_table(
    title: str,
    df: pd.DataFrame,
    max_rows: int | None = DISPLAY_HTML_MAX_ROWS,
) -> None:
    """Display a dataframe as static HTML."""
    table = prepare_display_dataframe(df, max_rows=max_rows)

    display(Markdown(f"### {title}"))

    html = table.to_html(index=False, escape=True)
    wrapped_html = (
        '<div style="max-height: 520px; overflow: auto; '
        'border: 1px solid #444; padding: 8px;">'
        f"{html}"
        "</div>"
    )
    display(HTML(wrapped_html))


configure_colab_dataframe_display()

# =============================================================================
# SEARCH TOPIC MODEL
# =============================================================================


@dataclass(frozen=True)
class SearchTopic:
    """One search topic for discovery."""

    family_hint: str
    category_hint: str
    query: str


@dataclass(frozen=True)
class DiscoveryProfile:
    """Brand-directed search expansion profile."""

    brand_name: str
    category_hint: str
    family_hint: str
    official_domain_fragment: str
    query_terms: tuple[str, ...]


FOLDING_BIKE_PROFILES: Final[tuple[DiscoveryProfile, ...]] = (
    DiscoveryProfile(
        brand_name="Montague",
        category_hint="folding bicycle",
        family_hint="foldable",
        official_domain_fragment="montaguebikes.com",
        query_terms=(
            "Montague folding bike official",
            "Montague full size folding bicycle product",
            "Montague Paratrooper folding bike",
            "Montague Navigator folding bike",
            "Montague Crosstown folding bike",
            "site:montaguebikes.com folding bike",
        ),
    ),
    DiscoveryProfile(
        brand_name="Brompton",
        category_hint="folding bicycle",
        family_hint="foldable",
        official_domain_fragment="brompton.com",
        query_terms=(
            "Brompton folding bike official",
            "Brompton foldable city bike product",
            "Brompton electric folding bike official",
            "Brompton C Line folding bike",
            "Brompton P Line folding bike",
            "Brompton T Line folding bike",
            "site:brompton.com folding bike",
        ),
    ),
    DiscoveryProfile(
        brand_name="Tern",
        category_hint="folding bicycle",
        family_hint="foldable",
        official_domain_fragment="ternbicycles.com",
        query_terms=(
            "Tern folding bike official",
            "Tern Vektron folding electric bike",
            "Tern BYB folding bike",
            "Tern Link folding bike",
            "Tern Verge folding bike",
            "site:ternbicycles.com folding bike",
        ),
    ),
    DiscoveryProfile(
        brand_name="DAHON",
        category_hint="folding bicycle",
        family_hint="foldable",
        official_domain_fragment="dahon.com",
        query_terms=(
            "DAHON folding bike official",
            "DAHON folding electric bike official",
            "DAHON Mariner folding bike",
            "DAHON K-Feather folding electric bike",
            "DAHON Vigor folding bike",
            "site:dahon.com folding bike",
            "site:usa.dahon.com folding bike",
        ),
    ),
    DiscoveryProfile(
        brand_name="Bike Friday",
        category_hint="folding bicycle",
        family_hint="foldable",
        official_domain_fragment="bikefriday.com",
        query_terms=(
            "Bike Friday folding travel bike official",
            "Bike Friday New World Tourist folding bike",
            "Bike Friday pakiT folding bike",
            "Bike Friday All-Packa folding bike",
            "Bike Friday folding e-bike",
            "site:bikefriday.com folding bike",
        ),
    ),
    DiscoveryProfile(
        brand_name="Birdy",
        category_hint="folding bicycle",
        family_hint="foldable",
        official_domain_fragment="birdybicycle.com",
        query_terms=(
            "Birdy folding bike official",
            "Birdy bicycle folding bike product",
            "site:birdybicycle.com folding bike",
        ),
    ),
    DiscoveryProfile(
        brand_name="Vello",
        category_hint="folding bicycle",
        family_hint="foldable",
        official_domain_fragment="vello.bike",
        query_terms=(
            "Vello folding bike official",
            "Vello Bike foldable bicycle product",
            "Vello folding e-bike official",
            "site:vello.bike folding bike",
        ),
    ),
    DiscoveryProfile(
        brand_name="Helix",
        category_hint="folding bicycle",
        family_hint="foldable",
        official_domain_fragment="helix.ca",
        query_terms=(
            "Helix folding bike official",
            "Helix titanium folding bicycle product",
            "site:helix.ca folding bike",
        ),
    ),
    DiscoveryProfile(
        brand_name="Gocycle",
        category_hint="folding e-bike",
        family_hint="foldable",
        official_domain_fragment="gocycle.com",
        query_terms=(
            "Gocycle folding electric bike official",
            "Gocycle foldable e-bike product",
            "site:gocycle.com folding electric bike",
        ),
    ),
    DiscoveryProfile(
        brand_name="ZiZZO",
        category_hint="folding bicycle",
        family_hint="foldable",
        official_domain_fragment="zizzo.bike",
        query_terms=(
            "ZiZZO folding bike official",
            "ZiZZO foldable bicycle product",
            "site:zizzo.bike folding bike",
        ),
    ),
    DiscoveryProfile(
        brand_name="Citizen Bike",
        category_hint="folding bicycle",
        family_hint="foldable",
        official_domain_fragment="citizenbike.com",
        query_terms=(
            "Citizen Bike folding bike official",
            "Citizen foldable bicycle product",
            "site:citizenbike.com folding bike",
        ),
    ),
    DiscoveryProfile(
        brand_name="STRiDA",
        category_hint="folding bicycle",
        family_hint="foldable",
        official_domain_fragment="strida.com",
        query_terms=(
            "STRiDA folding bike official",
            "STRiDA triangular folding bicycle product",
            "site:strida.com folding bike",
        ),
    ),
    DiscoveryProfile(
        brand_name="Hummingbird",
        category_hint="folding bicycle",
        family_hint="foldable",
        official_domain_fragment="hummingbirdbike.com",
        query_terms=(
            "Hummingbird folding bike official",
            "Hummingbird lightweight folding bicycle product",
            "site:hummingbirdbike.com folding bike",
        ),
    ),
    DiscoveryProfile(
        brand_name="Decathlon Tilt",
        category_hint="folding bicycle",
        family_hint="foldable",
        official_domain_fragment="decathlon.",
        query_terms=(
            "Decathlon Tilt folding bike official",
            "Btwin Tilt folding bicycle product",
            "site:decathlon.com folding bike Tilt",
        ),
    ),
)

FOLDABLE_HEADPHONE_PROFILES: Final[tuple[DiscoveryProfile, ...]] = (
    DiscoveryProfile(
        brand_name="Sony",
        category_hint="foldable headphones",
        family_hint="foldable",
        official_domain_fragment="sony.com",
        query_terms=(
            "Sony foldable headphones official",
            "Sony WH-1000XM4 swivel foldable structure",
            "Sony WH-1000XM4 headphones folded inwards",
            "site:sony.com WH-1000XM4 swivel foldable headphones",
        ),
    ),
    DiscoveryProfile(
        brand_name="Bose",
        category_hint="foldable headphones",
        family_hint="foldable",
        official_domain_fragment="bose.com",
        query_terms=(
            "Bose foldable headphones official",
            "Bose QuietComfort foldable headphones",
            "Bose QuietComfort Ultra foldable design",
            "site:bose.com foldable headphones QuietComfort",
        ),
    ),
    DiscoveryProfile(
        brand_name="soundcore",
        category_hint="foldable headphones",
        family_hint="foldable",
        official_domain_fragment="soundcore.com",
        query_terms=(
            "soundcore foldable headphones official",
            "soundcore Space Q45 foldable hinges",
            "soundcore Space One Pro super foldable headphones",
            "site:soundcore.com foldable headphones",
        ),
    ),
    DiscoveryProfile(
        brand_name="Sennheiser",
        category_hint="foldable headphones",
        family_hint="foldable",
        official_domain_fragment="sennheiser-hearing.com",
        query_terms=(
            "Sennheiser foldable headphones official",
            "Sennheiser HD 400S lightweight foldable headphones",
            "Sennheiser HD 100 folded headphones",
            "site:sennheiser-hearing.com foldable headphones",
        ),
    ),
    DiscoveryProfile(
        brand_name="JBL",
        category_hint="foldable headphones",
        family_hint="foldable",
        official_domain_fragment="jbl.com",
        query_terms=(
            "JBL foldable headphones official",
            "JBL folding headphones product",
            "site:jbl.com foldable headphones",
        ),
    ),
    DiscoveryProfile(
        brand_name="Audio-Technica",
        category_hint="foldable headphones",
        family_hint="foldable",
        official_domain_fragment="audio-technica.com",
        query_terms=(
            "Audio-Technica foldable headphones official",
            "Audio-Technica fold-flat headphones product",
            "site:audio-technica.com foldable headphones",
        ),
    ),
    DiscoveryProfile(
        brand_name="Beats",
        category_hint="foldable headphones",
        family_hint="foldable",
        official_domain_fragment="beatsbydre.com",
        query_terms=(
            "Beats foldable headphones official",
            "Beats Solo foldable headphones",
            "site:beatsbydre.com foldable headphones",
        ),
    ),
    DiscoveryProfile(
        brand_name="Skullcandy",
        category_hint="foldable headphones",
        family_hint="foldable",
        official_domain_fragment="skullcandy.com",
        query_terms=(
            "Skullcandy foldable headphones official",
            "Skullcandy folding headphones product",
            "site:skullcandy.com foldable headphones",
        ),
    ),
    DiscoveryProfile(
        brand_name="Marshall",
        category_hint="foldable headphones",
        family_hint="foldable",
        official_domain_fragment="marshallheadphones.com",
        query_terms=(
            "Marshall foldable headphones official",
            "Marshall headphones collapsible design product",
            "site:marshallheadphones.com foldable headphones",
        ),
    ),
)

BASE_SEARCH_TOPICS: Final[tuple[SearchTopic, ...]] = (
    SearchTopic("foldable", "computer mouse", "foldable computer mouse product"),
    SearchTopic("foldable", "bluetooth keyboard", "foldable bluetooth keyboard product"),
    SearchTopic("foldable", "travel keyboard", "foldable travel keyboard product"),
    SearchTopic("foldable", "headphones", "foldable headphones product official"),
    SearchTopic("foldable", "headphones", "folding headphones product official"),
    SearchTopic("foldable", "headphones", "fold-flat headphones product official"),
    SearchTopic("foldable", "headphones", "travel headphones foldable design product"),
    SearchTopic("foldable", "bicycle", "foldable bicycle product official"),
    SearchTopic("foldable", "bicycle", "folding bicycle product official"),
    SearchTopic("foldable", "e-bike", "folding e-bike product official"),
    SearchTopic("foldable", "bicycle helmet", "foldable bicycle helmet product official"),
    SearchTopic("foldable", "motorcycle helmet", "foldable motorcycle helmet product"),
    SearchTopic("foldable", "kayak", "foldable kayak product official"),
    SearchTopic("foldable", "canoe", "folding canoe product official"),
    SearchTopic("foldable", "paddle board", "foldable paddle board product"),
    SearchTopic("foldable", "solar panel", "foldable solar panel product official"),
    SearchTopic("foldable", "display", "foldable display product official"),
    SearchTopic("foldable", "OLED laptop", "foldable OLED laptop product"),
    SearchTopic("foldable", "smartphone", "foldable smartphone product official"),
    SearchTopic("foldable", "monitor", "foldable monitor product official"),
    SearchTopic("foldable", "drone", "folding drone product official"),
    SearchTopic("foldable", "stroller", "foldable stroller product official"),
    SearchTopic("foldable", "wagon", "folding wagon product official"),
    SearchTopic("foldable", "camping chair", "folding camping chair product official"),
    SearchTopic("foldable", "camping table", "folding camping table product official"),
    SearchTopic("foldable", "treadmill", "folding treadmill product official"),
    SearchTopic("foldable", "desk", "foldable desk product official"),
    SearchTopic("foldable", "bed", "foldable bed product official"),
    SearchTopic("foldable", "wheelchair", "foldable wheelchair product official"),
    SearchTopic("foldable", "crutch", "foldable crutch product official"),
    SearchTopic("foldable", "paper microscope", "foldable paper microscope product"),
    SearchTopic("collapsible", "water bottle", "collapsible water bottle product official"),
    SearchTopic("collapsible", "soft flask", "collapsible soft flask product"),
    SearchTopic("collapsible", "cup", "collapsible cup product official"),
    SearchTopic("collapsible", "coffee cup", "collapsible coffee cup product official"),
    SearchTopic("collapsible", "bowl", "collapsible bowl product official"),
    SearchTopic("collapsible", "food container", "collapsible food container product"),
    SearchTopic("collapsible", "lunch box", "collapsible lunch box product"),
    SearchTopic("collapsible", "cooler", "collapsible cooler product official"),
    SearchTopic("collapsible", "bucket", "collapsible bucket product official"),
    SearchTopic("collapsible", "laundry basket", "collapsible laundry basket product"),
    SearchTopic("collapsible", "storage bin", "collapsible storage bin product"),
    SearchTopic("collapsible", "crate", "collapsible crate product official"),
    SearchTopic("collapsible", "traffic cone", "collapsible traffic cone product"),
    SearchTopic("collapsible", "lantern", "collapsible lantern product official"),
    SearchTopic("collapsible", "trekking pole", "collapsible trekking pole product"),
    SearchTopic("collapsible", "fishing rod", "collapsible fishing rod product"),
    SearchTopic("collapsible", "pet bowl", "collapsible pet bowl product official"),
    SearchTopic("collapsible", "measuring cups", "collapsible measuring cups product"),
    SearchTopic("collapsible", "kettle", "collapsible kettle product"),
    SearchTopic("collapsible", "camping sink", "collapsible sink product camping"),
    SearchTopic("collapsible", "shovel", "collapsible shovel product"),
    SearchTopic("collapsible", "cart", "collapsible cart product official"),
    SearchTopic("rollable", "display", "rollable display product official"),
    SearchTopic("rollable", "keyboard", "rollable keyboard product"),
    SearchTopic("rollable", "piano keyboard", "roll up piano keyboard product"),
    SearchTopic("rollable", "solar charger", "roll up solar charger product"),
    SearchTopic("inflatable", "kayak", "inflatable kayak product official"),
    SearchTopic("inflatable", "paddle board", "inflatable paddle board product official"),
    SearchTopic("inflatable", "air mattress", "inflatable air mattress product official"),
    SearchTopic("modular", "furniture", "flat pack furniture product official"),
    SearchTopic("modular", "furniture", "modular collapsible furniture product"),
)

ALL_DISCOVERY_PROFILES: Final[tuple[DiscoveryProfile, ...]] = (
    FOLDING_BIKE_PROFILES + FOLDABLE_HEADPHONE_PROFILES
)

KNOWN_OFFICIAL_DOMAIN_FRAGMENTS: Final[tuple[str, ...]] = tuple(
    sorted(
        {
            profile.official_domain_fragment
            for profile in ALL_DISCOVERY_PROFILES
            if profile.official_domain_fragment
        }
    )
)

KNOWN_BRAND_TERMS: Final[tuple[str, ...]] = tuple(
    sorted({profile.brand_name.lower() for profile in ALL_DISCOVERY_PROFILES})
)

DENYLIST_DOMAINS: Final[tuple[str, ...]] = (
    "facebook.com",
    "instagram.com",
    "pinterest.com",
    "tiktok.com",
    "youtube.com",
    "reddit.com",
    "x.com",
    "twitter.com",
    "linkedin.com",
    "wikipedia.org",
    "quora.com",
)

MARKETPLACE_DOMAINS: Final[tuple[str, ...]] = (
    "amazon.",
    "aliexpress.",
    "alibaba.",
    "ebay.",
    "etsy.",
    "temu.",
    "walmart.",
    "target.",
    "bestbuy.",
)

LISTICLE_DOMAINS: Final[tuple[str, ...]] = (
    "bestproducts.",
    "gearpatrol.",
    "thewirecutter.",
    "nytimes.",
    "forbes.",
    "cnet.",
    "tomsguide.",
    "techradar.",
    "rtings.",
    "bicycling.",
)

PRODUCT_KEYWORDS: Final[tuple[str, ...]] = (
    "product",
    "shop",
    "buy",
    "store",
    "official",
    "foldable",
    "folding",
    "fold-flat",
    "fold flat",
    "folded",
    "collapsible",
    "rollable",
    "inflatable",
    "portable",
    "compact",
    "headphones",
    "headphone",
    "bicycle",
    "bike",
    "e-bike",
)

# =============================================================================
# TRANSFORMATION TAXONOMY
# =============================================================================


@dataclass(frozen=True)
class TransformationRule:
    """Rule for classifying compacting product transformations."""

    family: str
    subtype: str
    keywords: tuple[str, ...]
    priority: int


TRANSFORMATION_RULES: Final[tuple[TransformationRule, ...]] = (
    TransformationRule(
        family="foldable",
        subtype="flexible-display-fold",
        keywords=(
            "foldable display",
            "folding display",
            "flexible display",
            "foldable oled",
            "folding oled",
            "foldable phone",
            "foldable smartphone",
            "foldable laptop",
            "foldable computer",
            "foldable monitor",
            "rollable display",
            "ultra-thin glass",
        ),
        priority=120,
    ),
    TransformationRule(
        family="foldable",
        subtype="headphone-hinge-fold",
        keywords=(
            "foldable headphones",
            "folding headphones",
            "foldable headphone",
            "folding headphone",
            "fold-flat headphones",
            "fold flat headphones",
            "folded headphones",
            "headphones folded",
            "folded flat ear cup",
            "folded flat earcup",
            "swivel foldable",
            "swivel foldable structure",
            "ear cups swivel inwards",
            "earcups swivel inwards",
            "foldable hinges",
            "folding design",
            "collapsible headphones",
            "travel headphones",
            "compact case",
        ),
        priority=118,
    ),
    TransformationRule(
        family="foldable",
        subtype="bicycle-frame-fold",
        keywords=(
            "foldable bicycle",
            "folding bicycle",
            "foldable bike",
            "folding bike",
            "folding e-bike",
            "foldable e-bike",
            "folding electric bike",
            "foldable electric bike",
            "folds in",
            "folded size",
            "folded dimensions",
            "folding frame",
            "frame hinge",
            "hinged frame",
            "compact folding bike",
            "full size folding bike",
            "travel bike",
            "multimodal transportation",
            "train car trunk storage",
            "montague",
            "brompton",
            "tern",
            "dahon",
            "bike friday",
            "birdy",
            "vello",
            "helix",
            "gocycle",
            "zizzo",
            "citizen bike",
            "strida",
            "hummingbird",
        ),
        priority=116,
    ),
    TransformationRule(
        family="foldable",
        subtype="general-hinge-fold",
        keywords=(
            "folding stroller",
            "foldable stroller",
            "folding wheelchair",
            "foldable wheelchair",
            "folding chair",
            "folding table",
            "folding treadmill",
            "folding drone",
            "hinge",
            "hinged",
            "book-fold",
            "clamshell",
        ),
        priority=105,
    ),
    TransformationRule(
        family="foldable",
        subtype="origami-fold",
        keywords=(
            "origami",
            "folding kayak",
            "foldable kayak",
            "folding canoe",
            "foldable canoe",
            "paper microscope",
            "foldscope",
            "folded paper",
            "folded panel",
            "crease pattern",
        ),
        priority=100,
    ),
    TransformationRule(
        family="foldable",
        subtype="flexure-fold",
        keywords=(
            "foldable mouse",
            "arc mouse",
            "snap-flat",
            "snap flat",
            "flexure",
            "flexible shell",
            "bendable shell",
        ),
        priority=95,
    ),
    TransformationRule(
        family="collapsible",
        subtype="squeeze-collapse",
        keywords=(
            "collapsible water bottle",
            "collapsible bottle",
            "soft flask",
            "hydration flask",
            "soft bottle",
            "water pouch",
            "squeeze bottle",
            "reservoir",
        ),
        priority=115,
    ),
    TransformationRule(
        family="collapsible",
        subtype="telescoping-collapse",
        keywords=(
            "collapsible cup",
            "collapsible coffee cup",
            "collapsible bowl",
            "collapsible food container",
            "collapsible measuring cup",
            "collapsible kettle",
            "telescoping",
            "telescopic",
            "nested",
            "nesting",
        ),
        priority=110,
    ),
    TransformationRule(
        family="collapsible",
        subtype="segmented-shell-collapse",
        keywords=(
            "foldable helmet",
            "folding helmet",
            "collapsible helmet",
            "segmented helmet",
            "collapsible shell",
            "nested shell",
        ),
        priority=108,
    ),
    TransformationRule(
        family="collapsible",
        subtype="frame-assisted-collapse",
        keywords=(
            "collapsible chair",
            "collapsible table",
            "collapsible wagon",
            "collapsible cart",
            "collapsible laundry basket",
            "collapsible storage bin",
            "collapsible crate",
            "collapsible cooler",
            "collapsible traffic cone",
            "pop-up",
            "pop up",
        ),
        priority=95,
    ),
    TransformationRule(
        family="rollable",
        subtype="rollable-sheet",
        keywords=(
            "rollable",
            "roll-up",
            "roll up",
            "rolltop",
            "roll-top",
            "rollable keyboard",
            "roll up keyboard",
            "roll up solar",
        ),
        priority=85,
    ),
    TransformationRule(
        family="inflatable",
        subtype="deflation-compaction",
        keywords=(
            "inflatable",
            "deflatable",
            "air mattress",
            "inflatable kayak",
            "inflatable paddle board",
            "air beam",
        ),
        priority=80,
    ),
    TransformationRule(
        family="modular",
        subtype="flat-pack-or-disassemblable",
        keywords=(
            "flat pack",
            "flat-pack",
            "modular",
            "disassemblable",
            "breakdown",
            "snap together",
            "snap-together",
        ),
        priority=60,
    ),
)

FAMILY_ORDER: Final[list[str]] = [
    "foldable",
    "collapsible",
    "rollable",
    "inflatable",
    "modular",
    "hybrid",
    "unclassified",
]

# =============================================================================
# TEXT AND URL UTILITIES
# =============================================================================


def clean_text(value: Any, max_chars: int | None = None) -> str:
    """Normalize text to one line and optionally truncate it."""
    if value is None:
        return ""

    if isinstance(value, float) and pd.isna(value):
        return ""

    try:
        is_missing = pd.isna(value)
        if isinstance(is_missing, bool) and is_missing:
            return ""
    except (TypeError, ValueError):
        pass

    text = re.sub(r"\s+", " ", str(value)).strip()

    if max_chars is not None and len(text) > max_chars:
        return text[: max_chars - 1].rstrip() + "…"

    return text


def normalize_url(url: str) -> str:
    """Normalize a URL for deduplication."""
    parsed = urllib.parse.urlparse(clean_text(url))
    scheme = parsed.scheme.lower() or "https"
    netloc = parsed.netloc.lower().removeprefix("www.")
    path = re.sub(r"/+$", "", parsed.path)

    return urllib.parse.urlunparse((scheme, netloc, path, "", "", ""))


def extract_domain(url: str) -> str:
    """Extract the normalized domain from a URL."""
    parsed = urllib.parse.urlparse(clean_text(url))
    return parsed.netloc.lower().removeprefix("www.")


def domain_is_denied(domain: str) -> bool:
    """Return whether a domain should be excluded from discovery."""
    return any(blocked in domain for blocked in DENYLIST_DOMAINS)


def domain_contains(domain: str, fragments: tuple[str, ...]) -> bool:
    """Return whether a domain contains any configured fragment."""
    return any(fragment in domain for fragment in fragments)


def first_nonempty(*values: Any) -> str:
    """Return the first nonempty cleaned value."""
    for value in values:
        text = clean_text(value)
        if text:
            return text
    return ""


def safe_bool(value: Any) -> bool:
    """Convert missing or scalar values into a stable boolean."""
    if value is None:
        return False

    try:
        if pd.isna(value):
            return False
    except (TypeError, ValueError):
        pass

    if isinstance(value, str):
        return value.strip().lower() in {"true", "1", "yes", "y"}

    return bool(value)


def topic_to_dict(topic: SearchTopic) -> dict[str, str]:
    """Convert one SearchTopic into a table row."""
    return {
        "family_hint": topic.family_hint,
        "category_hint": topic.category_hint,
        "query": topic.query,
    }


def build_profile_topics(
    profiles: tuple[DiscoveryProfile, ...],
) -> list[SearchTopic]:
    """Build search topics from discovery profiles."""
    topics: list[SearchTopic] = []

    for profile in profiles:
        for query in profile.query_terms:
            topics.append(
                SearchTopic(
                    family_hint=profile.family_hint,
                    category_hint=profile.category_hint,
                    query=query,
                )
            )

    return topics


def build_all_search_topics() -> list[dict[str, str]]:
    """Build generic, brand-directed, and model-aware query topics."""
    topics = list(BASE_SEARCH_TOPICS)
    topics.extend(build_profile_topics(FOLDING_BIKE_PROFILES))
    topics.extend(build_profile_topics(FOLDABLE_HEADPHONE_PROFILES))

    rows = [topic_to_dict(topic) for topic in topics]
    seen: set[tuple[str, str, str]] = set()
    unique_rows: list[dict[str, str]] = []

    for row in rows:
        key = (
            row["family_hint"].lower(),
            row["category_hint"].lower(),
            row["query"].lower(),
        )
        if key in seen:
            continue
        seen.add(key)
        unique_rows.append(row)

    return unique_rows


# =============================================================================
# SEARCH DISCOVERY
# =============================================================================


def build_query_table() -> pd.DataFrame:
    """Build the search query table from configured topics."""
    query_df = pd.DataFrame(build_all_search_topics()).drop_duplicates()

    if MAX_QUERIES_TO_RUN is not None:
        query_df = query_df.head(MAX_QUERIES_TO_RUN)

    query_df["query_id"] = [
        f"q{index:04d}" for index in range(1, len(query_df) + 1)
    ]

    return query_df[["query_id", "family_hint", "category_hint", "query"]]


def run_single_search(query: str, max_results: int) -> list[dict[str, str]]:
    """Run one DDGS web search and return result dictionaries."""
    try:
        with DDGS(timeout=SEARCH_TIMEOUT_S) as ddgs:
            results = ddgs.text(
                query,
                region=SEARCH_REGION,
                safesearch=SAFESEARCH,
                max_results=max_results,
            )
        return list(results)
    except Exception as exc:
        print(f"Search failed for query: {query}")
        print(f"Reason: {exc}")
        return []


def run_discovery_searches(query_df: pd.DataFrame) -> pd.DataFrame:
    """Run configured search queries and collect raw candidates."""
    rows: list[dict[str, Any]] = []
    total_queries = len(query_df)

    for index, query_row in query_df.iterrows():
        query = clean_text(query_row["query"])
        started_at = time.perf_counter()

        print(f"[{index + 1}/{total_queries}] Searching: {query}")

        results = run_single_search(query, RESULTS_PER_QUERY)
        elapsed_s = time.perf_counter() - started_at
        print(f"    Results: {len(results)} | elapsed: {elapsed_s:.2f} s")

        for rank, result in enumerate(results, start=1):
            href = clean_text(result.get("href") or result.get("url"))
            title = clean_text(result.get("title"))
            body = clean_text(result.get("body"))

            if href == "":
                continue

            domain = extract_domain(href)

            if domain_is_denied(domain):
                continue

            rows.append(
                {
                    "query_id": query_row["query_id"],
                    "query": query,
                    "family_hint": query_row["family_hint"],
                    "category_hint": query_row["category_hint"],
                    "search_rank": rank,
                    "title": title,
                    "body": body,
                    "url": href,
                    "normalized_url": normalize_url(href),
                    "domain": domain,
                }
            )

        if index < total_queries - 1:
            time.sleep(SEARCH_DELAY_S)

    columns = [
        "query_id",
        "query",
        "family_hint",
        "category_hint",
        "search_rank",
        "title",
        "body",
        "url",
        "normalized_url",
        "domain",
    ]

    if len(rows) == 0:
        return pd.DataFrame(columns=columns)

    raw_df = pd.DataFrame(rows)
    raw_df = raw_df.drop_duplicates(subset=["normalized_url"])
    raw_df = raw_df.reset_index(drop=True)
    return raw_df[columns]


# =============================================================================
# CLASSIFICATION AND RANKING
# =============================================================================


def find_transformation_matches(
    text: str,
) -> list[tuple[TransformationRule, list[str]]]:
    """Find transformation rules whose keywords appear in text."""
    lowered = text.lower()
    matches: list[tuple[TransformationRule, list[str]]] = []

    for rule in TRANSFORMATION_RULES:
        matched_keywords = [
            keyword for keyword in rule.keywords if keyword.lower() in lowered
        ]

        if matched_keywords:
            matches.append((rule, matched_keywords))

    matches.sort(
        key=lambda item: (item[0].priority, len(item[1])),
        reverse=True,
    )
    return matches


def classify_candidate_text(text: str) -> dict[str, Any]:
    """Classify candidate text into transformation family and subtype."""
    matches = find_transformation_matches(text)

    if len(matches) == 0:
        return {
            "transformation_family": "unclassified",
            "transformation_subtype": "unclassified",
            "classification_confidence": 0.0,
            "matched_keywords": "",
            "secondary_family_candidates": "",
        }

    best_rule, best_keywords = matches[0]
    families = {rule.family for rule, _ in matches}
    family = "hybrid" if len(families) >= 2 else best_rule.family

    confidence = min(1.0, 0.35 + 0.15 * len(best_keywords))
    secondary_families = sorted(families.difference({best_rule.family}))

    all_keywords: list[str] = []
    for _, keywords in matches[:5]:
        all_keywords.extend(keywords)

    unique_keywords: list[str] = []
    for keyword in all_keywords:
        if keyword in unique_keywords:
            continue
        unique_keywords.append(keyword)

    return {
        "transformation_family": family,
        "transformation_subtype": best_rule.subtype,
        "classification_confidence": round(confidence, 3),
        "matched_keywords": ", ".join(unique_keywords[:16]),
        "secondary_family_candidates": ", ".join(secondary_families),
    }


def add_candidate_classification(candidate_df: pd.DataFrame) -> pd.DataFrame:
    """Classify each search candidate by text evidence."""
    output_df = candidate_df.copy()
    classification_rows = []

    for _, row in output_df.iterrows():
        combined_text = " | ".join(
            [
                clean_text(row.get("query")),
                clean_text(row.get("family_hint")),
                clean_text(row.get("category_hint")),
                clean_text(row.get("title")),
                clean_text(row.get("body")),
                clean_text(row.get("url")),
            ]
        )
        classification_rows.append(classify_candidate_text(combined_text))

    classification_df = pd.DataFrame(classification_rows)
    return pd.concat([output_df, classification_df], axis=1)


def row_has_exact_category_signal(row: pd.Series) -> bool:
    """Return whether a row has direct category evidence."""
    category_hint = clean_text(row.get("category_hint")).lower()
    title = clean_text(row.get("title")).lower()
    body = clean_text(row.get("body")).lower()
    query = clean_text(row.get("query")).lower()
    url = clean_text(row.get("url")).lower()
    text = f"{title} {body} {query} {url}"

    if category_hint in text:
        return True

    if category_hint in ("foldable headphones", "headphones"):
        return any(
            phrase in text
            for phrase in (
                "foldable headphones",
                "folding headphones",
                "fold-flat headphones",
                "headphones folded",
            )
        )

    if category_hint in ("folding bicycle", "folding e-bike", "bicycle", "e-bike"):
        return any(
            phrase in text
            for phrase in (
                "folding bike",
                "folding bicycle",
                "foldable bike",
                "foldable bicycle",
                "folding e-bike",
                "folding electric bike",
            )
        )

    return False


def row_has_known_brand_signal(row: pd.Series) -> bool:
    """Return whether a known brand name appears in row text."""
    domain = clean_text(row.get("domain")).lower()
    title = clean_text(row.get("title")).lower()
    body = clean_text(row.get("body")).lower()
    query = clean_text(row.get("query")).lower()
    url = clean_text(row.get("url")).lower()
    text = f"{title} {body} {query} {url} {domain}"
    normalized_text = text.replace("-", "").replace(".", "").replace("_", "")

    for brand_name in KNOWN_BRAND_TERMS:
        normalized_brand = brand_name.replace(" ", "")
        if brand_name in text or normalized_brand in normalized_text:
            return True

    return False


def score_candidate_source(row: pd.Series) -> float:
    """Score a candidate URL for manual review priority."""
    score = 0.0

    title = clean_text(row.get("title")).lower()
    body = clean_text(row.get("body")).lower()
    query = clean_text(row.get("query")).lower()
    domain = clean_text(row.get("domain")).lower()
    url = clean_text(row.get("url")).lower()
    combined = f"{title} {body} {query} {url}"

    keyword_hits = sum(keyword in combined for keyword in PRODUCT_KEYWORDS)
    score += min(0.35, 0.05 * keyword_hits)
    score += 0.30 * float(row.get("classification_confidence", 0.0))

    if domain_contains(domain, MARKETPLACE_DOMAINS):
        score -= MARKETPLACE_DOMAIN_PENALTY

    if domain_contains(domain, LISTICLE_DOMAINS):
        score -= LISTICLE_DOMAIN_PENALTY

    if any(term in url for term in ("/product", "/products", "/shop", "/store")):
        score += OFFICIAL_LOOKING_DOMAIN_BONUS

    if domain_contains(domain, KNOWN_OFFICIAL_DOMAIN_FRAGMENTS):
        score += OFFICIAL_DOMAIN_BONUS

    if row_has_exact_category_signal(row):
        score += EXACT_CATEGORY_QUERY_BONUS

    if row_has_known_brand_signal(row):
        score += KNOWN_BRAND_TEXT_BONUS

    rank = float(row.get("search_rank", RESULTS_PER_QUERY))
    score += max(0.0, 0.15 * (1.0 - (rank - 1.0) / max(RESULTS_PER_QUERY, 1)))

    return round(float(np.clip(score, 0.0, 1.0)), 3)


def rank_candidates(candidate_df: pd.DataFrame) -> pd.DataFrame:
    """Rank candidates for manual review."""
    output_df = candidate_df.copy()
    output_df["source_priority_score"] = output_df.apply(
        score_candidate_source,
        axis=1,
    )

    family_dtype = pd.CategoricalDtype(categories=FAMILY_ORDER, ordered=True)
    output_df["transformation_family"] = output_df[
        "transformation_family"
    ].astype(family_dtype)

    output_df = output_df.sort_values(
        [
            "source_priority_score",
            "classification_confidence",
            "transformation_family",
            "category_hint",
            "search_rank",
        ],
        ascending=[False, False, True, True, True],
        kind="stable",
    ).reset_index(drop=True)

    output_df["candidate_id"] = [
        f"cand-{index:05d}" for index in range(1, len(output_df) + 1)
    ]

    return output_df[
        [
            "candidate_id",
            "transformation_family",
            "transformation_subtype",
            "classification_confidence",
            "source_priority_score",
            "matched_keywords",
            "secondary_family_candidates",
            "family_hint",
            "category_hint",
            "query",
            "search_rank",
            "title",
            "body",
            "domain",
            "url",
            "normalized_url",
        ]
    ]


# =============================================================================
# BOUNDED METADATA SCRAPING
# =============================================================================


@dataclass(frozen=True)
class FetchResult:
    """Result from one HTML fetch operation."""

    url: str
    final_url: str
    status_code: int | None
    html: str
    ok: bool
    error_message: str


def build_http_session() -> requests.Session:
    """Create a retry-enabled HTTP session."""
    retry_strategy = Retry(
        total=RETRY_TOTAL,
        connect=RETRY_TOTAL,
        read=RETRY_TOTAL,
        backoff_factor=RETRY_BACKOFF_FACTOR,
        status_forcelist=(429, 500, 502, 503, 504),
        allowed_methods=("GET", "HEAD"),
    )

    adapter = HTTPAdapter(max_retries=retry_strategy)
    session = requests.Session()
    session.headers.update({"User-Agent": USER_AGENT})
    session.mount("https://", adapter)
    session.mount("http://", adapter)
    return session


def build_robots_url(url: str) -> str:
    """Build the robots.txt URL for a page."""
    parsed = urllib.parse.urlparse(url)
    root = f"{parsed.scheme}://{parsed.netloc}"
    return urllib.parse.urljoin(root, "/robots.txt")


def can_fetch_url(
    session: requests.Session,
    url: str,
    robots_cache: dict[str, tuple[bool, str]],
) -> tuple[bool, str]:
    """Check robots.txt with an explicit timeout and a domain cache."""
    if RUN_ROBOTS_CHECK is False:
        return True, "robots_check_disabled"

    domain = extract_domain(url)
    if domain in robots_cache:
        return robots_cache[domain]

    robots_url = build_robots_url(url)

    try:
        response = session.get(
            robots_url,
            timeout=(ROBOTS_CONNECT_TIMEOUT_S, ROBOTS_READ_TIMEOUT_S),
            allow_redirects=True,
        )

        if response.ok is False:
            if ALLOW_WHEN_ROBOTS_UNAVAILABLE:
                result = (
                    True,
                    f"robots_unavailable_allowed: {response.status_code}",
                )
            else:
                result = (
                    False,
                    f"robots_unavailable_blocked: {response.status_code}",
                )
            robots_cache[domain] = result
            return result

        parser = urllib.robotparser.RobotFileParser()
        parser.parse(response.text.splitlines())

        allowed = parser.can_fetch(USER_AGENT, url)
        status = "allowed" if allowed else "blocked_by_robots_txt"
        result = (allowed, status)
        robots_cache[domain] = result
        return result

    except requests.RequestException as exc:
        if ALLOW_WHEN_ROBOTS_UNAVAILABLE:
            result = (True, f"robots_unavailable_allowed: {exc}")
        else:
            result = (False, f"robots_unavailable_blocked: {exc}")
        robots_cache[domain] = result
        return result


def response_looks_like_html(response: requests.Response) -> bool:
    """Return whether an HTTP response appears to contain HTML."""
    content_type = response.headers.get("content-type", "").lower()
    return "html" in content_type or "text" in content_type


def read_limited_response_text(response: requests.Response) -> str:
    """Read response text while capping the amount of downloaded HTML."""
    chunks: list[bytes] = []
    total_bytes = 0

    for chunk in response.iter_content(chunk_size=32_768):
        if chunk:
            chunks.append(chunk)
            total_bytes += len(chunk)

        if total_bytes >= MAX_HTML_BYTES:
            break

    raw = b"".join(chunks)
    encoding = response.encoding or response.apparent_encoding or "utf-8"
    return raw.decode(encoding, errors="replace")


def fetch_html(
    session: requests.Session,
    url: str,
    robots_cache: dict[str, tuple[bool, str]],
) -> FetchResult:
    """Fetch one HTML page with bounded robots and request timeouts."""
    allowed, robots_status = can_fetch_url(session, url, robots_cache)

    if allowed is False:
        return FetchResult(
            url=url,
            final_url=url,
            status_code=None,
            html="",
            ok=False,
            error_message=robots_status,
        )

    try:
        response = session.get(
            url,
            timeout=(REQUEST_CONNECT_TIMEOUT_S, REQUEST_READ_TIMEOUT_S),
            allow_redirects=True,
            stream=True,
        )

        if response.ok is False:
            return FetchResult(
                url=url,
                final_url=response.url,
                status_code=response.status_code,
                html="",
                ok=False,
                error_message=(
                    f"{robots_status}; http_status: {response.status_code}"
                ),
            )

        if response_looks_like_html(response) is False:
            return FetchResult(
                url=url,
                final_url=response.url,
                status_code=response.status_code,
                html="",
                ok=False,
                error_message=f"{robots_status}; non_html_content",
            )

        html = read_limited_response_text(response)
        html_is_valid = html.lstrip().startswith("<") or "<html" in html[:1000].lower()

        return FetchResult(
            url=url,
            final_url=response.url,
            status_code=response.status_code,
            html=html if html_is_valid else "",
            ok=bool(html_is_valid),
            error_message=robots_status,
        )

    except requests.RequestException as exc:
        return FetchResult(
            url=url,
            final_url=url,
            status_code=None,
            html="",
            ok=False,
            error_message=f"{robots_status}; request_error: {exc}",
        )


def parse_json_safely(raw_json: str) -> Any:
    """Parse JSON while tolerating common control-character issues."""
    text = raw_json.strip()

    if text == "":
        return None

    try:
        return json.loads(text)
    except json.JSONDecodeError:
        text = re.sub(r"[\x00-\x1f\x7f]", " ", text)
        try:
            return json.loads(text)
        except json.JSONDecodeError:
            return None


def as_list(value: Any) -> list[Any]:
    """Convert scalar or list-like metadata into a Python list."""
    if value is None:
        return []

    if isinstance(value, list):
        return value

    return [value]


def flatten_json_ld(value: Any) -> list[dict[str, Any]]:
    """Flatten JSON-LD objects and graphs into record dictionaries."""
    flattened: list[dict[str, Any]] = []

    if isinstance(value, list):
        for item in value:
            flattened.extend(flatten_json_ld(item))
        return flattened

    if isinstance(value, dict):
        flattened.append(value)

        for key in ("@graph", "mainEntity", "itemListElement", "offers"):
            child = value.get(key)
            if isinstance(child, (dict, list)):
                flattened.extend(flatten_json_ld(child))

    return flattened


def extract_json_ld_objects(soup: BeautifulSoup) -> list[dict[str, Any]]:
    """Extract parseable JSON-LD records from HTML."""
    objects: list[dict[str, Any]] = []

    for script in soup.find_all("script", attrs={"type": "application/ld+json"}):
        parsed = parse_json_safely(script.get_text())
        if parsed is None:
            continue
        objects.extend(flatten_json_ld(parsed))

    return objects


def record_has_type(record: dict[str, Any], target_type: str) -> bool:
    """Check JSON-LD @type value for a target type."""
    raw_type = record.get("@type")
    types = [str(item).lower() for item in as_list(raw_type)]
    return target_type.lower() in types


def select_json_ld_product(records: list[dict[str, Any]]) -> dict[str, Any]:
    """Select the first JSON-LD Product record."""
    for record in records:
        if record_has_type(record, "Product"):
            return record
    return {}


def select_json_ld_offer(product: dict[str, Any]) -> dict[str, Any]:
    """Select the first JSON-LD Offer record from a Product."""
    for offer in as_list(product.get("offers")):
        if isinstance(offer, dict):
            return offer
    return {}


def extract_brand_name(value: Any) -> str:
    """Extract a brand name from JSON-LD brand metadata."""
    if isinstance(value, dict):
        return clean_text(value.get("name"))

    if isinstance(value, list):
        for item in value:
            brand_name = extract_brand_name(item)
            if brand_name:
                return brand_name

    return clean_text(value)


def extract_image_url(value: Any) -> str:
    """Extract the first image URL from JSON-LD image metadata."""
    for item in as_list(value):
        if isinstance(item, dict):
            image_url = clean_text(item.get("url") or item.get("contentUrl"))
        else:
            image_url = clean_text(item)

        if image_url:
            return image_url

    return ""


def get_meta_content(soup: BeautifulSoup, *names: str) -> str:
    """Read content from common metadata tags."""
    for name in names:
        for attrs in ({"property": name}, {"name": name}, {"itemprop": name}):
            tag = soup.find("meta", attrs=attrs)
            if tag is not None and tag.get("content"):
                return clean_text(tag.get("content"))

    return ""


def get_heading_title(soup: BeautifulSoup) -> str:
    """Extract a visible page title fallback."""
    h1 = soup.find("h1")
    title = soup.find("title")

    for tag in (h1, title):
        if tag is None:
            continue

        text = clean_text(tag.get_text(" ", strip=True))
        if text:
            return text

    return ""


def extract_page_metadata(fetch_result: FetchResult) -> dict[str, Any]:
    """Extract metadata from one fetched HTML page."""
    if fetch_result.ok is False:
        return {
            "fetch_ok": False,
            "final_url": fetch_result.final_url,
            "status_code": fetch_result.status_code,
            "scraped_name": "",
            "scraped_brand": "",
            "scraped_description": "",
            "scraped_price": "",
            "scraped_currency": "",
            "scraped_availability": "",
            "scraped_image_url": "",
            "json_ld_product_found": False,
            "error_message": fetch_result.error_message,
        }

    soup = BeautifulSoup(fetch_result.html, "lxml")
    json_ld_records = extract_json_ld_objects(soup)
    product = select_json_ld_product(json_ld_records)
    offer = select_json_ld_offer(product)

    availability = clean_text(offer.get("availability"))
    if availability.startswith("http"):
        availability = availability.rstrip("/").split("/")[-1]

    return {
        "fetch_ok": True,
        "final_url": fetch_result.final_url,
        "status_code": fetch_result.status_code,
        "scraped_name": first_nonempty(
            product.get("name"),
            get_meta_content(soup, "og:title", "twitter:title"),
            get_heading_title(soup),
        ),
        "scraped_brand": first_nonempty(
            extract_brand_name(product.get("brand")),
            get_meta_content(soup, "product:brand", "brand"),
        ),
        "scraped_description": clean_text(
            first_nonempty(
                product.get("description"),
                get_meta_content(
                    soup,
                    "og:description",
                    "twitter:description",
                    "description",
                ),
            ),
            MAX_DESCRIPTION_CHARS,
        ),
        "scraped_price": first_nonempty(
            offer.get("price"),
            offer.get("lowPrice"),
            offer.get("highPrice"),
            get_meta_content(soup, "product:price:amount"),
        ),
        "scraped_currency": first_nonempty(
            offer.get("priceCurrency"),
            get_meta_content(soup, "product:price:currency"),
        ),
        "scraped_availability": availability,
        "scraped_image_url": first_nonempty(
            extract_image_url(product.get("image")),
            get_meta_content(soup, "og:image", "twitter:image"),
        ),
        "json_ld_product_found": bool(product),
        "error_message": fetch_result.error_message,
    }


def compute_metadata_score(row: pd.Series) -> float:
    """Compute a simple metadata completeness score."""
    fields = [
        "scraped_name",
        "scraped_brand",
        "scraped_description",
        "scraped_price",
        "scraped_currency",
        "scraped_availability",
        "scraped_image_url",
    ]
    present = sum(clean_text(row.get(field)) != "" for field in fields)
    return round(present / len(fields), 3)


def choose_scrape_rows(candidate_df: pd.DataFrame) -> pd.DataFrame:
    """Choose a bounded set of candidates for enrichment scraping."""
    scrape_df = candidate_df.copy()

    if SCRAPE_ONLY_NON_SEED_ROWS and "already_in_seed_catalog" in scrape_df.columns:
        scrape_df = scrape_df.loc[~scrape_df["already_in_seed_catalog"]].copy()

    scrape_df = scrape_df.loc[
        scrape_df["source_priority_score"].astype(float) >= SCRAPE_PRIORITY_MIN
    ].copy()

    scrape_df = scrape_df.sort_values(
        ["source_priority_score", "classification_confidence"],
        ascending=[False, False],
        kind="stable",
    )

    return scrape_df.head(MAX_CANDIDATES_TO_SCRAPE).reset_index(drop=True)


def build_skip_metadata(row: pd.Series, reason: str) -> dict[str, Any]:
    """Build metadata for a skipped candidate."""
    return {
        "candidate_id": row["candidate_id"],
        "fetch_ok": False,
        "final_url": clean_text(row["url"]),
        "status_code": None,
        "scraped_name": "",
        "scraped_brand": "",
        "scraped_description": "",
        "scraped_price": "",
        "scraped_currency": "",
        "scraped_availability": "",
        "scraped_image_url": "",
        "json_ld_product_found": False,
        "error_message": reason,
    }


def scrape_candidate_pages(candidate_df: pd.DataFrame) -> pd.DataFrame:
    """Scrape selected candidate pages and merge metadata."""
    session = build_http_session()
    robots_cache: dict[str, tuple[bool, str]] = {}
    domain_failures: dict[str, int] = {}
    scrape_df = choose_scrape_rows(candidate_df)

    metadata_rows: list[dict[str, Any]] = []
    total_to_scrape = len(scrape_df)

    print(
        "Selected scrape candidates: "
        f"{total_to_scrape} of {len(candidate_df)} total candidates"
    )

    for local_index, row in scrape_df.iterrows():
        url = clean_text(row["url"])
        domain = clean_text(row["domain"])
        started_at = time.perf_counter()

        if domain_failures.get(domain, 0) >= BLOCK_DOMAIN_AFTER_FAILURES:
            print(
                f"[{local_index + 1}/{total_to_scrape}] Skipping domain "
                f"after repeated failures: {domain}"
            )
            metadata_rows.append(
                build_skip_metadata(row, "skipped_after_domain_failures")
            )
            continue

        print(f"[{local_index + 1}/{total_to_scrape}] Scraping: {url}")

        fetch_result = fetch_html(session, url, robots_cache)
        metadata = extract_page_metadata(fetch_result)
        metadata["candidate_id"] = row["candidate_id"]
        metadata_rows.append(metadata)

        if fetch_result.ok is False:
            domain_failures[domain] = domain_failures.get(domain, 0) + 1

        elapsed_s = time.perf_counter() - started_at
        print(
            "    "
            f"ok={fetch_result.ok} | "
            f"status={fetch_result.status_code} | "
            f"elapsed={elapsed_s:.2f} s | "
            f"message={clean_text(fetch_result.error_message, 160)}"
        )

        if local_index < total_to_scrape - 1:
            time.sleep(SCRAPE_DELAY_S)

    metadata_df = pd.DataFrame(metadata_rows)
    enriched_df = candidate_df.merge(metadata_df, on="candidate_id", how="left")

    for column in [
        "fetch_ok",
        "json_ld_product_found",
        "status_code",
        "scraped_name",
        "scraped_brand",
        "scraped_description",
        "scraped_price",
        "scraped_currency",
        "scraped_availability",
        "scraped_image_url",
        "error_message",
    ]:
        if column not in enriched_df.columns:
            enriched_df[column] = ""

    enriched_df["fetch_ok"] = enriched_df["fetch_ok"].map(safe_bool).astype(bool)
    enriched_df["json_ld_product_found"] = (
        enriched_df["json_ld_product_found"].map(safe_bool).astype(bool)
    )

    enriched_df["metadata_completeness_score"] = enriched_df.apply(
        compute_metadata_score,
        axis=1,
    )

    combined_text = (
        enriched_df["title"].fillna("")
        + " | "
        + enriched_df["body"].fillna("")
        + " | "
        + enriched_df["category_hint"].fillna("")
        + " | "
        + enriched_df["scraped_name"].fillna("")
        + " | "
        + enriched_df["scraped_brand"].fillna("")
        + " | "
        + enriched_df["scraped_description"].fillna("")
    )

    refined_rows = [classify_candidate_text(text) for text in combined_text]
    refined_df = pd.DataFrame(refined_rows).add_prefix("refined_")
    enriched_df = pd.concat([enriched_df, refined_df], axis=1)

    enriched_df["final_priority_score"] = (
        0.52 * enriched_df["source_priority_score"].astype(float)
        + 0.25 * enriched_df["metadata_completeness_score"].astype(float)
        + 0.23 * enriched_df["refined_classification_confidence"].astype(float)
    ).round(3)

    enriched_df = enriched_df.sort_values(
        ["final_priority_score", "metadata_completeness_score"],
        ascending=[False, False],
        kind="stable",
    ).reset_index(drop=True)

    return enriched_df


# =============================================================================
# SEED COMPARISON
# =============================================================================


def load_seed_urls() -> set[str]:
    """Load normalized URLs already present in the seed catalog."""
    if SEED_CSV_PATH.exists() is False:
        return set()

    seed_df = pd.read_csv(SEED_CSV_PATH)

    if "source_url" not in seed_df.columns:
        return set()

    return {
        normalize_url(url)
        for url in seed_df["source_url"].dropna().astype(str).tolist()
        if clean_text(url) != ""
    }


def mark_existing_seed_matches(candidate_df: pd.DataFrame) -> pd.DataFrame:
    """Mark candidate URLs that already appear in the seed set."""
    seed_urls = load_seed_urls()
    output_df = candidate_df.copy()
    output_df["already_in_seed_catalog"] = output_df["normalized_url"].isin(
        seed_urls
    )
    return output_df


# =============================================================================
# DISPLAY AND PLOTS
# =============================================================================


def build_summary_table(df: pd.DataFrame) -> pd.DataFrame:
    """Build family summary counts."""
    return (
        df.groupby("transformation_family", observed=False)
        .agg(
            candidate_count=("candidate_id", "count"),
            mean_priority=("source_priority_score", "mean"),
            mean_confidence=("classification_confidence", "mean"),
        )
        .reset_index()
        .assign(
            mean_priority=lambda table: table["mean_priority"].round(3),
            mean_confidence=lambda table: table["mean_confidence"].round(3),
        )
        .query("candidate_count > 0")
    )


def build_category_summary_table(df: pd.DataFrame) -> pd.DataFrame:
    """Build category summary counts."""
    return (
        df.groupby("category_hint", observed=False)
        .agg(
            candidate_count=("candidate_id", "count"),
            mean_priority=("source_priority_score", "mean"),
            mean_confidence=("classification_confidence", "mean"),
        )
        .reset_index()
        .assign(
            mean_priority=lambda table: table["mean_priority"].round(3),
            mean_confidence=lambda table: table["mean_confidence"].round(3),
        )
        .sort_values(
            ["candidate_count", "mean_priority"],
            ascending=[False, False],
            kind="stable",
        )
    )


def display_candidate_tables(candidate_df: pd.DataFrame) -> None:
    """Display candidate discovery results."""
    display_table(
        "Discovery summary by transformation family",
        build_summary_table(candidate_df),
        max_rows=None,
    )

    display_table(
        "Discovery summary by category",
        build_category_summary_table(candidate_df),
        max_rows=80,
    )

    review_columns = [
        "candidate_id",
        "transformation_family",
        "transformation_subtype",
        "source_priority_score",
        "classification_confidence",
        "already_in_seed_catalog",
        "category_hint",
        "title",
        "domain",
        "query",
        "url",
    ]

    display_table(
        "Highest-priority candidate URLs",
        candidate_df[review_columns],
        max_rows=60,
    )


def display_enriched_tables(enriched_df: pd.DataFrame) -> None:
    """Display enriched candidate tables after metadata scraping."""
    review_columns = [
        "candidate_id",
        "final_priority_score",
        "transformation_family",
        "transformation_subtype",
        "refined_transformation_family",
        "refined_transformation_subtype",
        "metadata_completeness_score",
        "fetch_ok",
        "json_ld_product_found",
        "category_hint",
        "scraped_name",
        "scraped_brand",
        "scraped_price",
        "scraped_currency",
        "title",
        "domain",
        "url",
        "error_message",
    ]

    display_table(
        "Enriched candidate review table",
        enriched_df[review_columns],
        max_rows=80,
    )

    scrape_status_df = (
        enriched_df["fetch_ok"]
        .value_counts()
        .rename_axis("fetch_ok")
        .reset_index(name="candidate_count")
    )

    display_table(
        "Scrape status summary",
        scrape_status_df,
        max_rows=None,
    )


def plot_family_counts(df: pd.DataFrame) -> None:
    """Render candidate counts by transformation family."""
    counts = (
        df["transformation_family"]
        .astype(str)
        .value_counts()
        .reindex(FAMILY_ORDER)
        .dropna()
    )
    counts = counts[counts > 0].sort_values()

    fig, ax = plt.subplots(figsize=(9.5, 5.4))
    counts.plot(kind="barh", ax=ax)
    ax.set_title("Discovered Candidate Count by Transformation Family")
    ax.set_xlabel("Candidate count")
    ax.set_ylabel("Transformation family")
    ax.grid(axis="x", alpha=0.25)
    fig.tight_layout()
    plt.show()


def plot_priority_distribution(df: pd.DataFrame) -> None:
    """Render priority-score distribution."""
    fig, ax = plt.subplots(figsize=(8.6, 4.8))
    ax.hist(df["source_priority_score"], bins=18)
    ax.set_title("Candidate Source Priority Distribution")
    ax.set_xlabel("Source priority score")
    ax.set_ylabel("Candidate count")
    ax.grid(axis="y", alpha=0.25)
    fig.tight_layout()
    plt.show()


def plot_top_domains(df: pd.DataFrame) -> None:
    """Render the most common source domains."""
    counts = df["domain"].value_counts().head(24).sort_values()

    fig, ax = plt.subplots(figsize=(10.4, 6.4))
    counts.plot(kind="barh", ax=ax)
    ax.set_title("Top Discovery Domains")
    ax.set_xlabel("Candidate count")
    ax.set_ylabel("Domain")
    ax.grid(axis="x", alpha=0.25)
    fig.tight_layout()
    plt.show()


def plot_top_categories(df: pd.DataFrame) -> None:
    """Render frequent category hints."""
    counts = df["category_hint"].fillna("unknown").value_counts()
    counts = counts.head(24).sort_values()

    fig, ax = plt.subplots(figsize=(10.8, 7.0))
    counts.plot(kind="barh", ax=ax)
    ax.set_title("Top Discovery Categories")
    ax.set_xlabel("Candidate count")
    ax.set_ylabel("Category")
    ax.grid(axis="x", alpha=0.25)
    fig.tight_layout()
    plt.show()


# =============================================================================
# EXPORTS
# =============================================================================


def build_discovery_protocol_markdown() -> str:
    """Build a GitHub documentation page for discovery workflow."""
    return """# Discovery Protocol

This document shows how the repository expands beyond manually curated seed
products.

## Goal

The discovery notebook searches broad product-category queries, brand-directed
queries, and model-aware queries. It captures candidate URLs, classifies each
candidate by physical transformation family, scrapes metadata when allowed, and
exports review tables.

## Candidate status

Rows in `data/discovery_candidates.csv` and
`data/discovery_candidates_enriched.csv` are leads. A lead becomes a canonical
catalog record only after manual review confirms the product exists, the source
is credible, and the transformation mechanism is correctly classified.

## Transformation families

- Foldable products compact through hinges, creases, flexures, flexible display
  laminates, folding frames, swivel-fold headphone structures, or origami-like
  panel networks.
- Collapsible products compact through nesting, telescoping, shell collapse,
  soft-vessel volume evacuation, or frame-assisted collapse.
- Rollable products compact by winding a flexible sheet or membrane.
- Inflatable products compact through deflation.
- Modular products compact through disassembly or flat-pack conversion.
- Hybrid products combine multiple transformation families.

## Review checklist

1. Verify the URL points to a real product, official page, standards page,
   manufacturer page, or high-quality retailer listing.
2. Confirm whether the transformation is foldable, collapsible, rollable,
   inflatable, modular, or hybrid.
3. Record dimensions, mass, folded volume, deployed volume, price, release year,
   certifications, and material stack when available.
4. Promote reviewed rows into `data/products_seed.csv` or a future canonical
   `data/products.csv`.
"""


def export_outputs(
    query_df: pd.DataFrame,
    candidate_df: pd.DataFrame,
    enriched_df: pd.DataFrame | None,
) -> None:
    """Write discovery outputs to repository files."""
    if EXPORT_FILES is False:
        return

    CANDIDATE_CSV_PATH.parent.mkdir(parents=True, exist_ok=True)
    DISCOVERY_DOC_PATH.parent.mkdir(parents=True, exist_ok=True)

    query_df.to_csv(QUERY_LOG_CSV_PATH, index=False)
    candidate_df.to_csv(CANDIDATE_CSV_PATH, index=False)

    if enriched_df is not None:
        enriched_df.to_csv(ENRICHED_CSV_PATH, index=False)

    DISCOVERY_DOC_PATH.write_text(
        build_discovery_protocol_markdown(),
        encoding="utf-8",
    )

    display(Markdown(f"Query log written to `{QUERY_LOG_CSV_PATH}`."))
    display(Markdown(f"Candidate CSV written to `{CANDIDATE_CSV_PATH}`."))

    if enriched_df is not None:
        display(Markdown(f"Enriched CSV written to `{ENRICHED_CSV_PATH}`."))

    display(Markdown(f"Discovery protocol written to `{DISCOVERY_DOC_PATH}`."))


# =============================================================================
# MAIN WORKFLOW
# =============================================================================


def main() -> tuple[pd.DataFrame, pd.DataFrame | None]:
    """Run discovery, classification, optional scraping, plotting, and export."""
    total_started_at = time.perf_counter()

    query_df = build_query_table()
    print(f"Discovery queries loaded: {len(query_df)}")
    display_table("Query table preview", query_df, max_rows=80)

    raw_candidate_df = run_discovery_searches(query_df)
    print(f"Raw unique candidate URLs: {len(raw_candidate_df)}")

    if len(raw_candidate_df) == 0:
        raise RuntimeError("No candidates were discovered. Try fewer queries later.")

    candidate_df = add_candidate_classification(raw_candidate_df)
    candidate_df = rank_candidates(candidate_df)
    candidate_df = mark_existing_seed_matches(candidate_df)

    new_count = int((~candidate_df["already_in_seed_catalog"]).sum())
    print(f"Candidate URLs after deduplication: {len(candidate_df)}")
    print(f"Candidate URLs outside seed catalog: {new_count}")

    display_candidate_tables(candidate_df)
    plot_family_counts(candidate_df)
    plot_priority_distribution(candidate_df)
    plot_top_domains(candidate_df)
    plot_top_categories(candidate_df)

    enriched_df = None
    if SCRAPE_CANDIDATES:
        enriched_df = scrape_candidate_pages(candidate_df)
        display_enriched_tables(enriched_df)

    export_outputs(query_df, candidate_df, enriched_df)

    elapsed_min = (time.perf_counter() - total_started_at) / 60.0
    print(f"Workflow complete in {elapsed_min:.2f} min")

    return candidate_df, enriched_df


discovery_candidates, discovery_enriched = main()
```
